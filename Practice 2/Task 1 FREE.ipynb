### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
p = int(input("Principle amount?"))
r = int(input("Rate of intrest?"))
t = int(input("Time period?"))

si = (p*r*t)/100

print(f"The simple intrest on {p} rupees on {r}% intrest rate for a period of {t} years is {si} rupees")
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
n = int(input("Enter number of terms:"))

sum = 0

for i in range (1,n+1):
    sum += i**2

    print(i**2, end = "")
    if i!=n:
        print("+", end="")


print()
print(sum)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
a = int(input("Enter 1st term:"))
b = int(input("Enter 2nd term:"))
n = int(input("Which term you wanna find:"))

print(f"{n}th term of series is {a+(n-1)*(b-a)}.")
````
