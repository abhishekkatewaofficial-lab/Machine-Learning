````python
# Q46 — Pipeline as a plain dict structure
def make_pipeline():
    steps = []
    def add(func):
        steps.append(func)
    def run(value):
        for i, f in enumerate(steps, 1):
            old = repr(value)
            value = f(value)
            print(f"Step {i} ({f.__name__:<12}): {old} \u2192 {repr(value)}")
        print(f"Final result: {value}")
        return value
    return type('Pipeline', (), {'add': staticmethod(add), 'run': staticmethod(run)})()

p = make_pipeline()
p.add(str.strip)
p.add(str.upper)
p.add(lambda s: s.replace(" ", "_"))
p.run("  hello world  ")
````

````python
# Q50
def min_coins(amount, coins, memo={}):
    if amount == 0: return 0
    if amount in memo: return memo[amount]
    best = float('inf')
    for c in coins:
        if c <= amount:
            result = min_coins(amount - c, coins, memo)
            if result + 1 < best:
                best = result + 1
    memo[amount] = best
    return best
coins = [1, 5, 10, 25]
for amt in [41, 30, 0]:
    print(f"Min coins for {amt}: {min_coins(amt, coins, {})}")
````

# Notebook 9 — Wrap-Up

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| `def` + `return` | Always prefer explicit `return`; without it, the function returns `None` |
| Default args | Evaluated ONCE at definition — never use mutable defaults (use `None` instead) |
| `*args` | Collects extra positional args as a **tuple** inside the function |
| `**kwargs` | Collects extra keyword args as a **dict** inside the function |
| `global` / `nonlocal` | `global` for module-level; `nonlocal` for enclosing function scope |
| Lambda | Anonymous one-liner function — best used inline, not for complex logic |
| Closure | Inner function captures variables from outer function scope |
| Higher-order function | Functions that accept or return other functions |
| Recursion | Always define base case first; every recursive call should reduce the problem |
| Memoization | Store already-computed results in a dict to avoid redundant recursion |

## Common Mistakes
1. **Mutable default argument**: `def f(lst=[])` — the list is shared across all calls. Use `None`.
2. **Forgetting `return`**: A function that prints but doesn't return gives `None`.
3. **Missing base case in recursion**: Causes infinite recursion and `RecursionError`.
4. **Using `global` too much**: Prefer returning values and passing them as args instead.
5. **`*args` vs `*` (unpacking)**: `def f(*args)` collects; `f(*lst)` unpacks when calling.
