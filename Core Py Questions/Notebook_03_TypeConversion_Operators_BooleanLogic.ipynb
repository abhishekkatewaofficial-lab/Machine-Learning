### Question 33
**Difficulty:** Medium &nbsp;&nbsp; **Topic:** De Morgan's Laws

Verify De Morgan's Laws in Python for `a = True` and `b = False`:
1. `not (a and b)` == `(not a) or (not b)`

**Expected Output:**
```
True
True
All 4 combinations verified: True
```


````python
a, b = True, False
print(not (a and b) == (not a) or (not b))
print(not (a or b) == ((not a) and (not b)))
pairs = [(True, True), (True, False), (False, True), (False, False)]
law1 = all(not (a and b) == ((not a) or (not b))  for a, b in pairs)
law2 = all(not (a or b)  == ((not a) and (not b)) for a, b in pairs)
print(f"All 4 combinations verified: {law1 and law2}")
````

# Notebook 3 — Wrap-Up

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| `int()` truncates | Always towards zero — NOT rounding |
| `and` / `or` return values | Not just booleans — they return one of their operands |
| Short-circuit evaluation | Right side is skipped if left side determines result |
| `not` precedence | `not` > `and` > `or` — memorize this |
| `isinstance()` > `type()` | Use `isinstance()` — it handles inheritance correctly |
| Bitwise `&` for masking | `n & (n-1)` clears the lowest set bit — used in power-of-2 checks |
| `bool` is subclass of `int` | `True == 1`, `False == 0`, `isinstance(True, int)` is `True` |
| `or`-default pattern risk | Breaks when valid values are falsy (0, "", False) |
| NaN is never equal to itself | `float('nan') == float('nan')` is always `False` |

---

## Common Mistakes
1. **Using `or` for defaults when 0 or `""` are valid values** — use `if x is None` instead.
2. **Assuming `and`/`or` always return `True`/`False`** — they return operands.
3. **Forgetting `not` has higher precedence than `and`** — `not a and b` is `(not a) and b`.
4. **Using `type(x) == int` instead of `isinstance(x, int)`** — misses subclasses.
5. **Not handling `float('nan')` equality** — always use `math.isnan()` to check for NaN.
