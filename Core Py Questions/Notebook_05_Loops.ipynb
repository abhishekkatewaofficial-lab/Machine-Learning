````python
# Q49
A = [[1,2,3],[4,5,6],[7,8,9]]
B = [[9,8,7],[6,5,4],[3,2,1]]
n = 3

def print_matrix(m, title):
    print(title)
    for row in m:
        print(*row)
    print()

sum_m  = [[A[i][j] + B[i][j] for j in range(n)] for i in range(n)]
prod_m = [[A[i][j] * B[i][j] for j in range(n)] for i in range(n)]
trans  = [[A[j][i]            for j in range(n)] for i in range(n)]
trace  = sum(A[i][i] for i in range(n))

print_matrix(sum_m,  "Sum:")
print_matrix(prod_m, "Element-wise Product:")
print_matrix(trans,  "Transpose of A:")
print(f"Trace of A: {trace}")
````

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| `for` vs `while` | Use `for` when you know the count; `while` when you loop until a condition |
| `break` | Exits the loop immediately |
| `continue` | Skips the rest of the current iteration |
| `pass` | Does nothing — placeholder |
| `for-else` / `while-else` | `else` runs only if loop exits normally (no `break`) |
| `enumerate()` | Gives index + value together — avoids manual counters |
| `zip()` | Pairs up multiple iterables — stops at the shortest |
| Loop variable scope | Loop variable lives after the loop ends in Python |
| Accumulator pattern | Initialize before loop, update inside, use after |
| Nested loop break | Need a flag or exception to break outer loop from inner |

---

## Common Mistakes
1. **Off-by-one in `range()`** — `range(1, n)` excludes `n`; use `range(1, n+1)` to include.
2. **Modifying a list while iterating over it** — always iterate over a copy or collect changes separately.
3. **Forgetting `else` only runs on natural loop exit** — if you `break`, `else` is skipped.
4. **Using `zip()` with unequal lists** — data can be silently lost. Check lengths first or use `zip_longest`.
5. **Infinite `while` loops** — always ensure the condition can become `False`.
