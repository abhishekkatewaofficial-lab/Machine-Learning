### Question 8
**Difficulty:** Easy &nbsp;&nbsp; **Topic:** Leap year check

A year is a leap year if:
- Divisible by 4 **AND** (not divisible by 100 **OR** divisible by 400)

Check and print for years: `2000`, `1900`, `2024`, `2023`.

**Expected Output:**
```
2000: Leap Year
1900: Not a Leap Year
2024: Leap Year
2023: Not a Leap Year
```


````python
for year in [2000, 1900, 2024, 2023]:
    if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
        print(f"{year}: Leap Year")
    else:
        print(f"{year}: Not a Leap Year")
````

````python
# Q50 — vending machine state machine
events = [
    ("IDLE",       "select",  "cola"),
    ("SELECTED",   "pay",     50),
    ("PAID",       "confirm", None),
    ("DISPENSING", "collect", None),
    ("IDLE",       "pay",     50),
    ("SELECTED",   "cancel",  None),
]

for state, action, value in events:
    val_str = f" {value}" if value is not None else ""
    match (state, action):
        case ("IDLE", "select"):
            print(f"[IDLE] select {value} \u2192 Item selected. Next: SELECTED")
        case ("SELECTED", "pay"):
            print(f"[SELECTED] pay {value} \u2192 Payment received. Next: PAID")
        case ("PAID", "confirm"):
            print(f"[PAID] confirm \u2192 Dispensing item. Next: DISPENSING")
        case ("DISPENSING", "collect"):
            print(f"[DISPENSING] collect \u2192 Item collected. Next: IDLE")
        case ("SELECTED", "cancel"):
            print(f"[SELECTED] cancel \u2192 Cancelled. Next: IDLE")
        case _:
            print(f"[{state}] {action}{val_str} \u2192 Invalid action in current state.")
````

# Notebook 4 — Wrap-Up

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| `elif` order matters | Always check most specific/highest condition first |
| Nested `if` → guard clauses | Flip nested ifs to fail-fast style for cleaner code |
| `match-case` basics | Cleaner than `if-elif` for many-branch equality checks |
| `match-case` with `\|` | Combine multiple patterns in one case |
| `match-case` with guards | Add `if condition` after the pattern for extra filtering |
| Sequence patterns in `match` | `case ["go", direction]` captures list elements automatically |
| Dict as branch table | Replacing `if-elif` chains with a dict lookup is often cleaner |
| `next()` with generator | Finds first matching item — avoids if-elif for lookup tables |
| bool-first in isinstance | Always check `bool` before `int` since `bool` subclasses `int` |

---

## Common Mistakes
1. **Wrong `elif` order** — checking lower thresholds before higher ones.
2. **Deep nesting instead of guard clauses** — hard to read, hard to debug.
3. **Forgetting `_` wildcard in `match-case`** — unmatched values cause silent pass-throughs.
4. **`isinstance(val, int)` before `isinstance(val, bool)`** — True/False incorrectly classified as Integer.
5. **Using `or` chains instead of `in` for membership** — `if x == 1 or x == 2` → use `if x in [1, 2]`.
