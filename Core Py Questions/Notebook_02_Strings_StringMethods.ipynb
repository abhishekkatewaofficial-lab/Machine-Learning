# Notebook 2 — Wrap-Up

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| Strings are immutable | You can never change a character in-place; always build a new string |
| `.split()` default | Splits on any whitespace and discards empty strings — great for cleanup |
| `.join()` | The Pythonic way to build strings from a list — faster than `+=` in loops |
| Slicing `[::-1]` | The idiomatic way to reverse a string |
| f-string specs | `{val:>10.2f}`, `{val:,}`, `{val:.2e}` — powerful and concise |
| `ord()` / `chr()` | Bridge between characters and their integer codes |
| String interning | CPython interns simple literals; dynamically built strings are not interned |
| Lexicographic comparison | Strings compare character by character by ASCII value |

---

## Common Mistakes
1. **Using `+=` to build strings in a loop** — use `"".join(list)` instead.
2. **Forgetting `.strip()` after `input()`** — user input often has trailing newlines.
3. **Assuming `.find()` raises an error** — it returns `-1`; `.index()` raises `ValueError`.
4. **Using `is` to compare string values** — always use `==`.
5. **Forgetting string comparison is lexicographic** — `"10" < "9"` is `True`.
