
- `ref` is also changed (reference)
- `cpy` is NOT changed (copy)

# Notebook 6 — Wrap-Up

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| `append` vs `extend` | `append` nests; `extend` unpacks and adds items individually |
| `remove` vs `pop` vs `del` | By value / by index with return / by index without return |
| `sort()` vs `sorted()` | In-place returns None; `sorted()` returns a new list |
| Reference trap | `b = a` shares the same object — use `a.copy()` or `a[:]` |
| Slice assignment | `lst[1:3] = [x, y, z]` replaces a range — even different lengths |
| Star unpacking | `first, *rest = lst` — grabs one and the rest cleanly |
| `zip(*matrix)` | Transpose a matrix using zip and unpacking |
| Two pointers | Efficient pattern for palindrome check, merge, partition |
| Stack with list | `append()` to push, `pop()` to pop, `[-1]` to peek |
| Merge intervals | Sort + greedy merge — a must-know pattern for interviews |

---

## Common Mistakes
1. `b = a` does NOT copy — both point to the same list.
2. Modifying a list while iterating over it — leads to bugs.
3. `sort()` returns `None` — `result = lst.sort()` gives `None`.
4. `remove(x)` removes only the **first** occurrence, not all.
5. `append([1,2])` nests the list — use `extend([1,2])` to add elements.
