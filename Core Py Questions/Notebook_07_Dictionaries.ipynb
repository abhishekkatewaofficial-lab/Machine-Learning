````python
# Q43
roman_map = {"I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000}
for s in ["III","IV","IX","LVIII","MCMXCIV"]:
    total = 0
    for i in range(len(s)):
        if i + 1 < len(s) and roman_map[s[i]] < roman_map[s[i+1]]:
            total -= roman_map[s[i]]
        else:
            total += roman_map[s[i]]
    print(f"{s} \u2192 {total}")
````

````python
# Q44
for s in ["abcabcbb", "bbbbb", "pwwkew"]:
    last_seen = {}
    start = max_len = 0
    for i, ch in enumerate(s):
        if ch in last_seen and last_seen[ch] >= start:
            start = last_seen[ch] + 1
        last_seen[ch] = i
        max_len = max(max_len, i - start + 1)
    print(f"{s} \u2192 {max_len}")
````

````python
# Q49
sentences = ["python is great","python is easy","java is great"]
matrix = {}
for sent in sentences:
    words = list(dict.fromkeys(sent.split()))  # unique words in sentence
    for i in range(len(words)):
        for j in range(len(words)):
            if i != j:
                matrix.setdefault(words[i], {})
                matrix[words[i]][words[j]] = matrix[words[i]].get(words[j], 0) + 1

print("python co-occurs with:")
for word, cnt in sorted(matrix.get("python", {}).items(), key=lambda x: -x[1]):
    print(f"  {word}: {cnt}")
````

````python
# Q50
commands = [
    "ADD apple 50 10","ADD banana 30 5","SELL apple 15",
    "RESTOCK banana 20","SELL cherry 5","SELL banana 60",
    "ADD cherry 100 8","REPORT",
]
inventory = {}  # {item: {qty, price}}
for cmd in commands:
    parts = cmd.split()
    action = parts[0]
    if action == "ADD":
        _, item, qty, price = parts
        inventory[item] = {"qty": int(qty), "price": int(price)}
        print(f"Added {item}: qty={qty}, price=\u20b9{price}")
    elif action == "SELL":
        _, item, qty = parts
        qty = int(qty)
        if item not in inventory:
            print(f"Error: {item} not in inventory")
        elif inventory[item]["qty"] < qty:
            print(f"Error: only {inventory[item]['qty']} {item} in stock")
        else:
            inventory[item]["qty"] -= qty
            print(f"Sold {qty} {item}. Remaining: {inventory[item]['qty']}")
    elif action == "RESTOCK":
        _, item, qty = parts
        qty = int(qty)
        inventory[item]["qty"] += qty
        print(f"Restocked {item} by {qty}. Total: {inventory[item]['qty']}")
    elif action == "REPORT":
        print("--- INVENTORY REPORT ---")
        total_val = 0
        for item, info in sorted(inventory.items()):
            val = info["qty"] * info["price"]
            total_val += val
            print(f"{item:<7}| qty={info['qty']:<4}| price=\u20b9{info['price']:<3}| value=\u20b9{val}")
        print(f"Total inventory value: \u20b9{total_val}")
````

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| `d[key]` vs `.get()` | `[]` raises `KeyError`; `.get()` returns a default safely |
| `.setdefault()` | Sets a key only if it doesn't exist yet — perfect for grouping |
| `.update()` / `\|` | Merge dicts — right side wins for duplicate keys |
| `.pop()` vs `del` | `pop` returns the value; `del` doesn't |
| `.popitem()` | Removes and returns the **last inserted** pair (Python 3.7+) |
| Reference trap | `d2 = d1` shares the object — use `.copy()` for independence |
| Dict comprehension | `{k: v for k, v in ...}` — clean, fast, Pythonic |
| Freq counter | `d.get(k, 0) + 1` — the most common dict pattern |
| Grouping | `d.setdefault(key, []).append(val)` — group items by property |
| Two-sum pattern | Dict for O(n) lookup — store `{value: index}` as you iterate |

---

## Common Mistakes
1. **Modifying a dict while iterating** — use dict comprehension or iterate over a copy.
2. **`d[key]` on missing key** — use `.get(key, default)` to avoid `KeyError`.
3. **`d = other_dict` is NOT a copy** — use `.copy()` or `dict(other_dict)`.
4. **`d.keys()` returns a view, not a list** — wrap with `list()` when needed.
5. **`d.sort()`** — dicts have no `.sort()` method. Use `sorted(d.items(), key=...)`.
