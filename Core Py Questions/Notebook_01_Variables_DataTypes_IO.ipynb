| Concept | One-liner Summary |
|---|---|
| Variables | Labels that point to objects in memory |

## Common Mistakes
1. **Forgetting to convert `input()`** — `int(input(...))` not just `input(...)` for numbers.
2. **Confusing `=` (assignment) with `==` (comparison)**.
3. **Using `is` to compare values** — use `==` for equality checks.
4. **Expecting `int(3.9)` to round** — it truncates towards zero.
5. **Assuming `bool("False")` is `False`** — any non-empty string is truthy.
