````python
# Q47
docs = {
    "doc1": "python is great and easy to learn",
    "doc2": "python is powerful and fast to use",
    "doc3": "java is great but verbose and slow",
}
doc_sets = {name: set(text.split()) for name, text in docs.items()}
names = list(doc_sets.keys())
best_pair, best_score = "", 0
for i in range(len(names)):
    for j in range(i+1, len(names)):
        a, b = names[i], names[j]
        score = len(doc_sets[a] & doc_sets[b]) / len(doc_sets[a] | doc_sets[b])
        print(f"{a} vs {b}: {score:.2f}")
        if score > best_score:
            best_score, best_pair = score, f"{a} & {b}"
print(f"Most similar: {best_pair} ({best_score:.2f})")
````

````python
# Q48
arrivals   = sorted([900, 940, 950, 1100, 1500, 1800])
departures = sorted([910, 1200, 1120, 1130, 1900, 2000])
platforms = 0
max_platforms = 0
i = j = 0
while i < len(arrivals):
    if arrivals[i] <= departures[j]:
        platforms += 1
        max_platforms = max(max_platforms, platforms)
        i += 1
    else:
        platforms -= 1
        j += 1
print(f"Minimum platforms needed: {max_platforms}")
````

````python
# Q49
recipes = {
    "pasta":    {"flour","egg","water","salt"},
    "omelette": {"egg","butter","salt","pepper"},
    "cake":     {"flour","egg","butter","sugar","salt"},
    "salad":    {"lettuce","tomato","cucumber","salt"},
}
# Most common ingredient
freq = {}
for ings in recipes.values():
    for ing in ings:
        freq[ing] = freq.get(ing, 0) + 1
most = max(freq, key=lambda k: freq[k])
print(f"Most common ingredient: {most} (in {freq[most]} recipes)")
# Makeable
available = {"egg","flour","water","salt","butter","pepper"}
makeable  = [name for name, ings in recipes.items() if ings.issubset(available)]
print(f"Makeable recipes: {makeable}")
# Pasta ∩ Cake
print(f"Pasta \u2229 Cake: {recipes['pasta'] & recipes['cake']}")
# Unique to omelette
others = set()
for name, ings in recipes.items():
    if name != "omelette": others |= ings
print(f"Unique to omelette: {recipes['omelette'] - others}")
````

````python
# Q50
network = {
    "Alice": {"Bob","Carol","Dave"},
    "Bob":   {"Alice","Carol","Eve"},
    "Carol": {"Alice","Bob","Dave","Eve"},
    "Dave":  {"Alice","Carol","Frank"},
    "Eve":   {"Bob","Carol"},
    "Frank": {"Dave"},
}
# Mutual friends of Alice & Bob
print(f"Alice & Bob mutual friends: {network['Alice'] & network['Bob']}")
# 2nd degree of Alice
fof = set()
for friend in network["Alice"]:
    fof |= network[friend]
second_degree = fof - network["Alice"] - {"Alice"}
print(f"Alice's 2nd-degree (not direct): {second_degree}")
# Most connected
most = max(network, key=lambda k: len(network[k]))
print(f"Most connected: {most} ({len(network[most])} friends)")
# Friends of everyone
all_people = set(network.keys())
friends_of_all = {p for p in all_people if network[p] == all_people - {p}}
print(f"Friends of everyone: {friends_of_all}")
````

# Notebook 8 — Wrap-Up

## Key Takeaways
| Concept | One-liner Summary |
|---|---|
| `{}` vs `set()` | `{}` creates a dict — always use `set()` for empty sets |
| `add()` vs `update()` | `add` takes one hashable item; `update` takes any iterable |
| `remove()` vs `discard()` | `remove` raises `KeyError` if missing; `discard` is silent |
| `\|`, `&`, `-`, `^` | Union, intersection, difference, symmetric difference |
| In-place: `\|=`, `&=`, `-=`, `^=` | Modify the set in place (same as the above but mutating) |
| `issubset()`/`issuperset()` | Check containment; a set is always a subset of itself |
| `isdisjoint()` | True if two sets share NO elements |
| `frozenset` | Immutable set — hashable, can be used as dict key |
| Dedup preserving order | `seen = set()` + loop — O(n) with set membership O(1) |
| Sets vs lists membership | `in` on a set is O(1); on a list is O(n) — always prefer set for lookups |

---

## Common Mistakes
1. `{}` creates a **dict** — use `set()` for an empty set.
2. Sets are **unordered** — no indexing (`s[0]` raises `TypeError`).
3. Sets can only contain **hashable** items — no lists or dicts as elements.
4. `remove(x)` raises `KeyError` on missing key — use `discard(x)` for safe removal.
5. Converting to a set and back **loses order** — use `seen = set()` + loop to deduplicate while preserving order.
