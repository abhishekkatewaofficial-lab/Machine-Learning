### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
ctc = float(input("Enter CTC (in lakhs): "))

if ctc < 5:
    tax = 0

elif ctc < 10:
    tax = ctc*.10

elif ctc < 20:
    tax = ctc*.20

else:
    tax = ctc*.30

in_hand = (ctc - .1*ctc - .05*ctc - .03*ctc - tax)/12

print(f"Monthly In-hand Salary: {in_hand:.2f} Lakhs")
````

### `Problem 3`: Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit.

````python
cp = int(input("Cost Price: "))
sp = int(input("Selling Price: "))

profit_loss = sp - cp

if profit_loss > 0:
    print(f"Profit = ₹{profit_loss}")
elif profit_loss < 0:
    print(f"Loss = ₹{abs(profit_loss)}")
else:
    print("No Profit, No Loss")
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
number = int(input("Kiska Factorial Nikalna Hai?"))
fact = 1

for num in range(1,number+1):
    fact *= num

print(fact)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
n = int(input("Enter N: "))

i = 1
total = 0

while i <= n:
    if i % 5 == 0:
        i += 1
        continue

    total += i

    if total > 300:
        break

    i += 1

print(total)
````

### `Problem 9`: Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

````python
sum = 0
count = 0

while True:

    num = int(input("Enter a Number:"))

    if num == 0:

        break

    sum += num
    count += 1

print(f"Total Sum is {sum} and Average is {sum/count if count else 0} ")
````

###`Problem 9`: Write  program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

````python
list = []

for num in range (2000,3201):
    if num%7 == 0 and num%5 != 0:
        list.append(str(num))

print(",".join(list))
````

###`Problem 10`: Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
nums = []

for num in range(1000, 3001):
    valid = True

    for digit in str(num):
        if int(digit) % 2 != 0:
            valid = False
            break

    if valid:
        nums.append(str(num))

print(" ".join(nums))
````

# Pattern: "Check if ALL elements satisfy a condition"

Use this pattern whenever you need to verify that **every** item in a collection meets a condition.

## Template

```python
valid = True

for item in collection:
    if condition_fails:
        valid = False
        break

if valid:
    # Do something
```

## Why use `valid`?

* Assume the answer is **True** initially.
* If even **one** element fails the condition:

  * Set `valid = False`
  * Stop checking using `break`
* After the loop:

  * If `valid` is still `True`, then **all** elements passed.

###`Problem 12`:Write a program to print whether a given number is a prime number or not

````python
number = int(input("Enter a number:"))

if number < 2:
    print(f"{number} is not a prime number" )

else:
    is_prime = True

    for i in range(2,number):
        if number % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{number} is a Prime Number")

    else:
        print(f"{number} is a non-Prime Number")

````
