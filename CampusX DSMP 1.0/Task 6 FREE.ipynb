### **`Problem 3:`** Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
Sample_String = 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'

upper  = 0
lower  = 0

for char in Sample_String:
    if char.isupper():
        upper+=1
    if char.islower():
        lower+=1
print(f"No. of Upper case characters : {upper}")
print(f"No. of Upper case characters : {lower}")


````

`Problem-7` Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
text = "hello how are you i am fine thank you"

words = text.split()

freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

max_word = ""
max_count = 0

for key, value in freq.items():
    if value > max_count:
        max_count = value
        max_word = key

print(max_word, "->", max_count)


print('='*8)

from collections import Counter

text = "hello how are you i am fine thank you"

freq = Counter(text.split())

word, count = freq.most_common(1)[0]

print(word, "->", count)
````

`Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python

nums = [13, 42, 15, 37, 22, 39, 41, 50]

hist = {}

for num in nums:
    start = (num // 10) * 10 + 1
    end = start + 9

    key = f"{start}-{end}"
    hist[key] = hist.get(key, 0) + 1


for key in sorted(hist):
    print(f"{key} : {hist[key]}")
````

`Problem-9` Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.

```
List of Coordinates
[(1,1),(2,2),(3,3),(4,4)]
Query Point
(0,0)
```

```
Output
Nearest to (0,0) is (1,1)
```

````python
import math

points = [(1, 1), (2, 2), (3, 3), (4, 4)]
query  = (0, 0)

nearest          = None
minimum_distance = float("inf")

for point in points:
    distance = math.dist(point,query)

    if distance<minimum_distance:
        minimum_distance=distance
        nearest         =point

print(f"Nearest to {query} is {nearest}")
````

`Problem 15`- A dictionary contains following information about 5 employees:
- First name
- Last name
- Age
- Grade(Skilled,Semi-skilled,Highly skilled)<br>
Write a program using map/filter/reduce to a list of employees(first name + last name) who are highly skilled

````python
# Write code here
employees = [
    {
        'fname':'Nitish',
        'lname':'Singh',
        'age' : 33,
        'grade':'skilled'
    },
    {
        'fname':'Ankit',
        'lname':'Verma',
        'age' : 34,
        'grade':'semi-skilled'
    },
    {
        'fname':'Neha',
        'lname':'Singh',
        'age' : 35,
        'grade':'highly-skilled'
    },
    {
        'fname':'Anurag',
        'lname':'Kumar',
        'age' : 30,
        'grade':'skilled'
    },
    {
        'fname':'Abhinav',
        'lname':'Sharma',
        'age' : 37,
        'grade':'highly-skilled'
    }

]
````

````python
highly_skilled = list(filter(lambda emp: emp["grade"] == "highly-skilled", employees))
print(highly_skilled)

result = list(map(lambda emp: emp["fname"] + " " + emp["lname"], highly_skilled))

print(result)

````
