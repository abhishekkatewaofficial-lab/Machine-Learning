### `Q1`: Key with maximum unique values

Given a dictionary with values list, extract key whose value has most unique values.

**Example 1:**

Input:

```bash
test_dict = {"CampusX" : [5, 7, 9, 4, 0], "is" : [6, 7, 4, 3, 3], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
CampusX
```

**Example 2:**

Input:
```bash
test_dict = {"CampusX" : [5, 7, 7, 7, 7], "is" : [6, 7, 7, 7], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
Best
```

````python

def biggest(dict):
    result = []

    for word,list in dict.items():
        result.append([len(set(list)) , word])

    result.sort()
    return result[-1][1]

answer1  = biggest(dict = {
    "CampusX": [5, 7, 9, 4, 0],
    "is": [6, 7, 4, 3, 3],
    "Best": [9, 9, 6, 5, 5]
})

answer2 = biggest(dict = 
    {"CampusX" : [5, 7, 7, 7, 7], 
    "is" : [6, 7, 7, 7],
    "Best" : [9, 9, 6, 5, 5]
})

print(answer1)
print(answer2)
print("="*15)
#####################################
def biggest(data):
    max_count = 0
    answer = ""

    for word, values in data.items():
        unique_count = len(set(values))

        if unique_count > max_count:
            max_count = unique_count
            answer = word

    return answer

answer1  = biggest(data = {
    "CampusX": [5, 7, 9, 4, 0],
    "is": [6, 7, 4, 3, 3],
    "Best": [9, 9, 6, 5, 5]
})

answer2 = biggest(data = 
    {"CampusX" : [5, 7, 7, 7, 7], 
    "is" : [6, 7, 7, 7],
    "Best" : [9, 9, 6, 5, 5]
})

print(answer1)
print(answer2)
````

### `Q2`: Replace words from Dictionary. Given String, replace it’s words from lookup dictionary.

**Example 1:**

Input:

```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}
```

Output:

```bash
CampusX is the best channel for Data-Science students.
```

**Example 2:**

Input:
```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"good" : "is the best channel", "ds" : "Data-Science"}
```

Output:
```bash
CampusX best for DS students.
```

````python
test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}

words = test_str.split()

result = []

for word in words:
    if word in repl_dict:
        result.append(repl_dict[word])
    else:
        result.append(word)

print(" ".join(result))
````

### `Q3`: Convert List to List of dictionaries. Given list values and keys list, convert these values to key value pairs in form of list of dictionaries.

**Example 1:**

Input:
```bash
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'DataScience', 'id': 3}, {'name': 'is', 'id': 8}]
```

**Example 2:**

Input:
```bash
test_list = ["CampusX", 10]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'CampusX', 'id': 10}]
```

````python
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]

result = []

for i in range(0, len(test_list), 2):
    temp = {}

    temp[key_list[0]] = test_list[i]
    temp[key_list[1]] = test_list[i + 1]

    result.append(temp)

print(result)
````

### `Q4`: Convert a list of Tuples into Dictionary.

Input:

```bash
[('A', 1), ('B', 2), ('C', 3)]
```

### `Q5`: Sort Dictionary key and values List.


Output:

```bash
{'a': [4, 19], 'b': [10, 12], 'c': [3]}
```

**Example 2:**

Input:

```bash
{'c': [10, 34, 3]}
```

Output:

```bash
{'c': [3, 10, 34]}
```
