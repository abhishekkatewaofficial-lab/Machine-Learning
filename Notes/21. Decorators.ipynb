# Step 6 : functools.wraps

Python provides

```
functools.wraps
```

to copy metadata from the Base Function to the Enhanced Function.

````python
from functools import wraps


def decorator(base_function):

    @wraps(base_function)
    def enhanced_function(*args, **kwargs):

        print("🚀 Starting")

        result = base_function(*args, **kwargs)

        print("🏁 Finished")

        return result

    return enhanced_function

@decorator
def add(a, b):

    """Adds two numbers"""

    return a + b


print(add.__name__)

print(add.__doc__)
````

# Final Production Template

```python

        # Before

        result = base_function(*args, **kwargs)

        # After

        return result

    return enhanced_function
```

Memorizing this template is not important.

Understanding **why each line exists** is.

# Google Interview Answer

If an interviewer asks:

"What is a decorator?"

A complete answer would be:

> A decorator is a higher-order function that receives a Base Function, creates an Enhanced Function, adds behavior before or after the Base Function, and returns the Enhanced Function.

It works because Python supports:

1. Functions as first-class objects
2. Higher-order functions
3. Closures
4. The `@` syntax as syntactic sugar
5. `*args` and `**kwargs` for generic decorators
6. `functools.wraps` to preserve metadata

````python
import time


def timer_decorator(base_function):

    def enhanced_function(*args, **kwargs):

        print("🚀 Function Started")

        start = time.perf_counter()

        result = base_function(*args, **kwargs)

        end = time.perf_counter()

        print("🏁 Function Finished")
        print(f"⏱ Execution Time: {end - start:.6f} seconds")

        return result

    return enhanced_function


@timer_decorator
def calculate_sum(a, b):

    time.sleep(2)

    return a + b


answer = calculate_sum(10, 20)

print(f"Result: {answer}")
````

````python
import time


def production_decorator(base_function):

    def enhanced_function(*args, **kwargs):

        print("=" * 50)
        print(f"🚀 Starting: {base_function.__name__}")

        start_time = time.perf_counter()

        try:
            result = base_function(*args, **kwargs)

            print("✅ Function Executed Successfully")

            return result

        except Exception as error:

            print(f"❌ Error: {error}")

        finally:

            end_time = time.perf_counter()

            print(f"⏱ Execution Time: {end_time - start_time:.6f} seconds")
            print("=" * 50)

    return enhanced_function


@production_decorator
def process_payment(customer_name, amount):

    print(f"💳 Processing payment of ₹{amount} for {customer_name}")

    time.sleep(2)

    return "Payment Successful"


@production_decorator
def divide(a, b):

    return a / b


@production_decorator
def greet(name, age=20):

    print(f"Hello {name}, Age: {age}")


payment_status = process_payment("Abhi", 2500)
print(payment_status)

print()

print(divide(20, 5))

print()

divide(20, 0)

print()

greet("Rahul", age=22)

print()

# Without @wraps, metadata belongs to the enhanced function
print(process_payment.__name__)
print(process_payment.__doc__)
````
