# Python Generators

## What is a Generator?

A Generator is a special type of function that returns values one at a time instead of returning all values at once.

Generators use the `yield` keyword.

### Why use Generators?

- Memory Efficient
- Faster for large datasets
- Values are generated only when needed (Lazy Evaluation)

### Normal Function vs Generator

| Normal Function | Generator |
|---------------|------------|
| Uses `return` | Uses `yield` |
| Returns entire result at once | Returns one value at a time |
| More memory usage | Less memory usage |
| Execution ends after return | Execution pauses after yield |

```python
# yield = pause + save state + return value
```

```python
# Normal Function

def square_numbers(n):
    result = []

    for i in range(n):
        result.append(i * i)

    return result

print(square_numbers(5))
```

## Generator Example

Instead of storing all values in memory, a generator produces values one by one using `yield`.

```python
def square_generator(n):
    for i in range(n):
        yield i * i

gen = square_generator(5)

print(gen)
```

## Getting Values Using next()

The `next()` function is used to get the next value from a generator.

```python
def square_generator(n):
    for i in range(n):
        yield i * i

gen = square_generator(5)

print(next(gen))
print(next(gen))
print(next(gen))
```

## Generator Execution Flow

When a generator encounters `yield`:

1. It returns the current value.
2. Saves its current state.
3. Pauses execution.
4. Continues from the same place on the next call.

```python
def demo():
    print("Step 1")
    yield 10

    print("Step 2")
    yield 20

    print("Step 3")
    yield 30

g = demo()

print(next(g))
print(next(g))
print(next(g))
```

## Using Generator with for Loop

The most common way to use generators is with a for loop.

```python
def count_up_to(n):
    for i in range(1, n + 1):
        yield i

for num in count_up_to(5):
    print(num)
```

## Generator Expression

Generator expressions are similar to list comprehensions but use parentheses `()`.

```python
gen = (x * x for x in range(5))

print(next(gen))
print(next(gen))
print(next(gen))
```

## List Comprehension vs Generator Expression

```python
import sys

lst = [x*x for x in range(1000)]
gen = (x*x for x in range(1000))

print(sys.getsizeof(lst))
print(sys.getsizeof(gen))
```

## Infinite Generator

Generators can generate infinite sequences because values are produced only when needed.

```python
def infinite_numbers():
    num = 1

    while True:
        yield num
        num += 1

gen = infinite_numbers()

for _ in range(5):
    print(next(gen))
```
