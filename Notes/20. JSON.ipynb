# JSON (JavaScript Object Notation)

## What is JSON?

JSON (JavaScript Object Notation) is a lightweight data-interchange format used to exchange data between applications.

Although JSON originated from JavaScript, today it is language-independent and supported by almost every programming language.

Examples:
- Python ↔ API
- Browser ↔ Server
- Mobile App ↔ Backend
- Database ↔ Application

JSON is one of the most widely used formats for data exchange.

# Why Do We Need JSON?

How will they communicate?

That common language is JSON.

# JSON Syntax Rules

1. Data is stored as key-value pairs.
2. Keys must always be strings.
3. Strings use double quotes.
4. Objects use {}
5. Arrays use []
6. Values can be:
   - String
   - Number
   - Boolean
   - null
   - Object
   - Array

````python
person = {
    "name": "Abhi",
    "age": 20,
    "student": True
}

print(person)
````

Although this looks like JSON,

it is actually a Python dictionary.

# Python JSON Module

Python provides a built-in module named `json`.

It helps us:

1. Convert Python objects → JSON
2. Convert JSON → Python objects
3. Read JSON files
4. Write JSON files

Import it using:


````python
import json
````

# Why Do We Need json Module?

Suppose we have a Python dictionary.

```python
student = {
    "name": "Abhi",
    "age": 20
}
```

A server cannot understand Python dictionaries directly.

It understands JSON.

So we convert:

Python Dictionary

↓

JSON String

using

```python
json.dumps()
```

# json.dumps()

`dumps()` means

**Dump String**

It converts a Python object into a JSON string.

Syntax

```python
json.dumps(python_object)
```

````python
import json

student = {
    "name": "Abhi",
    "age": 20,
    "cgpa": 8.7
}

json_data = json.dumps(student)

print(json_data)
print(type(student))
print(type(json_data))
````

# Interview Observation

Notice the type.

Even though it looks like a dictionary,

it is actually a **string**.

Always verify using:

```python
type(json_data)
```

````python
student = {
    "name": "Abhi",
    "skills": ["Python", "SQL"]
}

json_data = json.dumps(student)

print(type(student))
print(type(json_data))
````

# json.loads()

`loads()` means

**Load String**

It converts a JSON string into a Python object.

Syntax

```python
json.loads(json_string)
```

````python
import json

json_data = '{"name":"Abhi","age":20}'

python_data = json.loads(json_data)

print(python_data)
print(type(python_data))
````

# Flow

Python Dictionary

↓

json.dumps()

↓

JSON String

↓

json.loads()

↓

Python Dictionary

````python
import json

employee = {
    "id": 101,
    "name": "Abhi"
}

json_string = json.dumps(employee)

python_dict = json.loads(json_string)

print(type(employee))
print(type(json_string))
print(type(python_dict))
````

# Pretty Printing JSON

Large JSON data becomes difficult to read.

Use

```python
indent
```

to format it.

````python
import json

student = {
    "name": "Abhi",
    "age": 20,
    "skills": ["Python", "SQL", "Pandas"]
}

print(json.dumps(student, indent=4))
````

# sort_keys=True

This prints keys in alphabetical order.

Useful while debugging.

````python
import json

student = {
    "city": "Pune",
    "name": "Abhi",
    "age": 20
}

print(json.dumps(student, indent=4, sort_keys=True))
````

# Converting Lists

# Nested Objects

The `json` module automatically handles nested dictionaries and lists.

````python
import json

employee = {

    "name": "Abhi",

    "address": {

        "city": "Pune",

        "state": "MH"
    },

    "skills": [

        "Python",

        "Docker",

        "SQL"
    ]
}

json_data = json.dumps(employee, indent=4)

print(json_data)
````

# Common Mistake

Wrong

```python
json.loads(student)
```

because

`student`

is already a dictionary.

`loads()` expects a JSON string.

Correct

```python
json.loads(json_string)
```

````python
import json

student = {
    "name": "Abhi"
}

try:
    json.loads(student)

except Exception as error:
    print(error)
````

# dumps() vs loads()

| Function | Meaning | Input | Output |
|----------|---------|-------|--------|
| dumps() | Dict → JSON | Python Object | JSON String |
| loads() | JSON → Dict | JSON String | Python Object |
