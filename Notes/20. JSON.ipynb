# JSON (JavaScript Object Notation)

## What is JSON?

JSON (JavaScript Object Notation) is a lightweight data-interchange format used to exchange data between applications.

Although JSON originated from JavaScript, today it is language-independent and supported by almost every programming language.

Examples:
- Python ↔ API
- Browser ↔ Server
- Mobile App ↔ Backend
- Database ↔ Application

JSON is one of the most widely used formats for data exchange.

# Why Do We Need JSON?

Imagine two applications want to communicate.

Python Application

↓

Data

↓

Java Application

Both languages have different data structures.

Python uses:

- list
- tuple
- dictionary

Java uses:

- ArrayList
- HashMap

How will they communicate?

They need a common language.

That common language is JSON.

# JSON Syntax Rules

1. Data is stored as key-value pairs.
2. Keys must always be strings.
3. Strings use double quotes.
4. Objects use {}
5. Arrays use []
6. Values can be:
   - String
   - Number
   - Boolean
   - null
   - Object
   - Array

````python
person = {
    "name": "Abhi",
    "age": 20,
    "student": True
}

print(person)
````

Although this looks like JSON,

it is actually a Python dictionary.

Many beginners confuse these two.

# Python Dictionary vs JSON

Python Dictionary

```python
{
    "name": "Abhi",
    "age": 20,
    "student": True
}
```

JSON

```json
{
    "name": "Abhi",
    "age": 20,
    "student": true
}
```

Notice the difference:

Python

```python
True
False
None
```

JSON

```json
true
false
null
```

# JSON Data Types

| JSON Type | Python Type |
|------------|-------------|
| String | str |
| Number | int / float |
| Boolean | bool |
| null | None |
| Array | list |
| Object | dict |

````python
sample = {
    "name": "Abhi",
    "age": 20,
    "cgpa": 8.5,
    "student": True,
    "skills": ["Python", "SQL", "ML"],
    "address": {
        "city": "Pune",
        "country": "India"
    }
}

print(sample)
````

# Nested JSON

JSON can contain another JSON object.

Example

Person

↓

Address

↓

City

This is called nested JSON.

````python
student = {
    "name": "Rahul",

    "address": {

        "city": "Delhi",

        "state": "Delhi"
    }
}

print(student["address"]["city"])
````

# JSON Array

JSON arrays are equivalent to Python lists.

Example

{
    "skills":
    [
        "Python",
        "SQL",
        "Pandas"
    ]
}

````python
student = {
    "skills": [
        "Python",
        "SQL",
        "Pandas"
    ]
}

print(student["skills"][1])
````

# Complex JSON Example

Real APIs often return nested objects and arrays together.

````python
employee = {

    "id": 101,

    "name": "Abhi",

    "salary": 75000,

    "department": {

        "name": "AI",

        "location": "Bangalore"

    },

    "skills": [

        "Python",

        "SQL",

        "Docker"

    ]
}

print(employee["department"]["location"])

print(employee["skills"][2])
````

# Interview Questions

Q1. Is every Python dictionary valid JSON?

No.

Example

```python
{
    "is_student": True
}
```

Valid Python

Invalid JSON

Because JSON expects

```json
true
```

not

```json
True
```

---

Q2. Is every JSON object a Python dictionary?

After parsing,

Yes.

Python converts JSON objects into dictionaries.
