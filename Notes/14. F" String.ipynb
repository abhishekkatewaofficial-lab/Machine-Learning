# Formatted Strings (f-Strings) 

```python
# f"Text {variable}"
```

```python
name = "Abhi"

print(f"Hello {name}")
```

Explanation

* Prefix f before the string.
* Place variables inside {}.
* Python automatically inserts their values.

2. Why Use f-Strings?

Old Method: Concatenation

```python
name = "Abhi"
age = 20

print("My name is " + name + " and I am " + str(age) + " years old.")
```

# Using f-Strings

```python
name = "Abhi"
age = 20

print(f"My name is {name} and I am {age} years old.")
```

# 3. Multiple Variables

```python
name = "Abhi"
age = 20
city = "Mumbai"

print(f"My name is {name}, I am {age} years old, and I live in {city}.")
```

# 4. Using Expressions Inside f-Strings

```python
a = 10
b = 20

print(f"Sum = {a + b}")
```

Explanation

Anything inside {} is evaluated as a Python expression.

# 5. Mathematical Operations

```python
x = 15
y = 5

print(f"Addition: {x + y}")
print(f"Subtraction: {x - y}")
print(f"Multiplication: {x * y}")
print(f"Division: {x / y}")
```

# 6. Calling Functions Inside f-Strings

```python
name = "abhi"

print(f"Uppercase: {name.upper()}")
```

# Explanation

# Methods and functions can be used inside {}.

# 7. Formatting Floating Point Numbers

```python
pi = 3.1415926535

print(f"{pi:.2f}")
```

# 9. Width Formatting

```python
num = 25

print(f"|{num:50}|")
```

Explanation

* Total width = 5 characters
* Extra spaces added on left

# 10. Left Alignment

```python
name = "Abhi"

print(f"|{name:<10}|")
```

# 11. Right Alignment

```python
name = "Abhi"

print(f"|{name:>10}|")
```

# 12. Center Alignment

```python
name = "Abhi"

print(f"|{name:^30}|")
```

# 13. Adding Commas to Large Numbers

```python
salary = 123456789

print(f"{salary:,}")
```

# 14. Percentage Formatting

```python
score = 0.85

print(f"{score:.2%}")
```

# 15. Using Conditional Expressions

```python
age = 20

print(f"{'Adult' if age >= 18 else 'Minor'}")
```

# 16. Date Formatting

```python
from datetime import datetime

today = datetime.now()

print(f"{today:%d-%m-%Y}")
```

```python
print(today)
```

# 17. Debugging Feature (=)

```python
x = 10
y = 20

print(f"{x=}")
print(f"{y=}")
print(f"{x+y=}")
```

Explanation

Useful for debugging because it shows variable name and value together
