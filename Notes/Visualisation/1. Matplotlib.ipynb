# 6. Plotting Multiple Lines

A single graph can display multiple datasets.

This makes comparison easier.

## Syntax

```python
plt.plot(x, y1)

plt.plot(x, y2)

plt.show()
```

> Each call to `plt.plot()` adds a new line to the same figure.

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

sales_2024 = [120, 150, 180, 200, 230, 250]
sales_2025 = [140, 170, 210, 240, 280, 320]

plt.plot(months, sales_2024)
plt.plot(months, sales_2025)

plt.xlabel("Month")
plt.ylabel("Revenue (₹ Thousands)")
plt.title("Company Revenue Comparison")
plt.grid(visible=True,
         linestyle = '--',
         alpha = 0.5
        )

plt.show()
````

## Observation

The graph contains two lines.

However, we cannot identify which line belongs to which dataset.

To solve this problem, use **`plt.legend()`**.

The legend identifies each plotted line using its label.

## Syntax

```python
plt.plot(x, y, label="Label Name")

plt.legend()
```

---

## Parameters

### `plt.plot()`

| Parameter | Description |
|-----------|-------------|
| label | Name displayed inside the legend |

### `plt.legend()`

```python
plt.legend()
```

Common Syntax

```python
plt.legend(loc="best")
```

### Important Parameters

| Parameter | Description |
|-----------|-------------|
| loc | Position of the legend |

Common values:

- `"best"` *(default)*
- `"upper right"`
- `"upper left"`
- `"lower right"`
- `"lower left"`
- `"center"`

> Always provide a `label` while plotting if you want it to appear in the legend.

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

sales_2024 = [120, 150, 180, 200, 230, 250]
sales_2025 = [140, 170, 210, 240, 280, 320]

plt.plot(months, sales_2024, label="2024")

plt.plot(months, sales_2025, label="2025")

plt.xlabel("Month")
plt.ylabel("Revenue (₹ Thousands)")
plt.title("Company Revenue Comparison")
plt.grid(visible=True,
         linestyle = '--',
         alpha = 0.5
        )

plt.legend()

plt.show()
````

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

sales_2024 = [120, 150, 180, 200, 230, 250]
sales_2025 = [140, 170, 210, 240, 280, 320]

plt.plot(months, sales_2024, label="2024 Revenue")

plt.plot(months, sales_2025, label="2025 Revenue")

plt.xlabel("Month")
plt.ylabel("Revenue (₹ Thousands)")
plt.title("Revenue Comparison (2024 vs 2025)")

plt.grid(alpha=0.3,
        linestyle = '--')

plt.legend(loc="upper left")

plt.show()
````

## Key Points

- Use one `plt.plot()` call for each dataset.
- Use the `label` parameter while plotting.
- Call `plt.legend()` only after plotting all datasets.
- A legend makes multi-line graphs easy to understand.

# 8. Styling Line Plots

Matplotlib provides several parameters to customize the appearance of a line plot.

Instead of using separate functions, most styling options are passed directly to `plt.plot()`.

## Syntax

```python
plt.plot(
    x,
    y,
    color="blue",
    linewidth=2,
    linestyle="-",
    marker="o",
    markersize=6,
    markerfacecolor="white",
    markeredgecolor="blue",
    alpha=1.0,
    label="Data"
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| `color` | Changes the line color |
| `linewidth` | Changes line thickness |
| `linestyle` | Changes the line style |
| `marker` | Adds markers on data points |
| `markersize` | Changes marker size |
| `markerfacecolor` | Changes marker fill color |
| `markeredgecolor` | Changes marker border color |
| `alpha` | Controls transparency (0 to 1) |

## `color`

Changes the color of the plotted line.

### Syntax

```python
plt.plot(x, y, color="red")
```

### Common Colors

```python
"red"
"blue"
"green"
"black"
"orange"
"purple"
"pink"
"brown"
"gray"
"cyan"
"magenta"
"yellow"
```

You can also use:


````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 210, 250, 290]

plt.plot(months, sales, color="royalblue")

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue")

plt.grid(alpha=0.3)

plt.show()
````

## `linewidth`

Controls the thickness of the line.

### Syntax

```python
plt.plot(x, y, linewidth=2)
```

### Default

```python
linewidth=1.5
```

Larger values create thicker lines.

````python
plt.plot(
    months,
    sales,
    color="royalblue",
    linewidth=4
)

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue")

plt.grid(alpha=0.3)

plt.show()
````

````python
plt.plot(
    months,
    sales,
    color="royalblue",
    linewidth=2,
    linestyle="--"
)

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue")

plt.grid(alpha=0.3)

plt.show()
````

## `marker`

Displays a symbol at each data point.

### Syntax

```python
plt.plot(x, y, marker="o")
```

### Common Markers

| Marker | Shape |
|---------|-------|
| `"o"` | Circle |
| `"s"` | Square |
| `"^"` | Triangle |
| `"*"` | Star |
| `"D"` | Diamond |
| `"x"` | Cross |
| `"+"` | Plus |
| `"."` | Point |

````python
plt.plot(
    months,
    sales,
    marker="o"
)

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue")

plt.grid(alpha=0.3)

plt.show()
````

## `alpha`

Controls transparency.

### Syntax

```python
plt.plot(x, y, alpha=0.5)
```

### Range

```python
0.0  → Fully Transparent

1.0  → Fully Visible
```

````python
plt.plot(
    months,
    sales,
    color="royalblue",
    linewidth=3,
    marker="o",
    alpha=0.6
)

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue")

plt.grid(alpha=0.3)

plt.show()
````

## Combining Multiple Styling Parameters

In real projects, multiple styling parameters are used together to create clean and readable visualizations.

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 210, 250, 290]

plt.plot(
    months,
    sales,
    color="royalblue",
    linewidth=2.5,
    linestyle="-",
    marker="o",
    markersize=8,
    markerfacecolor="white",
    markeredgecolor="royalblue",
    alpha=0.9,
    label="Revenue"
)

plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue (₹ Thousands)")

plt.grid(alpha=0.3)
plt.legend()

plt.show()
````

# 9. Creating a Figure

Before plotting, Matplotlib automatically creates a figure.

To customize its size or resolution, use `plt.figure()`.

## Syntax

```python
plt.figure()

plt.figure(
    figsize=(width, height),
    dpi=100
)
```

---

## Parameters

| Parameter | Description |
|-----------|-------------|
| `figsize` | Figure size in inches `(width, height)` |
| `dpi` | Resolution (Dots Per Inch) |

---

### Default Values

```python
figsize = (6.4, 4.8)
dpi = 100
```

> Always call `plt.figure()` **before** `plt.plot()`.

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 210, 250, 290]

plt.figure()

plt.plot(months, sales)

plt.title("Default Figure")

plt.show()
````

## Example: Custom Figure Size

`figsize` controls the width and height of the figure.

The values are measured in **inches**.

````python
plt.figure(figsize=(8, 5))

plt.plot(months, sales)

plt.title("Figure Size: 8 × 5")

plt.show()
````

## Example: DPI

`dpi` controls the resolution of the figure.

Higher DPI produces sharper images, especially when saving plots.

````python
plt.figure(figsize=(8, 5), 
           dpi=100)

plt.plot(months, sales)

plt.title("Figure with Higher DPI")

plt.show()
````

## Best Practices

- Use `figsize` to control the plot size.
- Use higher `dpi` (150–300) for reports or presentations.
- Create the figure before plotting any data.

Axis limits define the visible range of the x-axis and y-axis.

## Functions

- `plt.xlim()`
- `plt.ylim()`

---

## Syntax

```python
plt.xlim(left, right)

plt.ylim(bottom, top)
```

---

## Parameters

### `plt.xlim()`

| Parameter | Description |
|-----------|-------------|
| `left` | Minimum x-axis value |
| `right` | Maximum x-axis value |

### `plt.ylim()`

| Parameter | Description |
|-----------|-------------|
| `bottom` | Minimum y-axis value |
| `top` | Maximum y-axis value |

````python
months = [1, 2, 3, 4, 5, 6]
sales = [120, 150, 180, 210, 250, 290]

plt.figure(figsize=(8, 5))

plt.plot(months, sales, marker="o")

plt.xlim(1, 8)
plt.ylim(100, 320)

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Custom Axis Limits")

plt.grid(alpha=0.3)

plt.show()
````

Ticks are the values displayed along the x-axis and y-axis.

You can customize both their positions and labels.

## Functions

- `plt.xticks()`
- `plt.yticks()`

---

## Syntax

```python
plt.xticks(ticks)

plt.yticks(ticks)
```

or

```python
plt.xticks(
    ticks,
    labels
)
```

---

## Parameters

| Parameter | Description |
|-----------|-------------|
| `ticks` | Positions of tick marks |
| `labels` | Custom labels (optional) |

````python
months = [1, 2, 3, 4, 5, 6]
sales = [120, 150, 180, 210, 250, 290]

plt.figure(figsize=(8, 5))

plt.plot(months, sales, marker="o")

plt.xticks(
    ticks=[1, 2, 3, 4, 5, 6],
    labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
)

plt.yticks([100, 150, 200, 250, 300])

plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Custom Tick Labels")

plt.grid(alpha=0.3)

plt.show()
````

## Complete Line Plot Example

This example combines the concepts learned in this notebook.

````python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
revenue = [120, 150, 180, 210, 250, 290]

plt.figure(figsize=(8, 5), dpi=100)

plt.plot(
    months,
    revenue,
    color="royalblue",
    linewidth=2.5,
    linestyle="-",
    marker="o",
    markersize=8,
    markerfacecolor="white",
    markeredgecolor="royalblue",
    alpha=0.9,
    label="Revenue"
)

plt.title("Monthly Revenue (First Half of 2025)")
plt.xlabel("Month")
plt.ylabel("Revenue (₹ Thousands)")

plt.grid(alpha=0.7,
         linestyle='--')
plt.legend(loc='best')


plt.show()
````
