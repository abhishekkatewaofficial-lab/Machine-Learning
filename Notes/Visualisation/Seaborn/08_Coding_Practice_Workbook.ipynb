```python
# ── GLOBAL IMPORTS ──────────────────────────────────────────────────────
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.patches as mpatches
import numpy as np
import os

sns.set_theme(
    style='whitegrid', context='notebook', font_scale=1.1,
    rc={'figure.facecolor':'white','axes.facecolor':'#FAFAFA',
        'grid.alpha':0.35,'figure.dpi':110,
        'axes.spines.top':False,'axes.spines.right':False}
)
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
os.makedirs('outputs/practice', exist_ok=True)
print("✅ Imports done.")
```

```python
# ── DATAFRAME 1: tips_df  (Restaurant Analytics) ────────────────────────
tips_df = sns.load_dataset('tips')

# Engineered features
tips_df['tip_pct']         = (tips_df['tip'] / tips_df['total_bill'] * 100).round(2)
tips_df['bill_per_person'] = (tips_df['total_bill'] / tips_df['size']).round(2)
tips_df['tip_category']    = pd.cut(
    tips_df['tip_pct'],
    bins=[0, 10, 15, 20, 100],
    labels=['Low (<10%)', 'Standard (10-15%)', 'Good (15-20%)', 'Excellent (>20%)']
)
tips_df['is_weekend']   = tips_df['day'].isin(['Sat', 'Sun'])
tips_df['large_party']  = tips_df['size'] >= 4

# Reference variables
day_order = ['Thur', 'Fri', 'Sat', 'Sun']
time_pal  = {'Lunch': '#1565C0', 'Dinner': '#E53935'}
sex_pal   = {'Male': '#1565C0', 'Female': '#C62828'}
smoke_pal = {'Yes': '#E53935', 'No': '#2E7D32'}

print("✅ tips_df:", tips_df.shape)
print(tips_df.dtypes.to_string())
print()
print(tips_df.head(3).to_string())
```

```python
# ── DATAFRAME 2: penguins_df  (Biological Classification) ───────────────
penguins_df = sns.load_dataset('penguins').dropna()

# Engineered features
penguins_df['bill_ratio']  = (penguins_df['bill_length_mm'] /
                               penguins_df['bill_depth_mm']).round(3)
penguins_df['size_index']  = ((penguins_df['flipper_length_mm'] +
                                penguins_df['body_mass_g'] / 100) / 2).round(2)
penguins_df['is_heavy']    = penguins_df['body_mass_g'] > penguins_df['body_mass_g'].median()
penguins_df['mass_kg']     = (penguins_df['body_mass_g'] / 1000).round(3)

# Reference
species_pal = {'Adelie':'#E53935', 'Chinstrap':'#1565C0', 'Gentoo':'#2E7D32'}
island_pal  = {'Torgersen':'#6A1B9A', 'Biscoe':'#E65100', 'Dream':'#00695C'}
peng_vars   = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']

print("✅ penguins_df:", penguins_df.shape)
print(penguins_df.dtypes.to_string())
print()
print(penguins_df.head(3).to_string())
```

```python
# ── DATAFRAME 3: titanic_df  (Survival Analysis) ─────────────────────────
titanic_df = sns.load_dataset('titanic')

# Engineered features (on the raw df — NaN preserved for missing value demo)
titanic_df['survived_label'] = titanic_df['survived'].map({0: 'Died', 1: 'Survived'})
titanic_df['class_label']    = titanic_df['pclass'].map(
    {1: '1st Class', 2: '2nd Class', 3: '3rd Class'})
titanic_df['age_group']      = pd.cut(
    titanic_df['age'],
    bins=[0, 12, 18, 35, 60, 100],
    labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
)
titanic_df['fare_quartile']  = pd.qcut(
    titanic_df['fare'].fillna(titanic_df['fare'].median()),
    q=4, labels=['Q1 (Low)', 'Q2', 'Q3', 'Q4 (High)']
)

# Clean subset (no NaN) for ML-style analysis
titanic_clean = titanic_df[['survived','pclass','sex','age',
                             'fare','sibsp','parch']].dropna()

# Survival rate matrix
surv_matrix = titanic_df.groupby(['pclass','sex'])['survived'].mean().unstack() * 100

# Reference
surv_pal    = {0: '#E53935', 1: '#2E7D32'}
class_order = ['1st Class', '2nd Class', '3rd Class']

print("✅ titanic_df:", titanic_df.shape, "| titanic_clean:", titanic_clean.shape)
print()
print("Missing values:")
print(titanic_df.isnull().sum()[titanic_df.isnull().sum() > 0].to_string())
print()
print("Survival matrix (%):")
print(surv_matrix.round(1).to_string())
```

```python
# ── DATAFRAME 4: flights_df + aggregations  (Time Series) ───────────────
flights_df = sns.load_dataset('flights')

# Aggregations
flights_yearly = (flights_df.groupby('year')['passengers']
                             .sum().reset_index()
                             .rename(columns={'passengers':'total_passengers'}))
flights_yearly['yoy_growth_pct'] = (
    flights_yearly['total_passengers'].pct_change() * 100).round(2)

month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']
flights_monthly_avg = (flights_df.groupby('month')['passengers']
                                  .mean().reindex(month_order)
                                  .reset_index()
                                  .rename(columns={'passengers':'avg_passengers'}))

# Pivot table for heatmap
flights_pivot = flights_df.pivot(index='month', columns='year', values='passengers')
flights_pivot = flights_pivot.reindex(month_order)

# Era segmentation
flights_df['era'] = pd.cut(flights_df['year'], bins=[1948,1952,1956,1960],
                            labels=['1949-52','1953-56','1957-60'])

print("✅ flights_df:", flights_df.shape)
print("   flights_yearly:", flights_yearly.shape)
print("   flights_pivot:", flights_pivot.shape)
print()
print(flights_yearly.to_string(index=False))
```

```python
# ── DATAFRAME 5: diamonds_df  (Luxury Retail — 53,940 rows) ─────────────
diamonds_df = sns.load_dataset('diamonds')

# Engineered features
diamonds_df['price_per_carat'] = (diamonds_df['price'] / diamonds_df['carat']).round(0).astype(int)
diamonds_df['cut_score']       = diamonds_df['cut'].map(
    {'Fair':1, 'Good':2, 'Very Good':3, 'Premium':4, 'Ideal':5})
diamonds_df['clarity_score']   = diamonds_df['clarity'].map(
    {'I1':1,'SI2':2,'SI1':3,'VS2':4,'VS1':5,'VVS2':6,'VVS1':7,'IF':8})
diamonds_df['total_size']      = (diamonds_df['x'] * diamonds_df['y'] * diamonds_df['z']).round(2)

# Reference
cut_order     = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
clarity_order = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
cut_pal       = {'Fair':'#E53935','Good':'#FB8C00','Very Good':'#FDD835',
                 'Premium':'#2E7D32','Ideal':'#1565C0'}

# Sample for scatter/pairplot (full 53k is slow)
diamonds_sample = diamonds_df.sample(n=3000, random_state=42)

print("✅ diamonds_df:", diamonds_df.shape)
print("   diamonds_sample:", diamonds_sample.shape)
print()
print(diamonds_df[['carat','cut','color','clarity','price','price_per_carat']].describe().round(1).to_string())
```

```python
# ── DATAFRAME 6: mpg_df  (Automotive Engineering) ────────────────────────
mpg_df = sns.load_dataset('mpg').dropna()

# Engineered features
mpg_df['efficiency']       = (mpg_df['mpg'] / mpg_df['weight'] * 1000).round(3)
mpg_df['power_to_weight']  = (mpg_df['horsepower'] / mpg_df['weight'] * 100).round(3)
mpg_df['era']              = pd.cut(mpg_df['model_year'],
                                     bins=[69, 73, 77, 83],
                                     labels=['Early 70s','Mid 70s','Late 70s-80s'])
mpg_df['is_efficient']     = mpg_df['mpg'] > mpg_df['mpg'].median()

# Reference
origin_pal = {'usa':'#E53935', 'europe':'#1565C0', 'japan':'#2E7D32'}
origin_order = ['usa', 'europe', 'japan']
mpg_num_cols = ['mpg','cylinders','displacement','horsepower','weight','acceleration']
corr_mpg     = mpg_df[mpg_num_cols].corr()

print("✅ mpg_df:", mpg_df.shape)
print()
print(mpg_df[mpg_num_cols].describe().round(2).to_string())
```

```python
# ── DATAFRAME 7: exercise_df  (Health / Fitness) ─────────────────────────
exercise_df = sns.load_dataset('exercise')

# Reference
diet_pal = {'no fat': '#E53935', 'low fat': '#2E7D32'}
kind_order = ['rest', 'walking', 'running']

print("✅ exercise_df:", exercise_df.shape)
print(exercise_df.dtypes.to_string())
print()
print(exercise_df.head(6).to_string())
```

```python
# ── DATAFRAME 8: anscombe DataFrames  (Statistics Proof) ─────────────────
anscombe_all = sns.load_dataset('anscombe')

# Separate each dataset
ans_I   = anscombe_all[anscombe_all['dataset'] == 'I'].reset_index(drop=True)
ans_II  = anscombe_all[anscombe_all['dataset'] == 'II'].reset_index(drop=True)
ans_III = anscombe_all[anscombe_all['dataset'] == 'III'].reset_index(drop=True)
ans_IV  = anscombe_all[anscombe_all['dataset'] == 'IV'].reset_index(drop=True)

print("✅ anscombe_all:", anscombe_all.shape, "| 4 subsets extracted")
print()
print("Stats per dataset (all nearly identical!):")
stats_anscombe = (anscombe_all.groupby('dataset')['x','y']
                               .agg(['mean','std'])
                               .round(2))
print(stats_anscombe.to_string())
print()
print("⚠️  Same stats, VERY different shapes — always visualize!")
```
