## 1. Why Seaborn? The Industry Case

### 🧠 Intuition First

Imagine you're a Data Scientist at a retail company. Your manager asks for a chart showing:
> *"How does total bill vary by day of the week, split by gender, with confidence intervals?"*

With **Seaborn**, it's **2 lines**.

That's the fundamental value proposition of Seaborn.

## 2. Seaborn's Relationship with Matplotlib

### 🏗️ The Architecture

```
┌─────────────────────────────────────────────────┐
│                   YOUR CODE                      │
│         sns.barplot(data=df, ...)                │
└────────────────────┬────────────────────────────┘
                     │  calls
┌────────────────────▼────────────────────────────┐
│               SEABORN LAYER                      │
│  - Handles DataFrames natively                   │
│  - Computes statistics (mean, CI, etc.)          │
│  - Applies beautiful defaults                    │
│  - Manages color palettes                        │
└────────────────────┬────────────────────────────┘
                     │  renders via
┌────────────────────▼────────────────────────────┐
│             MATPLOTLIB LAYER                     │
│  - The actual rendering engine                   │
│  - Figure, Axes, Artists                         │
│  - You can always access it for fine-tuning      │
└─────────────────────────────────────────────────┘
```

### 🔑 Critical Insight

Every Seaborn plot **returns a Matplotlib Axes object**. This means:
- You can use `plt.title()`, `plt.xlabel()`, `ax.set_title()` on any Seaborn plot
- You can mix Seaborn and Matplotlib in the same figure
- You are never "locked in" — full Matplotlib power is always available

**Mental Model:** Think of Seaborn as a **smart, opinionated wrapper** around Matplotlib. Seaborn handles the *what* (statistics, colors, layouts). Matplotlib handles the *how* (pixels, rendering).

## 5. Figure Aesthetics: Themes, Styles & Contexts

### 🧠 Intuition

Seaborn's aesthetic system has **three layers**:

1. **Style** — Controls the background and grid appearance (`whitegrid`, `darkgrid`, etc.)
2. **Context** — Controls the scale of plot elements for different audiences (`paper`, `notebook`, `talk`, `poster`)
3. **Theme (rc params)** — Fine-grained font, line-width overrides

Think of it like choosing a suit: **Style** = the fabric color, **Context** = the fit (slim fit for meetings, oversized for presentations).

---

### 5.1 Seaborn Styles

````python
# ─── Comparing All 5 Seaborn Styles Side by Side ─────────────────────
# Business Problem: Choosing the right style for a weekly sales report.

# Sample data: weekly average tip percentages
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100
weekly_avg = tips.groupby('day', observed=True)['tip_pct'].mean().reset_index()
weekly_avg.columns = ['Day', 'Avg Tip %']

styles = ['white', 'whitegrid', 'dark', 'darkgrid', 'ticks']

fig, axes = plt.subplots(
    nrows=1, ncols=5,
    figsize=(22, 5)
)

for ax, style in zip(axes, styles):
    with sns.axes_style(style):  # Apply style locally (non-destructive)
        sns.barplot(
            data=weekly_avg,
            x='Day',
            y='Avg Tip %',
            palette='viridis',
            ax=ax
        )
        ax.set_title(f"Style: '{style}'", fontsize=12, fontweight='bold')
        ax.set_xlabel('')
        ax.set_ylabel('Avg Tip %' if style == 'white' else '')
        ax.tick_params(axis='x', rotation=30)

fig.suptitle(
    "Seaborn Style Comparison — Restaurant Tip Analysis",
    fontsize=15, fontweight='bold', y=1.02
)
plt.tight_layout()
plt.show()
````

### 📌 Style Selection Guide

| Style | When to Use | Avoid When |
|-------|-------------|------------|
| `whitegrid` | Reports, dashboards, EDA | Dark-mode presentations |
| `white` | Clean publication figures | Need grid reference lines |
| `darkgrid` | Personal exploration | Printing (uses ink) |
| `dark` | Dramatic presentations | Readability-critical contexts |
| `ticks` | Scientific/academic papers | General business use |

> 🏆 **Industry Standard:** `whitegrid` is the most widely used style in professional Data Science work — clean, readable, and works well in both light and dark documents.

---

### 5.2 Seaborn Contexts — Scaling for Your Audience

````python
# ─── Comparing All 4 Seaborn Contexts ────────────────────────────────
# Use Case: Same chart, different audience (Jupyter → Conference Poster)

contexts = ['paper', 'notebook', 'talk', 'poster']

fig, axes = plt.subplots(1, 4, figsize=(24, 5))

for ax, context in zip(axes, contexts):
    with sns.plotting_context(context):
        sns.boxplot(
            data=tips,
            x='day',
            y='total_bill',
            palette='Set2',
            ax=ax
        )
        ax.set_title(f"Context: '{context}'", fontweight='bold')
        ax.set_xlabel('Day')
        ax.set_ylabel('Total Bill ($)' if context == 'paper' else '')

fig.suptitle(
    "Seaborn Context Comparison — Bill Distribution by Day",
    fontsize=14, fontweight='bold', y=1.02
)
plt.tight_layout()
plt.show()
````

### 📌 Context Quick Reference

| Context | Font Scale | Line Width | Best For |
|---------|-----------|------------|----------|
| `paper` | 0.8× | Thin | Academic papers, dense reports |
| `notebook` | 1.0× | Normal | **Default — Jupyter notebooks** |

## 6. Color Palettes

### 🧠 Intuition: Colors Tell Stories

Color is not decoration — it's **data encoding**. Using the wrong type of palette is one of the most common mistakes in data visualization:

| Data Type | Palette Type | Example |
|-----------|-------------|------|
| Categories with no order | **Qualitative** | Species, Gender, Country |
| Numeric data (low → high) | **Sequential** | Temperature, Sales Revenue |
| Numeric data with a meaningful midpoint | **Diverging** | Correlation (−1 to +1), Profit/Loss |

### 6.1 Qualitative (Categorical) Palettes

````python
# ─── Qualitative Palettes: For Unordered Categories ──────────────────
# Use Case: Comparing tip percentages across days (no inherent order)

qualitative_palettes = ['deep', 'muted', 'pastel', 'bright', 'dark', 'colorblind']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for ax, palette_name in zip(axes, qualitative_palettes):
    sns.barplot(
        data=tips,
        x='day',
        y='total_bill',
        hue='day',              # Color each day differently
        palette=palette_name,
        estimator='mean',       # Show the mean total bill
        legend=False,
        ax=ax
    )
    ax.set_title(f"Palette: '{palette_name}'", fontsize=12, fontweight='bold')
    ax.set_xlabel('Day of Week')
    ax.set_ylabel('Avg Total Bill ($)')
    ax.tick_params(axis='x', rotation=0)

fig.suptitle(
    "Qualitative Color Palettes — Average Total Bill by Day",
    fontsize=15, fontweight='bold'
)
plt.tight_layout()
plt.show()
````

> ♿ **Accessibility Best Practice:** Always use `colorblind` palette when your visualization will be seen by a broad audience. ~8% of men have color vision deficiency. Using `colorblind`-safe palettes makes your work more inclusive and professional.

````python
# ─── Sequential Palettes: For Ordered/Numeric Data ───────────────────
# Use Case: Visualizing passenger count over time (values go 0 → high)

sequential_palettes = ['Blues', 'viridis', 'magma', 'plasma', 'YlOrRd', 'BuGn']

fig, axes = plt.subplots(2, 3, figsize=(18, 8))
axes = axes.flatten()

# Pivot flights data for a heatmap
flights_pivot = flights.pivot_table(
    index='month', columns='year', values='passengers'
)

for ax, palette_name in zip(axes, sequential_palettes):
    sns.heatmap(
        flights_pivot,
        cmap=palette_name,
        annot=False,
        linewidths=0.3,
        ax=ax,
        cbar_kws={'shrink': 0.7}
    )
    ax.set_title(f"cmap='{palette_name}'", fontsize=11, fontweight='bold')
    ax.set_xlabel('Year')
    ax.set_ylabel('Month')

fig.suptitle(
    "Sequential Palettes — Monthly Flight Passengers Heatmap",
    fontsize=14, fontweight='bold'
)
plt.tight_layout()
plt.show()
````

````python
# ─── Diverging Palettes: For Data with a Meaningful Center ───────────
# Use Case: Correlation matrix (values go from -1 → 0 → +1)
# The CENTER (0 = no correlation) must be visually obvious

diverging_palettes = ['coolwarm', 'RdBu', 'RdYlGn', 'PiYG', 'BrBG', 'seismic']

# Correlation matrix of tips numeric columns
tips_numeric    = tips[['total_bill', 'tip', 'size', 'tip_pct']]
correlation_mtx = tips_numeric.corr()

fig, axes = plt.subplots(2, 3, figsize=(20, 12))
axes = axes.flatten()

for ax, palette_name in zip(axes, diverging_palettes):
    sns.heatmap(
        correlation_mtx,
        cmap=palette_name,
        annot=True,
        fmt='.2f',
        vmin=-1, vmax=1,    # CRITICAL: Always set vmin/vmax for correlation heatmaps
        center=0,           # CRITICAL: Center the colormap at 0
        square=True,
        linewidths=0.5,
        ax=ax
    )
    ax.set_title(f"cmap='{palette_name}'", fontsize=11, fontweight='bold')

fig.suptitle(
    "Diverging Palettes — Tips Dataset Correlation Matrix",
    fontsize=14, fontweight='bold'
)
plt.tight_layout()
plt.show()
````

````python
# ─── Viewing and Creating Custom Palettes ────────────────────────────

# 1. View any palette as swatches
print("=== Viewing Built-in Palettes ===")
for name in ['deep', 'colorblind', 'viridis']:
    palette = sns.color_palette(name, n_colors=6)
    print(f"  {name}: {[f'#{int(r*255):02X}{int(g*255):02X}{int(b*255):02X}' for r, g, b in palette]}")

# 2. Create a custom brand palette (e.g., company colors)
brand_palette = ['#2D4A8A', '#E8593C', '#3AAFA9', '#F4C430', '#6B4226']

fig, ax = plt.subplots(figsize=(10, 3))
sns.barplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='day',
    palette=brand_palette,
    legend=False,
    ax=ax
)
ax.set_title("Custom Brand Color Palette — Average Bill by Day", fontweight='bold')
ax.set_xlabel('Day of Week')
ax.set_ylabel('Average Total Bill ($)')
plt.tight_layout()
plt.show()
````

## 7. Figure Customization

### 🧠 Intuition

A plot without proper titles, labels, and formatting is just a drawing. A plot WITH them is a **communication tool**. The goal of customization is to make the chart **self-explanatory** — anyone should understand it without reading surrounding text.

### 7.1 The Complete Customization Template

````python
# ─── Complete Professional Figure Template ────────────────────────────
# Business Problem: Present average restaurant bill and tip by day
# Audience: Restaurant manager reviewing weekly performance

fig, ax = plt.subplots(figsize=(10, 6))      # (1) Set figure size explicitly

# ── Core plot ──
sns.barplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    palette='Set2',
    capsize=0.1,            # Add caps to error bars
    errwidth=1.5,           # Error bar line width
    ax=ax
)

# ── Titles and Labels ──
ax.set_title(
    "Average Total Bill by Day and Gender\n(with 95% Confidence Intervals)",
    fontsize=15,
    fontweight='bold',
    pad=15                  # Space between title and plot
)
ax.set_xlabel('Day of the Week', fontsize=12, labelpad=10)
ax.set_ylabel('Average Total Bill (USD)', fontsize=12, labelpad=10)

# ── Tick Customization ──
ax.tick_params(axis='x', rotation=0, labelsize=11)
ax.tick_params(axis='y', labelsize=11)

# ── Legend Customization ──
ax.legend(
    title='Customer Gender',
    title_fontsize=11,
    fontsize=10,
    loc='upper left',
    framealpha=0.8          # Semi-transparent legend box
)

# ── Grid Customization ──
ax.yaxis.grid(True, linestyle='--', alpha=0.6)  # Horizontal grid only
ax.set_axisbelow(True)       # Grid behind bars (not on top)

# ── Annotation: Add business insight directly on the chart ──
ax.annotate(
    'Highest bills on\nSat & Sun (weekend)',
    xy=(2.5, 22),             # Point to annotate
    xytext=(1.2, 27),          # Text location
    fontsize=9,
    color='#444',
    arrowprops=dict(arrowstyle='->', color='#888', lw=1.2),
    bbox=dict(boxstyle='round,pad=0.3', facecolor='lightyellow', alpha=0.8)
)

# ── Data Source Footer ──
fig.text(
    0.99, 0.01,
    'Source: Seaborn Tips Dataset',
    ha='right', va='bottom',
    fontsize=8, color='gray', style='italic'
)

plt.tight_layout()             # ALWAYS call before plt.show() or savefig()
plt.show()
````

### 📌 Customization Parameter Reference

| Parameter/Method | What it Controls | Example |
|-----------------|-----------------|--------|
| `figsize=(w, h)` | Figure dimensions in inches | `figsize=(10, 6)` |
| `ax.set_title()` | Plot title | `fontsize=14, fontweight='bold'` |
| `ax.set_xlabel()` | X-axis label | `labelpad=10` for spacing |
| `ax.set_ylabel()` | Y-axis label | `fontsize=12` |
| `ax.tick_params()` | Tick size, rotation | `axis='x', rotation=45` |
| `ax.legend()` | Legend position, title | `loc='upper right'` |
| `ax.set_xlim()` | X-axis range | `ax.set_xlim(0, 100)` |
| `ax.set_ylim()` | Y-axis range | `ax.set_ylim(0, 50)` |
| `ax.annotate()` | Add text annotations | With arrowprops |
| `fig.text()` | Figure-level text | For source, footnotes |
| `plt.tight_layout()` | Remove whitespace clipping | Always use before save/show |

### 7.2 Multi-Panel Figures (Subplots)

## 8. Saving Figures — For Reports, Dashboards & Publications

### 🧠 Why Saving Matters

In a real job, you almost never just `plt.show()` a chart. You **save** it and embed it in:
- PowerPoint presentations
- PDF reports
- Dashboards (Streamlit, Tableau, Power BI)
- Web pages
- Research papers

Knowing which format to use and how to set quality parameters is a **professional skill**.

````python
# ─── Saving a Figure: The Complete Professional Approach ─────────────

import os

# Create an output folder for saved figures
output_dir = 'outputs'
os.makedirs(output_dir, exist_ok=True)

# ── Build the figure ──
fig, ax = plt.subplots(figsize=(10, 6))

sns.boxplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='time',
    palette='Set2',
    ax=ax
)

ax.set_title('Total Bill Distribution by Day and Meal Time', fontweight='bold', fontsize=14)
ax.set_xlabel('Day of the Week', fontsize=12)
ax.set_ylabel('Total Bill ($)', fontsize=12)
ax.legend(title='Meal Time')

plt.tight_layout()

# ── Save in multiple formats ──

# PNG: for web, presentations, general use
fig.savefig(
    f'{output_dir}/bill_by_day.png',
    dpi=150,              # 72 = web, 150 = reports, 300 = print quality
    bbox_inches='tight',  # CRITICAL: prevents title/label clipping
    facecolor='white'     # Ensure white background (not transparent)
)

# PDF: for publications and high-quality reports (vector format)
fig.savefig(
    f'{output_dir}/bill_by_day.pdf',
    dpi=300,
    bbox_inches='tight',
    format='pdf'
)

# SVG: for web / Inkscape / Adobe Illustrator editing
fig.savefig(
    f'{output_dir}/bill_by_day.svg',
    bbox_inches='tight',
    format='svg'
)

plt.show()

print(f"\n✅ Figures saved to: ./{output_dir}/")
print(f"   PNG (150 DPI) → for reports and presentations")
print(f"   PDF (300 DPI) → for academic papers (vector)")
print(f"   SVG (vector)  → for web and design tools")
````

### 📌 `savefig()` Parameter Reference

```python
fig.savefig(
    fname       = 'chart.png',    # Filepath + extension
    dpi         = 150,            # Dots per inch: 72 web, 150 report, 300 print
    bbox_inches = 'tight',        # ⭐ Always use — prevents label cutoff
    facecolor   = 'white',        # Background color of saved image
    transparent = False,          # True = transparent PNG background
    format      = 'png',          # 'png', 'pdf', 'svg', 'jpg', 'eps'
    pad_inches  = 0.1,            # Extra padding around tight boundary
)
```

| Format | Use Case | Resizable? |
|--------|----------|------------|
| PNG | Presentations, GitHub, web | ❌ Raster |
| JPG | Photographs, web (small size) | ❌ Raster |
| PDF | Academic papers, print | ✅ Vector |
| SVG | Web, design editing | ✅ Vector |
| EPS | LaTeX, journals | ✅ Vector |

> ⚠️ **Critical Warning:** Always use `bbox_inches='tight'`. Without it, long axis labels and titles get **cut off** in saved files even if they look fine in the notebook. This is the #1 saving mistake beginners make.

## 9. Best Practices

### ✅ Seaborn Professional Standards

````python
# ─── Reusable Plotting Function: Industry Best Practice ──────────────
# In production, wrap your common plot configurations into functions.
# This avoids code repetition and ensures visual consistency.

def plot_bar_comparison(
    data: pd.DataFrame,
    x: str,
    y: str,
    hue: str = None,
    title: str = 'Bar Chart',
    xlabel: str = None,
    ylabel: str = None,
    palette: str = 'Set2',
    figsize: tuple = (10, 6),
    rotation: int = 0,
    save_path: str = None
) -> None:
    """
    Generate a publication-quality bar chart with professional formatting.

    Parameters
    ----------
    data      : Tidy (long-form) DataFrame
    x         : Column name for the x-axis (categorical)
    y         : Column name for the y-axis (numeric)
    hue       : Column name for color grouping (optional)
    title     : Chart title string
    xlabel    : X-axis label (defaults to x column name)
    ylabel    : Y-axis label (defaults to y column name)
    palette   : Seaborn/matplotlib palette name
    figsize   : Tuple (width, height) in inches
    rotation  : X-tick label rotation in degrees
    save_path : File path to save the figure (optional)
    """
    fig, ax = plt.subplots(figsize=figsize)

    sns.barplot(
        data=data,
        x=x,
        y=y,
        hue=hue,
        palette=palette,
        ax=ax
    )

    ax.set_title(title, fontsize=14, fontweight='bold', pad=12)
    ax.set_xlabel(xlabel or x, fontsize=12, labelpad=8)
    ax.set_ylabel(ylabel or y, fontsize=12, labelpad=8)
    ax.tick_params(axis='x', rotation=rotation)

    if hue:
        ax.legend(title=hue.replace('_', ' ').title(), fontsize=10)

    ax.yaxis.grid(True, linestyle='--', alpha=0.5)
    ax.set_axisbelow(True)

    plt.tight_layout()

    if save_path:
        fig.savefig(save_path, dpi=150, bbox_inches='tight', facecolor='white')
        print(f"  💾 Figure saved: {save_path}")

    plt.show()


# ── Usage Example ──
plot_bar_comparison(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    title='Average Total Bill by Day and Gender',
    xlabel='Day of the Week',
    ylabel='Average Total Bill ($)',
    palette='Set2',
    figsize=(10, 6)
)
````

## 10. Common Mistakes to Avoid

| Mistake | Problem | Fix |
|---------|---------|-----|
| Forgetting `plt.tight_layout()` | Labels/titles get clipped | Always add before `plt.show()` |
| Using sequential palette for categories | Implies order where there is none | Use qualitative palette for categories |
| Not setting `vmin`/`vmax` on correlation heatmaps | Colors are misleading | Always set `vmin=-1, vmax=1` for correlation |
| Forgetting `bbox_inches='tight'` when saving | Labels get cut off in saved file | Add to every `savefig()` call |
| Re-loading dataset in every cell | Slow, messy code | Load once at top, reuse |
| Using `plt.style.use()` AND `sns.set_theme()` | Style conflicts | Use only `sns.set_theme()` |
| Hardcoding colors as strings ('blue', 'red') | Not scalable, unprofessional | Use named palettes |
| Missing axis labels | Chart is uninterpretable | Always label both axes |
| Overcrowded subplots | Unreadable | Use `figsize` and `tight_layout` |
| Using `plt.show()` before `savefig()` | Saves an empty figure | Always `savefig()` BEFORE `plt.show()` |

---

````python
# ─── DEMO: Common Mistake #10 — plt.show() Before savefig() ─────────
# ❌ WRONG: This saves an empty white image
#
# fig, ax = plt.subplots()
# sns.histplot(data=tips, x='total_bill', ax=ax)
# plt.show()           ← This CLEARS the figure from memory
# fig.savefig(...)     ← Saves EMPTY figure!

# ✅ CORRECT: Save FIRST, then show
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='steelblue', ax=ax)
ax.set_title('Correct Order: Save Before Show', fontweight='bold')
ax.set_xlabel('Total Bill ($)')
ax.set_ylabel('Count')
plt.tight_layout()
fig.savefig(f'{output_dir}/correct_save_order.png', dpi=120, bbox_inches='tight')  # FIRST
plt.show()  # SECOND
print("✅ Figure saved correctly (savefig before show)")
````

## 11. 🎤 Interview Notes

These are the **exact talking points** that impress in Data Science interviews:

---

**Q: What is Seaborn and how does it relate to Matplotlib?**

> *"Seaborn is a high-level statistical visualization library built on top of Matplotlib. It uses Matplotlib as its rendering engine but adds a DataFrame-native API, built-in statistical aggregations, automatic confidence intervals, and beautiful default aesthetics. Every Seaborn plot returns a Matplotlib Axes object, so you can always mix both libraries."*

---

**Q: How do you choose a color palette?**

> *"The choice depends on the data type: qualitative/categorical data → use qualitative palettes like `deep` or `colorblind`; ordered numeric data → sequential palettes like `viridis`; data with a meaningful midpoint like correlation → diverging palettes like `coolwarm` or `RdBu`."*

---

**Q: What is the difference between `style` and `context` in Seaborn?**

> *"Style controls background and grid aesthetics (white, darkgrid, etc.), while context controls the scale of all plot elements for different audiences: `paper` for dense reports, `notebook` for Jupyter, `talk` for slides, `poster` for large displays."*

---

**Q: What does `tight_layout()` do?**

> *"It automatically adjusts subplot parameters to prevent overlapping of labels, titles, and tick marks. Without it, especially in multi-panel figures, elements often get clipped."*

---

**Q: What is a tidy/long-form dataset and why does Seaborn prefer it?**

> *"In a tidy dataset, each row is one observation, each column is one variable, and each value is one measurement. Seaborn's `x`, `y`, `hue`, `style`, and `size` parameters each expect a single column name, which only works naturally with long-form data. Wide-form data needs to be `pd.melt()`-ed first."*
