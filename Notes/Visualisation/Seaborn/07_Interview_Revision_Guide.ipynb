## ⏱ Section 1 — Read This 5 Minutes Before the Interview

> 📌 If you have 5 minutes, read ONLY this section. It covers the most likely interview topics.

---

### 🎯 The 7 Things That Impress Every Interviewer

1. **"I always mask the upper triangle"** — shows you know correlation heatmaps professionally
2. **"I use `select_dtypes(include='number')` before `.corr()`"** — shows you understand data types
3. **"r = 0.8 between two predictors is multicollinearity — I'd drop one"** — ML awareness
4. **"Correlation ≠ causation — I check with LOWESS before assuming linearity"** — statistical maturity
5. **"I use `kind='hex'` for datasets over 10,000 rows"** — performance awareness
6. **"FacetGrid gives me more control than relplot"** — advanced knowledge
7. **"I write reusable plotting functions so charts are consistent across the project"** — engineering mindset

---

### 🔑 5 Code Patterns — Write These from Memory

```python
# 1. MASKED CORRELATION HEATMAP (asked most often)
corr = df.select_dtypes(include='number').corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f',
            cmap='coolwarm', center=0, square=True)

# 2. FEATURE-TARGET CORRELATION RANKING
df.corr()['target'].drop('target').abs().sort_values(ascending=False)

# 3. MISSING VALUE HEATMAP
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='gray_r')

# 4. PAIRPLOT WITH HUE (KDE preferred)
sns.pairplot(df, hue='species', diag_kind='kde', palette='colorblind')

# 5. JOINTPLOT FOR DEEP BIVARIATE ANALYSIS
sns.jointplot(data=df, x='col1', y='col2', kind='reg')
```

---

### 🚨 3 Things That FAIL You Instantly

| ❌ Common Mistake | ✅ Fix |
|------------------|-------|
| `sns.heatmap(df.corr())` with full matrix | Add `mask = np.triu(np.ones_like(corr, dtype=bool))` |
| Using `fmt='d'` on correlation heatmap | Pearson r is float — use `fmt='.2f'` |
| Passing `ax=` to `relplot`, `lmplot`, `displot` | These are figure-level — they create their own figure |

## 📊 Section 2 — Plot Selection Decision Guide

### "Which Seaborn plot should I use?"

```
STEP 1: How many variables?
   │
   ├── ONE VARIABLE ─────────────────────────────────────────────────
   │     Numeric?  →  histplot  /  kdeplot  /  ecdfplot
   │     Category? →  countplot
   │
   ├── TWO VARIABLES ────────────────────────────────────────────────
   │     Both Numeric?
   │       ├── Want raw points?         →  scatterplot
   │       ├── Want trend + CI?         →  lineplot
   │       ├── Want linear regression?  →  regplot
   │       ├── Want density + margin?   →  jointplot(kind='kde')
   │       └── n > 10,000?              →  jointplot(kind='hex')
   │     Numeric + Category?
   │       ├── Compare medians/IQR?     →  boxplot
   │       ├── Compare distributions?   →  violinplot  /  kdeplot(hue=)
   │       ├── Compare means + CI?      →  barplot(estimator='mean')
   │       └── Show individual points?  →  stripplot  /  swarmplot
   │     Both Category?
   │       →  countplot(x=cat1, hue=cat2)  OR  heatmap of pivot table
   │
   ├── MANY VARIABLES ───────────────────────────────────────────────
   │     Correlations between all?      →  heatmap(df.corr(), mask=triu)
   │     All pairs scatter/kde?         →  pairplot  /  PairGrid
   │     Cluster similar variables?     →  clustermap
   │     Missing values?                →  heatmap(df.isnull())
   │
   └── FACETED / GROUPED ────────────────────────────────────────────
         By one category → relplot / displot / catplot (col= or row=)
         Custom layout   → FacetGrid.map_dataframe()
         Custom per-zone → PairGrid (diag / upper / lower)
```

---

### 🧩 Plot-to-Function Quick Match

| What you want | Axes-level function | Figure-level equivalent |
|---------------|--------------------|-----------------------|
| Scatter | `sns.scatterplot(ax=ax)` | `sns.relplot(kind='scatter')` |
| Line | `sns.lineplot(ax=ax)` | `sns.relplot(kind='line')` |
| Linear regression | `sns.regplot(ax=ax)` | `sns.lmplot()` |
| Histogram | `sns.histplot(ax=ax)` | `sns.displot(kind='hist')` |
| KDE | `sns.kdeplot(ax=ax)` | `sns.displot(kind='kde')` |
| Boxplot | `sns.boxplot(ax=ax)` | `sns.catplot(kind='box')` |
| Violin | `sns.violinplot(ax=ax)` | `sns.catplot(kind='violin')` |
| Bar | `sns.barplot(ax=ax)` | `sns.catplot(kind='bar')` |
| Count | `sns.countplot(ax=ax)` | `sns.catplot(kind='count')` |
| Heatmap | `sns.heatmap(ax=ax)` | — (no figure-level) |
| Pairplot | — | `sns.pairplot()` = `PairGrid` shortcut |

> 🎯 **Rule:** Axes-level → takes `ax=`. Figure-level → creates its own Figure. **Never mix them.**

             hue='cat', multiple='stack', fill=True)

            common_norm=False, bw_adjust=1.0, clip=(0, None))

---

### 📦 CATEGORICAL PLOTS (NB2)
```python
# countplot
sns.countplot(data=df, x='cat', hue='cat2', order=['A','B','C'],
              palette='Set2', stat='count')

# barplot — MEANS with CI
sns.barplot(data=df, x='cat', y='num', estimator='mean',
            errorbar=('ci', 95), hue='cat2', palette='deep')

# boxplot
sns.boxplot(data=df, x='cat', y='num', hue='cat2', width=0.5,
            medianprops=dict(color='black', lw=2.5),
            flierprops=dict(marker='.', alpha=0.3))

               split=True, inner='box', scale='width')

# stripplot / swarmplot (show all points)
sns.stripplot(data=df, x='cat', y='num', hue='cat2', jitter=True, alpha=0.6)
sns.swarmplot(data=df, x='cat', y='num', hue='cat2', size=4)
```

### 📈 RELATIONAL PLOTS (NB3)

### 📉 REGRESSION PLOTS (NB3)
```python

sns.regplot(data=df, x='x', y='y', ci=95, scatter=True,
            lowess=False,            # True = non-parametric curve

            scatter_kws={'alpha': 0.4, 's': 30},
            line_kws={'color': 'red', 'lw': 2}, ax=ax)

sns.lmplot(data=df, x='x', y='y', hue='cat', col='cat2',
           scatter_kws={'alpha': 0.4}, line_kws={'lw': 2},
           ci=95, height=4, aspect=1.2)
```

## 🪤 Section 5 — Trap Questions Gallery

> These are designed to catch freshers who memorized syntax but didn't understand concepts.

---

### Trap 1 — The Figure-Level ax= Trap
```python
fig, ax = plt.subplots()
sns.relplot(data=tips, x='total_bill', y='tip', ax=ax)  # ❌ TypeError!
```
**Why it fails:** `relplot` is figure-level — it creates its own Figure. It doesn't accept `ax=`.
**Fix:** Use `sns.scatterplot(data=tips, x='total_bill', y='tip', ax=ax)` (axes-level).

---

### Trap 2 — fmt='d' on Correlation
```python
sns.heatmap(df.corr(), annot=True, fmt='d')   # ❌ Prints 0 everywhere!
```
**Why it fails:** Pearson r values are floats (e.g., 0.674). `fmt='d'` rounds to integer → all show 0 or 1.
**Fix:** `fmt='.2f'`

---

### Trap 3 — Missing `center=0` in coolwarm heatmap
```python
sns.heatmap(corr, cmap='coolwarm', annot=True)  # ❌ Misleading colors!
```
**Why it fails:** Without `center=0`, if all correlations are positive, the colormap minimum is NOT at white — your lowest positive values appear blue (suggesting negative). **Always use `center=0`.**

---

### Trap 4 — Unmasked correlation = Redundant data
```python
sns.heatmap(df.corr(), annot=True)  # ❌ Upper triangle = mirror of lower
```
**Why it fails:** Shows every correlation twice — wastes space and confuses readers.
**Fix:** Always `mask = np.triu(np.ones_like(corr, dtype=bool))`.

---

### Trap 5 — Using pairplot on all columns
```python
sns.pairplot(titanic)   # ❌ Crashes or produces nonsense!
```
**Why it fails:** `pairplot` tries to scatter-plot categorical columns (sex, embarked, etc.) — crashes or produces meaningless outputs.
**Fix:** `sns.pairplot(titanic, vars=['age', 'fare', 'pclass', 'survived'])`

---

### Trap 6 — split=True with 3 hue groups
```python
sns.violinplot(data=penguins, x='sex', y='body_mass_g',
               hue='species', split=True)   # ❌ Ignored or error!
```
**Why it fails:** `split=True` only works with **exactly 2 hue categories**. With 3 species, it's meaningless.
**Fix:** Remove `split=True` or filter to 2 species.

---

### Trap 7 — FacetGrid with map() and kwargs
```python
g = sns.FacetGrid(df, col='time')
g.map(sns.scatterplot, 'total_bill', 'tip', hue='sex')  # ❌ hue= ignored!
```
**Why it fails:** `map()` can't pass `hue=` — it passes only positional Series.
**Fix:** `g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', hue='sex')`

---

### Trap 8 — Accessing FacetGrid panels with a loop (interview context)
```python
for ax in g.axes.flat:        # ❌ Works but violates no-loop rule
    ax.axhline(y=5)
```
**Better way (no loop):** Use `g.map(lambda **kw: plt.axhline(y=5, color='red'))` or `g.axes_dict['key']` for specific panels.

---

### Trap 9 — Treating correlated features as independent predictors
```python
# mpg dataset — all 4 features are r > 0.80 correlated!
model.fit(df[['cylinders', 'displacement', 'weight', 'horsepower']], df['mpg'])
# ❌ Multicollinearity! Coefficients are unstable.
```
**Fix:** Pick one representative feature per correlated cluster. Detect with heatmap.

---

### Trap 10 — Using countplot when you need a proportion
```python
sns.countplot(data=titanic, x='sex', hue='survived')
# ❌ Makes it LOOK like males survived less — but there were also MORE males!
```
**Fix:** Use `barplot(estimator='mean')` or compute `df.groupby('sex')['survived'].mean()` first. Show **rates**, not counts, for imbalanced groups.

---

### Trap 11 — Ignoring `common_norm` in kdeplot
```python
sns.kdeplot(data=tips, x='tip', hue='time')   # group sizes differ!
# KDE auto-normalizes but relative HEIGHTS may mislead about group sizes.
```
**Fix:** Add `common_norm=False` — each group normalized to area=1 independently. Always use this when groups have different n.

---

### Trap 12 — Using `stat='count'` in histplot for group comparison
```python
sns.histplot(data=titanic, x='age', hue='pclass', stat='count')
# ❌ 3rd class has 3× more passengers — bars look bigger not because age differs!
```
**Fix:** `stat='density'` or `stat='probability'` to normalize for group size.

---

### Trap 13 — Confusing regplot and lmplot
```python
fig, ax = plt.subplots()
sns.lmplot(data=tips, x='total_bill', y='tip', ax=ax)  # ❌ lmplot ignores ax=!
```
**Why it fails:** `lmplot` is figure-level — creates its own figure.
**Fix:** `sns.regplot(data=tips, x='total_bill', y='tip', ax=ax)` for axes-level regression.

---

### Trap 14 — Misinterpreting clustermap axis reordering
> *"The clustermap heatmap shows `weight` has high correlation with `mpg` based on its position!"*

**Why it's wrong:** Clustermap **reorders rows and columns** based on clustering similarity. The position no longer corresponds to original order. Always read the **dendrogram**, not the position.

---

### Trap 15 — Interpreting correlation as causation
> *"total_bill and tip are correlated at r=0.67, so increasing the bill causes higher tips."*

**Why it's wrong:** Correlation shows a statistical relationship — not a directional causal mechanism. Both could be caused by a third factor (party size, occasion). Always say: *"total_bill and tip are correlated — not necessarily causal."*

## 🐛 Section 6 — Bug Gallery with Fixes

````python
# 🟢 Bug Gallery — Run to see FIXED versions of common Seaborn errors
# Each cell shows the WRONG pattern (commented) and the FIXED version below it.

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style='whitegrid', font_scale=1.05)
tips     = sns.load_dataset('tips')
penguins = sns.load_dataset('penguins').dropna()
print("✅ Libraries loaded for Bug Gallery.")
````

````python
# 🐛 BUG 1: fmt='d' on float correlation heatmap
# ❌ WRONG:
# sns.heatmap(tips.select_dtypes('number').corr(), annot=True, fmt='d')
# Result: all cells show 0 or 1 (integers) — correlation values lost!

# ✅ FIX: use fmt='.2f' for 2 decimal float display
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

corr_tips = tips.select_dtypes(include='number').corr()
mask      = np.triu(np.ones_like(corr_tips, dtype=bool))

# Wrong fmt
sns.heatmap(corr_tips, mask=mask, annot=True, fmt='d',
            cmap='coolwarm', center=0, square=True, ax=axes[0])
axes[0].set_title('❌ fmt="d" — All values truncated to 0 or 1', color='#C62828', fontweight='bold')

# Correct fmt
sns.heatmap(corr_tips, mask=mask, annot=True, fmt='.2f',
            cmap='coolwarm', center=0, square=True, ax=axes[1])
axes[1].set_title('✅ fmt=".2f" — Correct float display', color='#2E7D32', fontweight='bold')

plt.suptitle('🐛 Bug 1: fmt="d" vs fmt=".2f" on Correlation Heatmap', fontweight='bold', y=1.02)
plt.tight_layout(); plt.show()
````

````python
# 🐛 BUG 2: center=0 missing — misleading colormap
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
corr = tips.select_dtypes(include='number').corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

# Without center=0 — all correlations look the same color
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f',
            cmap='coolwarm', square=True, ax=axes[0])
axes[0].set_title('❌ No center=0 — scale off-center
(Weak positives look bluish!)',
                  color='#C62828', fontweight='bold')

# With center=0 — proper diverging around zero
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f',
            cmap='coolwarm', center=0, vmin=-1, vmax=1,
            square=True, ax=axes[1])
axes[1].set_title('✅ center=0, vmin=-1, vmax=1 — Correct diverging scale',
                  color='#2E7D32', fontweight='bold')

plt.suptitle('🐛 Bug 2: Missing center=0 in Correlation Heatmap', fontweight='bold', y=1.02)
plt.tight_layout(); plt.show()
````

````python
# 🐛 BUG 3: Unmasked correlation heatmap
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
corr = tips.select_dtypes(include='number').corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

# Full matrix — redundant upper triangle
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, ax=axes[0])
axes[0].set_title('❌ Full matrix — upper triangle is redundant mirror',
                  color='#C62828', fontweight='bold')

# Masked — professional standard
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, vmin=-1, vmax=1, square=True, ax=axes[1])
axes[1].set_title('✅ Masked upper triangle — clean professional version',
                  color='#2E7D32', fontweight='bold')

plt.suptitle('🐛 Bug 3: Unmasked vs Masked Correlation Heatmap', fontweight='bold', y=1.02)
plt.tight_layout(); plt.show()
````

````python
# 🐛 BUG 4: stat='count' vs stat='density' for groups of different sizes
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# stat='count' — misleading! Group sizes differ
sns.histplot(data=tips, x='total_bill', hue='time', bins=18,
             stat='count', common_norm=True, ax=axes[0])
axes[0].set_title('❌ stat="count" — Dinner looks dominant
but only because more dinner visits',
                  color='#C62828', fontweight='bold')
axes[0].set_xlabel('Total Bill ($)')

# stat='density' — fair comparison
sns.histplot(data=tips, x='total_bill', hue='time', bins=18,
             stat='density', common_norm=False, ax=axes[1])
axes[1].set_title('✅ stat="density", common_norm=False
Fair comparison regardless of group size',
                  color='#2E7D32', fontweight='bold')
axes[1].set_xlabel('Total Bill ($)')

n_lunch  = (tips['time'] == 'Lunch').sum()
n_dinner = (tips['time'] == 'Dinner').sum()
plt.suptitle(f'🐛 Bug 4: stat="count" vs stat="density" (Lunch n={n_lunch} | Dinner n={n_dinner})',
             fontweight='bold', y=1.02)
plt.tight_layout(); plt.show()
````

````python
# 🐛 BUG 5: split=True violinplot with 3 hue groups
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ❌ split=True with 3 species — Seaborn either errors or ignores split
try:
    sns.violinplot(data=penguins, x='island', y='body_mass_g',
                   hue='species', split=True, ax=axes[0])
    axes[0].set_title('❌ split=True with 3 hue groups
(Seaborn ignores split or raises error)',
                      color='#C62828', fontweight='bold')
except Exception as e:
    axes[0].text(0.5, 0.5, f'Error: {str(e)[:60]}', transform=axes[0].transAxes,
                 ha='center', va='center', color='red', fontsize=10, wrap=True)
    axes[0].set_title('❌ split=True with 3 species → Error', color='#C62828', fontweight='bold')

# ✅ Without split — correct for 3+ groups
sns.violinplot(data=penguins, x='island', y='body_mass_g',
               hue='species', inner='box', palette='colorblind', ax=axes[1])
axes[1].set_title('✅ No split= — correct for 3 hue categories',
                  color='#2E7D32', fontweight='bold')
axes[1].legend(title='Species', fontsize=8)

plt.suptitle('🐛 Bug 5: split=True requires exactly 2 hue categories',
             fontweight='bold', y=1.02)
plt.tight_layout(); plt.show()
````

### THE 6 RULES OF PROFESSIONAL SEABORN

| # | Rule | Why |
|---|------|-----|
| 1 | Always mask correlation heatmap upper triangle | Mirror = redundant = unprofessional |
| 2 | Always use `center=0` with `coolwarm` | Without it, colormap is misleading |
| 3 | Always `select_dtypes(include='number')` before `.corr()` | Categoricals break Pearson r |
| 4 | Figure-level functions NEVER take `ax=` | They create their own figure |
| 5 | Use `common_norm=False` in kdeplot when group sizes differ | Prevents misleading normalization |
| 6 | Use `kind='hex'` when n > 10,000 in jointplot | Scatter is unreadable at that scale |

### INTERVIEW VOCABULARY (use these words)

| Technical Term | Use When |
|---------------|---------|
| **Multicollinearity** | Discussing high correlations between predictors |
| **Data leakage** | A feature derived from the target variable |
| **Marginal distribution** | The side panels in `jointplot` |
| **Simpson's Paradox** | A trend reverses when groups are disaggregated |
| **Anscombe's Quartet** | Explaining why visualization is always necessary |
| **Figure-level vs Axes-level** | Explaining FacetGrid / relplot architecture |
| **LOWESS** | Non-parametric regression — actual curve shape |
| **95% Confidence Interval** | Error bars in `barplot` — 95% of resampled means fall here |
| **Pearson r** | Linear correlation coefficient (ranges −1 to +1) |
| **R²** | Proportion of variance explained = r² |
