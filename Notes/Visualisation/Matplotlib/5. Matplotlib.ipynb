# Matplotlib with Pandas

In real-world projects, data is rarely stored as Python lists.

Instead, data usually comes from:

- CSV files
- Excel files
- SQL databases
- APIs

Pandas is used to load and manage this data, while Matplotlib is used to visualize it.

This combination is one of the most common workflows in Data Analysis.

````python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("dark_background")
````

# Step 2 — Create a Sample DataFrame

````python
import pandas as pd

df = pd.DataFrame({
    "Month": [
        "Jan",
        "Feb",
        "Mar",
        "Apr",
        "May",
        "Jun"
    ],
    "Sales": [
        120,
        150,
        180,
        170,
        210,
        250
    ],
    "Profit": [
        20,
        28,
        35,
        32,
        45,
        55
    ],
    "Orders": [
        80,
        95,
        110,
        108,
        130,
        150
    ]
})

display(df)
````

# Step 3 — Understanding the DataFrame

A DataFrame is a table consisting of rows and columns.

Each column is called a **Series**.

Example:

```python
df["Sales"]
```

returns only the Sales column.

# Step 4 — First Plot Using a DataFrame

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(
    figsize=(8, 5)
)

ax.plot(
    df["Month"],
    df["Sales"],
    marker="o"
)

ax.set_title("Monthly Sales")

ax.set_xlabel("Month")

ax.set_ylabel("Sales")

ax.grid(alpha=0.3)

plt.show()
````

# Step 5 — Plot Multiple Columns

````python
fig, ax = plt.subplots(
    figsize=(9, 5)
)

ax.plot(
    df["Month"],
    df["Sales"],
    marker="o",
    linewidth=2,
    label="Sales"
)

ax.plot(
    df["Month"],
    df["Profit"],
    marker="s",
    linewidth=2,
    label="Profit"
)

ax.legend()

ax.set_title("Sales vs Profit")

ax.set_xlabel("Month")

ax.set_ylabel("Amount")

ax.grid(alpha=0.3)

plt.show()
````

# Step 6 — Bar Chart from DataFrame

````python
fig, ax = plt.subplots(
    figsize=(8, 5)
)

ax.bar(
    df["Month"],
    df["Orders"]
)

ax.set_title("Monthly Orders")

ax.set_xlabel("Month")

ax.set_ylabel("Orders")

plt.show()
````

# Step 7 — Scatter Plot

````python
fig, ax = plt.subplots(
    figsize=(7, 5)
)

ax.scatter(
    df["Sales"],
    df["Profit"]
)

ax.set_title("Sales vs Profit")

ax.set_xlabel("Sales")

ax.set_ylabel("Profit")

ax.grid(alpha=0.3)

plt.show()
````

# Step 8 — Histogram

````python
fig, ax = plt.subplots(
    figsize=(7, 5)
)

ax.hist(
    df["Sales"],
    bins=5
)

ax.set_title("Sales Distribution")

plt.show()
````

# Step 9 — Box Plot

````python
fig, ax = plt.subplots(
    figsize=(6, 5)
)

ax.boxplot(
    df["Sales"]
)

ax.set_title("Sales Distribution")

plt.show()
````

# Step 2 — Load Dataset

````python
titanic = sns.load_dataset("titanic")

titanic.head()
````

# Which passenger class had the highest number of passengers on the Titanic?

````python
class_counts = titanic["class"].value_counts()
display(class_counts.to_frame())

fig, ax = plt.subplots(figsize=(7,5))

ax.bar(
    class_counts.index,
    class_counts.values
)

ax.set_title("Passengers by Class")
ax.set_xlabel("Passenger Class")
ax.set_ylabel("Number of Passengers")

plt.show()
````

# Did more passengers survive, or did more passengers lose their lives?

````python
survival = titanic["survived"].value_counts()
display(survival.to_frame())

fig, ax = plt.subplots(figsize=(6,6))

ax.pie(
    survival.values,
    labels=["Did Not Survive", "Survived"],
    autopct="%1.1f%%",
    startangle=90,
    shadow=True
)

ax.set_title("Passenger Survival Rate")

plt.show()
````
