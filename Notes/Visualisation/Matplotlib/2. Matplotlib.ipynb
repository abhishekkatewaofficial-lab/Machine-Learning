# 1. Bar Plot

A bar plot is used to compare values across different categories.

## Best Use Cases

- Product Sales
- Student Marks
- Survey Results
- Population Comparison
- Revenue Comparison

---

## When NOT to Use

- Continuous data → Use Histogram
- Relationship between two numerical variables → Use Scatter Plot

---

## Syntax

```python
plt.bar(x, height)
```

```python
plt.bar(
    x,
    height,
    width=0.8,
    color="royalblue",
    edgecolor="black",
    linewidth=1,
    label="Sales",
    alpha=0.9,
    bottom=None
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| x | Categories |
| height | Bar heights |
| width | Width of bars |
| color | Fill color |
| edgecolor | Border color |
| linewidth | Border thickness |
| alpha | Transparency |
| label | Legend |
| bottom | Starting position (used in stacked bars) |

````python
products = ["Laptop", "Phone", "Tablet", "Watch"]

sales = [120, 180, 90, 140]

plt.bar(products, sales)

plt.title("Product Sales")
plt.show()
````

````python
products = ["Laptop", "Phone", "Tablet", "Watch"]

sales = [120, 180, 90, 140]

plt.figure(figsize=(8, 5))

plt.bar(
    products,
    sales,
    width=0.6,
    color="royalblue",
    edgecolor="black",
    linewidth=1.2,
    alpha=0.9,
    label="Sales"
)

plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")

plt.grid(axis="y", alpha=0.3)

plt.legend()

plt.show()
````

## Best Practices

- Use bar plots only for categorical data.
- Keep bar widths consistent.
- Grid on the y-axis improves readability.

# Horizontal Bar Plot

A horizontal bar plot is useful when category names are long.

## Syntax

```python

```

---

## Parameters

| Parameter | Description |
|-----------|-------------|
| y | Categories |
| width | Bar length |

````python
products = [
    "MacBook Pro",
    "iPhone",
    "iPad",
    "Apple Watch"
]

sales = [120, 180, 90, 140]

plt.barh(products, sales)

plt.grid(axis="x",
         alpha = 0.3)

plt.show()
````

# Grouped Bar Chart

Grouped bar charts compare multiple datasets for the same categories.

## Syntax

```python
plt.bar(x1, data1)

plt.bar(x2, data2)
```

````python
import numpy as np

products = ["Laptop", "Phone", "Tablet", "Watch"]

sales_2024 = [120, 180, 90, 140]
sales_2025 = [140, 210, 110, 170]

x = np.arange(len(products))
width = 0.35

plt.figure(figsize=(8, 5))

plt.bar(
    x - width / 2,
    sales_2024,
    width,
    label="2024"
)

plt.bar(
    x + width / 2,
    sales_2025,
    width,
    label="2025"
)

plt.xticks(x, products)

plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.title("Sales Comparison")

plt.legend()

plt.grid(axis="y", alpha=0.3)

plt.show()
````

# Stacked Bar Chart

Stacked bars show the contribution of multiple datasets to the total value.

## Syntax

```python
plt.bar(x, data1)

plt.bar(
    x,
    data2,
    bottom=data1
)
```

````python
products = ["Laptop", "Phone", "Tablet", "Watch"]

online = [80, 120, 60, 90]

offline = [40, 60, 30, 50]

plt.figure(figsize=(8, 5))

plt.bar(
    products,
    online,
    label="Online"
)

plt.bar(
    products,
    offline,
    bottom=online,
    label="Offline"
)

plt.xlabel("Products")
plt.ylabel("Units Sold")

plt.title("Online vs Offline Sales")

plt.legend()

plt.grid(axis="y", alpha=0.3)

plt.show()
````

# Adding Value Labels

Matplotlib provides `bar_label()` to display values on top of bars.

## Syntax

```python
container = plt.bar(...)

plt.bar_label(container)
```

````python
products = ["Laptop", "Phone", "Tablet", "Watch"]

sales = [120, 180, 90, 140]

plt.figure(figsize=(8, 5))

bars = plt.bar(
    products,
    sales,
    color="royalblue"
)

plt.bar_label(
    bars,
    padding=3
)

plt.title("Product Sales")

plt.grid(axis="y", alpha=0.3)

plt.show()
````

# 2. Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables.

It is one of the most important plots in Data Analysis and Machine Learning.

---

## Best Use Cases

- Correlation Analysis
- Outlier Detection
- Machine Learning
- Data Exploration
- Pattern Recognition

---

## When NOT to Use

- Categorical comparison → Use Bar Plot
- Distribution of data → Use Histogram
- Trend over time → Use Line Plot

---

## Syntax

```python
plt.scatter(x, y)
```

```python
plt.scatter(
    x,
    y,
    s=80,
    c="royalblue",
    marker="o",
    alpha=0.8,
    cmap="viridis",
    label="Students"
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| x | X values |
| y | Y values |
| s | Marker size |
| c | Marker color or values for colormap |
| marker | Marker style |
| alpha | Transparency |
| cmap | Colormap |
| label | Legend label |

````python
hours = [1,2,3,4,5,6,7]

marks = [35,42,50,63,74,88,95]

plt.scatter(hours, marks)

plt.show()
````

````python
hours = [1, 2, 3, 4, 5, 6, 7]
marks = [35, 42, 50, 63, 74, 88, 95]

plt.figure(figsize=(8, 5))

plt.scatter(
    hours,
    marks,
    s=40,
    c="royalblue",
    marker="o",
    alpha=0.8,
    label="Students"
)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.grid(alpha=0.3)
plt.legend()

plt.show()
````

Instead of giving one color, you can provide a list of values.

Matplotlib automatically assigns colors using a colormap.

````python
hours = [1, 2, 3, 4, 5, 6, 7]

marks = [35, 42, 50, 63, 74, 88, 95]

attendance = [60, 65, 72, 80, 84, 90, 98]

plt.figure(figsize=(8, 5))

scatter = plt.scatter(
    hours,
    marks,
    c=attendance,
    s=100,
    cmap="viridis"
)

plt.colorbar(scatter, label="Attendance (%)")

plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.title("Scatter Plot with Color Mapping")

plt.show()
````

````python
boys_hours = [1, 2, 3, 4, 5]
boys_marks = [40, 52, 61, 72, 82]

girls_hours = [1, 2, 3, 4, 5]
girls_marks = [45, 58, 70, 81, 92]

plt.figure(figsize=(8, 5))

plt.scatter(
    boys_hours,
    boys_marks,
    color="royalblue",
    s=50,
    label="Boys"
)

plt.scatter(
    girls_hours,
    girls_marks,
    color="crimson",
    s=50,
    label="Girls"
)

plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.title("Study Hours Comparison")

plt.legend()

plt.grid(alpha=0.3)

plt.show()
````

````python
import numpy as np

np.random.seed(42)

hours = np.random.randint(1, 9, 60)

marks = hours * 10 + np.random.randint(-8, 9, 60)

attendance = np.random.randint(60, 101, 60)

plt.figure(figsize=(9, 5))

scatter = plt.scatter(
    hours,
    marks,
    c=attendance,
    s=120,
    cmap="viridis",
    alpha=0.8,
    edgecolors="black"
)

plt.colorbar(scatter, label="Attendance (%)")

plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.title("Student Performance Analysis")

plt.grid(alpha=0.3)

plt.show()
````

# 3. Histogram

A histogram is used to visualize the distribution of numerical data.

Unlike a bar chart, histograms group continuous values into intervals called **bins**.

---

## Best Use Cases

- Marks Distribution
- Salary Distribution
- Age Distribution
- Height & Weight Analysis
- Data Distribution

---

## When NOT to Use

- Comparing categories → Use Bar Plot

- Relationship between two numerical variables → Use Scatter Plot

- Comparing exact values → Use Line Plot or Bar Plot

---

## Syntax

```python
plt.hist(data)
```

```python
plt.hist(
    data,
    bins=10,
    color="royalblue",
    edgecolor="black",
    density=False,
    cumulative=False,
    alpha=0.8,
    histtype="bar",
    label="Marks"
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| data | Numerical values |
| bins | Number of intervals |
| color | Bar color |
| edgecolor | Border color |
| density | Show probability density |
| cumulative | Cumulative histogram |
| alpha | Transparency |
| histtype | Histogram style |
| label | Legend |

````python
marks = [
    45, 52, 61, 75, 68,
    81, 90, 54, 72, 66,
    58, 77, 84, 69, 93
]

plt.hist(marks)

plt.title("Marks Distribution")

plt.show()
````

````python
marks = [
    45, 52, 61, 75, 68,
    81, 90, 54, 72, 66,
    58, 77, 84, 69, 93,
    48, 63, 71, 86, 95
]

plt.figure(figsize=(8, 5))

plt.hist(
    marks,
    bins=6,
    color="royalblue",
    edgecolor="black",
    alpha=0.8,
    label="Marks"
)

plt.title("Student Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.grid(axis="y", alpha=0.3)
plt.legend()

plt.show()
````

`bins` controls how the data is divided into intervals.

- Smaller bins → More detail
- Larger bins → Less detail

````python
import numpy as np

np.random.seed(42)

marks = np.random.randint(35, 101, 200)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)

plt.hist(
    marks,
    bins=5,
    color="steelblue",
    edgecolor="black"
)

plt.title("5 Bins")

plt.subplot(1, 2, 2)

plt.hist(
    marks,
    bins=20,
    color="tomato",
    edgecolor="black"
)

plt.title("20 Bins")

plt.tight_layout()

plt.show()
````

Setting `density=True` normalizes the histogram.

Instead of frequencies, the y-axis represents probability density.

````python
plt.figure(figsize=(8, 5))

plt.hist(
    marks,
    bins=10,
    density=True,
    color="mediumseagreen",
    edgecolor="black"
)

plt.title("Probability Density Histogram")

plt.show()
````

A cumulative histogram shows the running total of observations.

````python
plt.figure(figsize=(8, 5))

plt.hist(
    marks,
    bins=10,
    cumulative=True,
    color="darkorange",
    edgecolor="black"
)

plt.title("Cumulative Histogram")

plt.show()
````

````python
import numpy as np

np.random.seed(42)

exam_scores = np.random.normal(
    loc=70,
    scale=12,
    size=500
)

plt.figure(figsize=(9, 5))

plt.hist(
    exam_scores,
    bins=20,
    color="royalblue",
    edgecolor="white",
    alpha=0.85
)

plt.title("Exam Score Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

plt.grid(axis="y", alpha=0.3)

plt.show()
````

# 4. Pie Chart

A pie chart shows how each category contributes to the whole.

The total of all slices represents **100%**.

---

## Best Use Cases

- Market Share
- Budget Allocation
- Election Results
- Expense Breakdown
- Population Percentage

---

## When NOT to Use

- Too many categories (more than 6–7)

- Exact value comparison → Use Bar Plot

- Time-series data → Use Line Plot

---

## Syntax

```python
plt.pie(x)
```

```python
plt.pie(
    x,
    labels=None,
    colors=None,
    autopct="%1.1f%%",
    explode=None,
    startangle=90,
    shadow=False,
    wedgeprops=None
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| x | Numerical values |
| labels | Slice labels |
| colors | Slice colors |
| autopct | Percentage format |
| explode | Offset slices |
| startangle | Starting angle |
| shadow | Shadow effect |
| wedgeprops | Border customization |

````python
expenses = [40, 25, 20, 15]

categories = [
    "Rent",
    "Food",
    "Transport",
    "Entertainment"
]

plt.pie(expenses)

plt.title("Monthly Expenses")

plt.show()
````

````python
expenses = [40, 25, 20, 15]

categories = [
    "Rent",
    "Food",
    "Transport",
    "Entertainment"
]

plt.figure(figsize=(7, 7))

plt.pie(
    expenses,
    labels=categories,
    autopct="%1.1f%%",
    startangle=90,
    colors=[
        "royalblue",
        "orange",
        "seagreen",
        "crimson"
    ]
)

plt.title("Monthly Expenses")

plt.show()
````

`explode` moves one or more slices away from the center to highlight them.

````python
expenses = [40, 25, 20, 15]

plt.figure(figsize=(7, 7))

plt.pie(
    expenses,
    labels=[
        "Rent",
        "Food",
        "Transport",
        "Entertainment"
    ],
    explode=[0.1, 
             0, 
             0, 
             0
    ],
    autopct="%1.1f%%",
    startangle=90,
    shadow=True
)

plt.title("Exploded Pie Chart")

plt.show()
````

Use `shadow` and `wedgeprops` to improve presentation quality.

````python
expenses = [40, 25, 20, 15]

plt.figure(figsize=(7, 7))

plt.pie(
    expenses,
    labels=[
        "Rent",
        "Food",
        "Transport",
        "Entertainment"
    ],
    autopct="%1.1f%%",
    shadow=True,
    wedgeprops={
        "edgecolor": "white",
        "linewidth": 2
    }
)

plt.title("Expenses")

plt.show()
````

Matplotlib doesn't have a separate donut chart function.

A donut chart is created using `wedgeprops`.

````python
expenses = [40, 25, 20, 15]

plt.figure(figsize=(7, 7))

plt.pie(
    expenses,
    labels= [
    "Rent",
    "Food",
    "Transport",
    "Entertainment"
],
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={
        "width": 0.45,
        "edgecolor": "white"
    }
)

plt.title("Monthly Expenses")

plt.show()
````

````python
market_share = [38, 27, 18, 10, 7]

companies = [
    "Apple",
    "Samsung",
    "Xiaomi",
    "OnePlus",
    "Others"
]

colors = [
    "#4E79A7",
    "#F28E2B",
    "#59A14F",
    "#E15759",
    "#B07AA1"
]

explode = [0.08, 0, 0, 0, 0]

plt.figure(figsize=(8, 8))

plt.pie(
    market_share,
    labels=companies,
    colors=colors,
    explode=explode,
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    wedgeprops={
        "edgecolor": "white",
        "linewidth": 2
    }
)

plt.title("Smartphone Market Share")

plt.show()
````

# 5. Box Plot & Violin Plot

Both Box Plot and Violin Plot are used to understand the distribution of numerical data.

They help identify:
- Spread of data
- Median
- Quartiles
- Outliers
- Distribution shape

---

## Box Plot vs Violin Plot

| Box Plot | Violin Plot |
|----------|-------------|
| Shows summary statistics | Shows full data distribution |
| Easy to read | More detailed |
| Detects outliers | Shows density of data |
| Preferred in reports | Preferred in exploratory analysis |

---

## Best Use Cases

- Salary Analysis
- Exam Marks
- Product Prices
- Machine Learning Feature Analysis
- Outlier Detection

---

## When NOT to Use

- Comparing categories → Bar Plot

- Relationship between variables → Scatter Plot

- Trend over time → Line Plot

## Syntax

```python
plt.boxplot(data)
```

```python
plt.boxplot(
    data,
    labels=None,
    vert=True,
    patch_artist=True,
    showmeans=False,
    showfliers=True
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| data | Numerical data |
| labels | Category labels |
| vert | Vertical/Horizontal |
| patch_artist | Fill box with color |
| showmeans | Display mean |
| showfliers | Show outliers |

````python
marks = [
    45, 52, 61, 63, 68,
    71, 75, 78, 81, 85,
    88, 90, 92, 95
]

plt.boxplot(marks)

plt.title("Student Marks")

plt.show()
````

````python
marks = [
    45, 52, 61, 63, 68,
    71, 75, 78, 81, 85,
    88, 90, 92, 95
]

plt.figure(figsize=(7, 5))

plt.boxplot(
    marks,
    patch_artist=True,
    showmeans=True,
    boxprops={
        "facecolor": "skyblue",
        "edgecolor": "black"
    },
    medianprops={
        "color": "red",
        "linewidth": 2
    }
)

plt.title("Student Marks Distribution")

plt.ylabel("Marks")

plt.grid(alpha=0.3)

plt.show()
````

````python
maths = [55, 61, 68, 72, 75, 81, 85, 90]
science = [48, 58, 66, 73, 77, 82, 89, 94]
english = [60, 64, 69, 74, 78, 84, 87, 91]

plt.figure(figsize=(8, 5))

plt.boxplot(
    [maths, science, english],
    labels=["Maths", "Science", "English"],
    patch_artist=True
)

plt.title("Subject-wise Marks")

plt.ylabel("Marks")

plt.grid(alpha=0.3)

plt.show()
````

### Parts of a Box Plot

- **Median (Q2)** → Middle value (50th percentile)
- **Q1** → 25th percentile
- **Q3** → 75th percentile
- **IQR (Interquartile Range)** → Q3 − Q1
- **Whiskers** → Range of most data points
- **Outliers** → Values far from the rest of the data

## Syntax

```python
plt.violinplot(data)
```

```python
plt.violinplot(
    dataset,
    showmeans=False,
    showmedians=True,
    showextrema=True
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| dataset | Numerical data |
| showmeans | Display mean |
| showmedians | Display median |
| showextrema | Show min & max |

````python
marks = [
    45, 52, 61, 63, 68,
    71, 75, 78, 81, 85,
    88, 90, 92, 95
]

plt.violinplot(marks)

plt.title("Student Marks")

plt.show()
````

````python
marks = [
    45, 52, 61, 63, 68,
    71, 75, 78, 81, 85,
    88, 90, 92, 95
]

plt.figure(figsize=(7, 5))

plt.violinplot(
    marks,
    showmeans=True,
    showmedians=True
)

plt.title("Student Marks Distribution")

plt.ylabel("Marks")

plt.grid(alpha=0.3)

plt.show()
````

````python
maths = [55, 61, 68, 72, 75, 81, 85, 90]
science = [48, 58, 66, 73, 77, 82, 89, 94]
english = [60, 64, 69, 74, 78, 84, 87, 91]

plt.figure(figsize=(8, 5))

plt.violinplot(
    [maths, science, english],
    showmedians=True
)

plt.xticks(
    [1, 2, 3],
    ["Maths", "Science", "English"]
)

plt.title("Subject-wise Marks")

plt.ylabel("Marks")

plt.grid(alpha=0.3)

plt.show()
````

````python
import numpy as np

np.random.seed(42)

salary = np.random.normal(
    loc=50000,
    scale=8000,
    size=300
)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)

plt.boxplot(
    salary,
    patch_artist=True
)

plt.title("Box Plot")

plt.ylabel("Salary")

plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)

plt.violinplot(
    salary,
    showmeans=True,
    showmedians=True
)

plt.title("Violin Plot")

plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()
````

# Interview & Real-World Tips

✅ Use Box Plot to quickly detect outliers.

✅ Use Violin Plot when you want to understand the distribution shape.

✅ Box Plot is preferred in reports because it's easier to read.

✅ Violin Plot is commonly used in Machine Learning and Exploratory Data Analysis (EDA).

✅ Comparing multiple categories side by side becomes much easier with Box/Violin plots.

---

## Common Mistakes

❌ Confusing Box Plot with Histogram.

❌ Interpreting whiskers as minimum and maximum in every case (their meaning depends on the plotting rule).

❌ Using Violin Plot with very small datasets.

❌ Forgetting axis labels and units.
