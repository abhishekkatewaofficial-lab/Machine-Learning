# Figure & Axes

This is the most important concept in Matplotlib.

Almost every professional Matplotlib project uses the Object-Oriented (OO) interface instead of directly calling `plt.*`.

If you understand Figure and Axes, creating complex dashboards becomes much easier.

---

## Think of it Like This

Imagine a photo frame hanging on a wall.

- The **photo frame** is the **Figure**.
- The **photo inside the frame** is an **Axes**.

A Figure can contain one or many Axes.

## Figure

A Figure is the entire canvas or window that holds one or more plots.

It controls things like:

- Overall size
- Resolution (DPI)
- Background
- Saving the figure

A Figure itself does **not** draw graphs.

It only acts as a container.

````python
import matplotlib.pyplot as plt

fig = plt.figure()

plt.show()
````

````python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 5))

plt.show()
````

````python
import matplotlib.pyplot as plt

fig = plt.figure(
    figsize=(8, 5),
    dpi=120
)

plt.show()
````

## Axes

An Axes is the actual plotting area.

Everything you draw happens inside an Axes.

Examples:

- Line Plot
- Scatter Plot
- Bar Plot
- Histogram
- Pie Chart

All are drawn on an Axes.

````python
import matplotlib.pyplot as plt

fig = plt.figure()

ax = fig.add_axes(
    [0.1, 0.1, 0.8, 0.8]
)

plt.show()
````

The list passed to `add_axes()` is:

```python
[left, bottom, width, height]
```

All values range from **0 to 1**.

Example:

```python
[0.1, 0.1, 0.8, 0.8]
```

means:

- Left margin = 10%
- Bottom margin = 10%
- Width = 80%
- Height = 80%

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 6, 5]

fig = plt.figure(figsize=(7, 5))

ax = fig.add_axes(
    [0.1, 0.1, 0.8, 0.8]
)

ax.plot(x, y)

plt.show()
````

````python
#adding lables

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 6, 5]

fig = plt.figure(figsize=(8, 5))

ax = fig.add_axes(
    [0.1, 0.1, 0.8, 0.8]
)

ax.plot(
    x,
    y,
    marker="o"
)

ax.set_title("Monthly Sales")

ax.set_xlabel("Month")

ax.set_ylabel("Revenue")

ax.grid(alpha=0.3)

plt.show()
````

Instead of:

```python
plt.plot(...)
plt.title(...)
plt.xlabel(...)
```

Professionals write:

```python
ax.plot(...)
ax.set_title(...)
ax.set_xlabel(...)
```

because:

- Easier to manage multiple plots
- Cleaner code
- More readable
- Better for dashboards
- Industry standard

# 2. plt.subplots()

`plt.subplots()` is the recommended and most commonly used way to create a Figure and one or more Axes.

Instead of creating a Figure first and then manually adding Axes, this function creates both in a single line.

---

## Syntax

```python
fig, ax = plt.subplots()
```

For multiple plots:

```python
fig, axs = plt.subplots(
    nrows=2,
    ncols=2
)
```

---

## Important Parameters

| Parameter | Description |
|-----------|-------------|
| nrows | Number of rows |
| ncols | Number of columns |
| figsize | Figure size |
| dpi | Resolution |
| sharex | Share X-axis |
| sharey | Share Y-axis |
| constrained_layout | Automatic spacing |

Advantages:

- Creates Figure and Axes together
- Cleaner code
- Easy to create dashboards
- Easy to manage multiple plots
- Industry standard
- Used in almost every professional project

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

print(type(fig))
print(type(ax))
````

````python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5]

sales = [20, 35, 28, 42, 50]

fig, ax = plt.subplots()

ax.plot(months, sales)

plt.show()
````

````python
#Adding Lables

import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5]

sales = [20, 35, 28, 42, 50]

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(
    months,
    sales,
    marker="o",
    linewidth=2
)

ax.set_title("Monthly Sales")

ax.set_xlabel("Month")

ax.set_ylabel("Sales")

ax.grid(alpha=0.3)

plt.show()
````

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(
    figsize=(10, 6)
)

ax.plot([1, 2, 3], [2, 5, 4],marker = 'o')

plt.show()
````

## Creating Multiple Rows

```python
fig, axs = plt.subplots(
    nrows=2,
    ncols=1
)
```

Notice the variable name changes from `ax` to `axs`.

This is because multiple Axes are returned.

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(7, 6)
)

axs[0].plot(
    [1, 2, 3],
    [2, 5, 3]
)

axs[0].set_title("First Plot")

axs[1].plot(
    [1, 2, 3],
    [5, 3, 6]
)

axs[1].set_title("Second Plot")

plt.show()
````

````python
#Multiple columns

import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    nrows=1,
    ncols=2,
    figsize=(9, 4)
)

axs[0].plot(
    [1, 2, 3],
    [4, 5, 3],marker = 'o'
)

axs[0].set_title("Left Plot")

axs[1].plot(
    [1, 2, 3],
    [2, 6, 4],marker = 'o'
)

axs[1].set_title("Right Plot")

plt.show()
````

````python
#2*2 GRID

import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 8)
)

axs[0, 0].plot([1, 2, 3], [2, 4, 3],marker = 'o')
axs[0, 0].set_title("Plot 1")

axs[0, 1].plot([1, 2, 3], [3, 5, 2],marker = 'o')
axs[0, 1].set_title("Plot 2")

axs[1, 0].plot([1, 2, 3], [5, 2, 4],marker = 'o')
axs[1, 0].set_title("Plot 3")

axs[1, 1].plot([1, 2, 3], [2, 6, 5],marker = 'o')
axs[1, 1].set_title("Plot 4")

plt.show()
````

# 3. Shared Axes & Layout Management

When working with multiple plots, it is common for graphs to represent related data.

Instead of giving every subplot its own X-axis or Y-axis, Matplotlib allows multiple plots to share axes.

It also provides automatic layout tools to prevent overlapping labels and titles.

---

## Topics Covered

- sharex
- sharey
- tight_layout()
- constrained_layout
- suptitle()

## Sharing V-axis

sharex=True

When all plots have the same X values, you can share one X-axis.

Benefits:

- Cleaner figure
- Less repetition
- Easier comparison

````python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5]

sales = [15, 22, 30, 27, 35]
profit = [2, 5, 7, 6, 9]

fig, axs = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(8, 6),
    sharex=True
)

axs[0].plot(months, sales, marker="o")
axs[0].set_title("Sales")

axs[1].plot(months, profit, marker="o")
axs[1].set_title("Profit")

axs[0].grid(axis='x')
axs[1].grid(axis='x')

plt.show()
````

## Sharing Y-axis 

sharey=True

When multiple graphs use the same scale, sharing the Y-axis makes comparison easier.

````python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5]

store_a = [12, 18, 24, 30, 36]
store_b = [10, 15, 22, 28, 33]

fig, axs = plt.subplots(
    nrows=1,
    ncols=2,
    figsize=(10, 4),
    sharey=True
)

axs[0].plot(months, store_a, marker="o")
axs[0].set_title("Store A")

axs[1].plot(months, store_b, marker="o")
axs[1].set_title("Store B")

plt.show()
````

````python
# SHARING BOTH AXES

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

y1 = [2, 4, 5, 4, 6]
y2 = [3, 5, 4, 6, 7]
y3 = [1, 3, 2, 5, 4]
y4 = [4, 6, 5, 7, 8]

fig, axs = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(10, 8),
    sharex=True,
    sharey=True
)

axs[0,0].plot(x, y1)
axs[0,1].plot(x, y2)
axs[1,0].plot(x, y3)
axs[1,1].plot(x, y4)

plt.show()
````

## tight_layout()

Sometimes titles, labels, and ticks overlap.

`tight_layout()` automatically adjusts spacing between subplots.

````python
#WITHOUT TIGHT LAYOUT
import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    2,
    2,
    figsize=(6, 4)
)

for ax in axs.flat:
    ax.plot([1,2,3],[2,5,4])
    ax.set_title("Very Long Title")

plt.show()
````

````python
##WITH TIGHT LAYOUT
import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(6, 4)
)

for ax in axs.flat:
    ax.plot([1,2,3],[2,5,4])
    ax.set_title("Very Long Title")

plt.tight_layout()

plt.show()
````

## constrained_layout=True

This is a newer layout engine.

Instead of calling `tight_layout()` manually, Matplotlib adjusts the layout automatically while creating the figure.

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    2,
    2,
    figsize=(8,6),
    constrained_layout=True
)

for ax in axs.flat:
    ax.plot([1,2,3],[4,5,3])
    ax.set_title("Sales Report")

plt.show()
````

## suptitle()

Each subplot can have its own title.

Sometimes the entire figure also needs one main title.

Use:

```python
fig.suptitle()
```

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(8,6),
    constrained_layout=True
)

for i, ax in enumerate(axs.flat, start=1):
    ax.plot([1,2,3],[i+1, i+3, i+2])
    ax.set_title(f"Plot {i}")

fig.suptitle(
    "Company Performance Dashboard",
    fontsize=16
)

plt.show()
````

````python
import matplotlib.pyplot as plt

months = [1,2,3,4,5,6]

sales = [22,26,31,34,40,45]
profit = [3,5,6,7,9,11]
customers = [110,135,150,170,195,220]
expenses = [18,20,25,27,31,34]

fig, axs = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(11,7),
    sharex=True,
    constrained_layout=True
)

axs[0,0].plot(months, sales, marker="o")
axs[0,0].set_title("Sales")

axs[0,1].plot(months, profit, marker="o")
axs[0,1].set_title("Profit")

axs[1,0].plot(months, customers, marker="o")
axs[1,0].set_title("Customers")

axs[1,1].plot(months, expenses, marker="o")
axs[1,1].set_title("Expenses")

fig.suptitle(
    "Business Dashboard",
    fontsize=18,
    color='blue'
)

plt.show()
````

# 4. Figure-level vs Axes-level Methods

In Matplotlib, some methods belong to the Figure, while others belong to an Axes.

Knowing where a method belongs helps you write cleaner and more professional code.

---

## Rule of Thumb

If something affects the **entire figure**, use **fig**.

If something affects **only one graph**, use **ax**.

## Common Figure Methods

| Method | Purpose |
|---------|---------|
| fig.suptitle() | Main title for the entire figure |
| fig.savefig() | Save the figure |
| fig.tight_layout() | Adjust spacing (or use plt.tight_layout()) |
| fig.set_facecolor() | Change figure background |

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7, 5))

ax.plot([1, 2, 3], [2, 5, 4])

fig.suptitle("Sales Dashboard")

plt.show()
````

## Common Axes Methods

| Method | Purpose |
|---------|---------|
| ax.plot() | Line plot |
| ax.scatter() | Scatter plot |
| ax.bar() | Bar chart |
| ax.hist() | Histogram |
| ax.boxplot() | Box plot |
| ax.set_title() | Title |
| ax.set_xlabel() | X-axis label |
| ax.set_ylabel() | Y-axis label |
| ax.set_xlim() | X-axis range |
| ax.set_ylim() | Y-axis range |
| ax.set_xticks() | X-axis ticks |
| ax.set_yticks() | Y-axis ticks |
| ax.legend() | Legend |
| ax.grid() | Grid |

````python
#Setting title
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1, 2, 3], [4, 6, 5])

ax.set_title("Monthly Sales")

plt.show()
````

````python
#Setting X,Y Lables
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1, 2, 3], [2, 4, 3])

ax.set_xlabel("Month")

ax.set_ylabel("Revenue")

plt.show()
````

````python
#Setting tick loation
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5], [2, 5, 4, 6, 5])

ax.set_xticks([1, 2, 3, 4, 5])

plt.show()
````

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 5], [20, 30, 28, 35, 40])

ax.set_yticks([0, 10, 20, 30, 40, 50])

plt.show()
````

````python
#Custom Tick Lables
import matplotlib.pyplot as plt

months = [1, 2, 3, 4]

sales = [25, 32, 30, 38]

fig, ax = plt.subplots()

ax.plot(months, sales, marker="o")

ax.set_xticks(months)

ax.set_xticklabels(
    ["Jan", "Feb", "Mar", "Apr"]
)

plt.show()
````

````python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5, 6]

sales = [20, 24, 30, 28, 36, 42]

fig, ax = plt.subplots(
    figsize=(9, 5)
)

ax.plot(
    months,
    sales,
    marker="o",
    linewidth=2,
    label="Sales"
)

ax.set_title("Monthly Sales Report")

ax.set_xlabel("Month")

ax.set_ylabel("Sales")

ax.set_xlim(1, 6)

ax.set_ylim(0, 50)

ax.set_xticks(months)

ax.set_xticklabels(
    ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
)

ax.set_yticks(
    [0, 10, 20, 30, 40, 50]
)

ax.legend()

ax.grid(alpha=0.3)

plt.show()
````

### Q1. Why is `ax.set_title()` preferred over `plt.title()`?

**Answer:**
Because it modifies a specific Axes instead of relying on the current active plot, making the code clearer and more reliable for multiple subplots.

---

### Q2. When should you use `fig.suptitle()`?

**Answer:**
When you need a single title for the entire figure that contains one or more subplots.

---

### Q3. What is the difference between `set_xlim()` and `set_xticks()`?

- `set_xlim()` controls the visible range of the X-axis.
- `set_xticks()` controls where tick marks are displayed within that range.

---

### Q4. What is the difference between `set_xticks()` and `set_xticklabels()`?

- `set_xticks()` defines the tick positions.
- `set_xticklabels()` defines the text shown at those positions.
