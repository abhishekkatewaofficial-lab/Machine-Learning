# LIST

```python
li = [9,1,8,2,7, 3,6,4,5]

new_list = sorted(li)
new_list
```

```python
li.sort()
li
```

Here sorted(li) is a function and
     li.sort() is a method

     so function will return value and
     method will change original list

     thats why function can be set as variable whereas method cant.
