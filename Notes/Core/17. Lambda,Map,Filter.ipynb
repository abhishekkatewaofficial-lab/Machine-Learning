# Lambda Function

A lambda function is an anonymous (nameless) function.

Instead of writing a complete function using `def`, we can write it in a single line using `lambda`.

## Syntax

```python
lambda arguments : expression
```

The expression is automatically returned.

````python
square = lambda x: x * x

print(square(5))
````

````python
add = lambda a, b: a + b

print(add(10, 20))
````

Lambda functions are mostly used for small, one-time operations.

For complex logic, use a normal function (`def`).

# map()

The `map()` function applies a function to every element of an iterable.

## Syntax

```python
map(function, iterable)
```

It returns a map object.

````python
numbers = [1, 2, 3, 4, 5]

result = map(lambda x: x * x, numbers)

print(result)
````

````python
numbers = [1, 2, 3, 4, 5]

result = map(lambda x: x * x, numbers)

print(list(result))
````

````python
names = ["abhi", "rahul", "aman"]

capital = map(str.capitalize, names)

print(list(capital))
````

# filter()

The `filter()` function selects only those elements that satisfy a condition.

## Syntax

```python
filter(function, iterable)
```

The function must return `True` or `False`.

````python
numbers = [1, 2, 3, 4, 5, 6]

even = filter(lambda x: x % 2 == 0, numbers)

print(list(even))
````

````python
numbers = [10, 25, 30, 45, 60]

greater = filter(lambda x: x > 30, numbers)

print(list(greater))
````

# map() vs filter()

`map()` transforms every element.

`filter()` selects some elements.

Example:

Original List

```python
[1,2,3,4,5]
```

Using map()

```python
[1,4,9,16,25]
```

Using filter()

```python
[2,4]
```

````python
numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x * x, numbers))

evens = list(filter(lambda x: x % 2 == 0, numbers))

print("Squares:", squares)
print("Even:", evens)
````

# Combining map() and filter()

We can first filter data and then transform it.

````python
numbers = [1, 2, 3, 4, 5, 6]

result = map(
    lambda x: x * x,
    list(filter(lambda x: x % 2 == 0, numbers))
)

print(list(result))
````

# List Comprehension Alternative

Many Python developers prefer list comprehensions because they are often easier to read.

````python
numbers = [1, 2, 3, 4, 5]

squares = [x * x for x in numbers]

print(squares)
````

# Difference Table

| Lambda | map() | filter() |
|---------|--------|-----------|
| Anonymous function | Applies a function to every element | Selects elements based on a condition |
| Single expression | Returns transformed values | Returns matching values |
| Uses `lambda` frequently | Returns map object | Returns filter object |

# Interview Notes

## Lambda

- Anonymous function
- One-line function
- Uses `lambda`
- Automatically returns the expression

## map()

- Applies a function to every element
- Returns a map object
- Convert to a list using `list()`

## filter()

- Filters elements based on a condition
- Function should return `True` or `False`
- Returns a filter object

## Which is Better?

- Use `lambda` for short operations.
- Use `map()` for transformations.
- Use `filter()` for selecting data.
- Use list comprehensions when readability is better.

````python
# One-line Examples

square = lambda x: x * x

print(square(5))

print(list(map(lambda x: x * x, [1, 2, 3])))

print(list(filter(lambda x: x % 2 == 0, [1, 2, 3, 4, 5, 6])))
````
