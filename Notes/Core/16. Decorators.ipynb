# Python Decorators

## What is a Decorator?

A Decorator is a function that modifies or extends the behavior of another function without changing its original code.

In Python, functions are first-class objects, meaning:

- Functions can be assigned to variables.
- Functions can be passed as arguments.
- Functions can be returned from other functions.

Decorators use this feature to wrap existing functions.

### Real-Life Analogy

Imagine a gift 🎁.

- Original function = Gift
- Decorator = Gift Wrapper

The gift remains the same, but extra functionality is added around it.

### Syntax

```python
@decorator_name
def function_name():
    pass
```

This is equivalent to:

```python
function_name = decorator_name(function_name)
```

## Functions are Objects

Before understanding decorators, let's verify that functions are objects in Python.

````python
def greet():
    print("Hello")

say_hello = greet

say_hello()
````

## Function Passed as Argument

A function can be passed to another function.

````python
def greet():
    print("Hello")

def execute(func):
    func()

execute(greet)
````

## Function Returning Another Function

A function can also return another function.

````python
def outer():

    def inner():
        print("Inside Inner Function")

    return inner

result = outer()

result()
````

# Creating Our First Decorator

Structure:

1. Decorator receives a function.
2. Creates a wrapper function.
3. Adds extra behavior.
4. Returns wrapper.

````python
def decorator(func):

    def wrapper():
        print("Before Function")

        func()

        print("After Function")

    return wrapper


def greet():
    print("Hello")


greet = decorator(greet)

greet()
````

## Using @ Syntax

Python provides a cleaner syntax using `@`.

````python
def decorator(func):

    def wrapper():
        print("Before Function")

        func()

        print("After Function")

    return wrapper


@decorator
def greet():
    print("Hello")


greet()
````

# Decorator with Arguments

Problem:

The previous decorator only works with functions having no arguments.

Solution:

Use `*args` and `**kwargs`.

````python
def decorator(func):

    def wrapper(*args, **kwargs):

        print("Before Function")

        func(*args, **kwargs)

        print("After Function")

    return wrapper


@decorator
def add(a, b):
    print(a + b)


add(10, 20)
````

# Returning Values from Decorated Functions

If the original function returns something, the wrapper should also return it.

````python
def decorator(func):

    def wrapper(*args, **kwargs):

        result = func(*args, **kwargs)

        return result

    return wrapper


@decorator
def add(a, b):
    return a + b


print(add(5, 7))
````

# Practical Example: Execution Timer

A common real-world use of decorators is measuring execution time.

````python
import time

def timer(func):

    def wrapper(*args, **kwargs):

        start = time.time()

        result = func(*args, **kwargs)

        end = time.time()

        print("Execution Time:", end - start)

        return result

    return wrapper


@timer
def work():

    time.sleep(2)

    print("Task Completed")


work()
````

# Multiple Decorators

Decorators can be stacked.

````python
def decorator1(func):

    def wrapper():
        print("Decorator 1")

        func()

    return wrapper


def decorator2(func):

    def wrapper():
        print("Decorator 2")

        func()

    return wrapper


@decorator1
@decorator2
def greet():
    print("Hello")


greet()
````
