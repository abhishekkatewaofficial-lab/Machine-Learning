# Python File Handling

Files are used to store data permanently on disk.

Python provides built-in support for:

- Reading files
- Writing files
- Appending data
- Reading large files efficiently
- Working with binary files (images, videos, etc.)

The built-in function used is:

```python
open()

file_object = open(filename, mode)

Common Modes:
Mode	Description
'r'	Read (default)
'w'	Write (creates/overwrites file)
'a'	Append
'r+'	Read and Write
'rb'	Read Binary
'wb'	Write Binary

````python
file = open("test.txt", "r")

print(file.name)
print(file.mode)

file.close()
````

````python
with open("test.txt", "r") as file:
    pass

print(file.closed)
````

# Reading Entire File

````python
with open("test.txt", "r") as file:
    contents = file.read()

print(contents)
````

⚠️ Problem:

For very large files, `read()` loads everything into memory.

This can cause memory issues.

# Reading All Lines

````python
with open("test.txt", "r") as file:
    lines = file.readlines()

print(lines)
````

# Reading One Line at a Time

````python
with open("test.txt", "r") as file:
    print(file.readline(), end="")
    print(file.readline(), end="")
    print(file.readline(), end="")
````

# Efficient Way: Iterate Through File

Best for large text files.

````python
with open("test.txt", "r") as file:
    for line in file:
        print(line, end="")
````

Advantages:
Memory efficient
Reads one line at a time

# Reading Fixed Number of Characters

````python
with open("test.txt", "r") as file:
    data = file.read(100)

print(data)
````

# File Pointer

Whenever data is read, Python moves an internal cursor (pointer).

Current position can be checked using:

```python
f.tell()

````python
with open("test.txt", "r") as file:
    file.read(10)
    print(file.tell())
````

# Reading Large Files in Chunks

Useful when file size is huge.

````python
with open("test.txt", "r") as file:

    #size_to_read = 10

    contents = file.read(10)

    while len(contents) > 0:
        print(contents, end="*")
        contents = file.read(10)
````

# Moving File Pointer

Use:

```python
f.seek(position)

````python
with open("test.txt", "r") as file:

    print(file.read(10))

    file.seek(0)

    print(file.read(10))
````

# Writing to a File

Mode:

```python
'w'

````python
with open("test2.txt", "w") as file:
    file.write("Hello")
    file.write(" Python")
````

# Appending to a File

Mode:

```python
'a'

Does not overwrite existing content.

````python
with open("test2.txt", "a") as f:
    f.write("\nNew Line")
````

# Using seek() While Writing

````python
with open("test2.txt", "w") as f:

    f.write("TEST")

    f.seek(0)

    f.write("R")
````

# Copying a Text File

````python
with open("test.txt", "r") as rf:
    with open("test_copy.txt", "w") as wf:

        for line in rf:
            wf.write(line)
````

# Interview / Production Takeaways

1. Prefer `with open()` over manual close.
2. Use iteration for large text files.
3. Use chunk reading for huge files.
4. Use binary mode for images/videos/PDFs.
5. `tell()` gives current cursor position.
6. `seek()` moves cursor to a specific position.
7. `w` overwrites files.
8. `a` appends data.
9. `read()` loads entire file into memory.
10. Chunk-based processing is the most scalable approach.
