# Understanding csv.reader()
Creates an iterator that reads rows from a CSV file.

```python
import csv

with open("iaf_officers.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

# Skipping Header Row


```python
import csv

with open("iaf_officers.csv") as file:
    reader = csv.reader(file)

    next(reader)

    for row in reader:
        print(row)
```

# Writing to a CSV File

```python
import csv

with open("new_file.csv", "w") as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Age", "City"])
    writer.writerow(["Abhi", 20, "Mumbai"])
    writer.writerow(["Summu" , 23 , "Pune"])
    writer.writerow(["Kittu" , 1.5 , "Persia"])
```

# Writing Multiple Rows

```python
import csv

data = [
    ["Name", "Age", "City"],
    ["Abhi", 20, "Mumbai"],
    ["Rahul", 21, "Pune"],
    ["Priya", 22, "Delhi"]
]

with open("new_file.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerows(data)
```

# csv.DictReader()

Reads CSV as dictionaries.

```python
import csv

with open("iaf_officers.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row)
```

# Accessing Specific Columns

```python
import csv

with open("iaf_officers.csv") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row["Name"])
```

# csv.DictWriter()

Writes dictionaries into CSV files.

```python
import csv

with open("employees.csv", "w", newline="") as file:

    fields = ["Name", "Salary"]

    writer = csv.DictWriter(file, fieldnames=fields)

    writer.writeheader()

    writer.writerow({
        "Name": "Abhi",
        "Salary": 50000
    })
```

# Appending Data to CSV

```python
import csv

with open("new_file.csv", "a", newline="") as file:

    writer = csv.writer(file)

    writer.writerow(["Aman", 23, "Nagpur"])
```

# Reading CSV into a List

```python
import csv

with open("new_file.csv") as file:
    reader = csv.reader(file)

    data = list(reader)

print(data)
```
