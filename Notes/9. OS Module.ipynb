```python
import datetime
```

```python
bday = datetime.date(2001 , 7  ,5)

print(bday)
```

```python
today = datetime.date.today()
print(today)
```

```python
till_day = today - bday
print(till_day)
```

```python
bday = datetime.datetime(2001 , 7  ,5 , 4 , 15 , 00)
print(bday)
```

```python
today = datetime.datetime.today()
print(today)
```

```python
till_day = today-bday
print(till_day)
```

```python
now = datetime.datetime.now()
print(now)
```
