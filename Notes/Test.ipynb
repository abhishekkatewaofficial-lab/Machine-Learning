```python
words = ['Mon', 'Tue', 'Wed']
nums = [1, 2, 3]
# Use zip to pack into a tuple list
for w, n in zip(words, nums):
    print('%d:%s, ' %(n, w))
```

```python
def varargs(*args):
    return args

varargs(1, 2, 3)  # => (1, 2, 3)
```

```python
def keyword_args(**kwargs):
    return kwargs

# => {"big": "foot", "loch": "ness"}
keyword_args(big="foot", loch="ness")
```

```python
def add(x, y=10):
    return x + y

add(5)
```
