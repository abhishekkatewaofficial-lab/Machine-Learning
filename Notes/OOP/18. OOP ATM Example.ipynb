````python
class ATM:

    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.menu()

    def menu(self):
        choice = input("""
Hi! How can I help you?

1. Create PIN
2. Change PIN
3. Check Balance
4. Withdraw Money
5. Exit

Enter your choice: """)

        if choice == "1":
            self.create_pin()
        elif choice == "2":
            self.change_pin()
        elif choice == "3":
            self.check_balance()
        elif choice == "4":
            self.withdraw()
        else:
            print("Thank you for using our ATM.")
            return

    def create_pin(self):
        self.pin = input("Create your PIN: ")
        self.balance = int(input("Enter initial balance: "))

        print("PIN created successfully.")
        self.menu()

    def change_pin(self):
        old_pin = input("Enter old PIN: ")

        if old_pin == self.pin:
            self.pin = input("Enter new PIN: ")
            print("PIN changed successfully.")
        else:
            print("Incorrect PIN.")

        self.menu()

    def check_balance(self):
        entered_pin = input("Enter your PIN: ")

        if entered_pin == self.pin:
            print(f"Your balance is ₹{self.balance}")
        else:
            print("Incorrect PIN.")

        self.menu()

    def withdraw(self):
        entered_pin = input("Enter your PIN: ")

        if entered_pin == self.pin:
            amount = int(input("Enter withdrawal amount: "))

            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrawal successful. Remaining balance: ₹{self.balance}")
            else:
                print("Insufficient balance.")
        else:
            print("Incorrect PIN.")

        self.menu()


atm = ATM()
````

# Step 8 : Change PIN

Algorithm

1. Ask old PIN
2. Verify
3. If correct
    - Ask new PIN
4. Otherwise
    - Show error

````python
def change_pin(self):

    old_pin = input("Enter old PIN : ")

    if old_pin == self.pin:

        self.pin = input("Enter new PIN : ")

        print("PIN Changed Successfully")

    else:

        print("Incorrect PIN")

    self.menu()
````

# Thought Process


# Step 9 : Check Balance

Algorithm

1. Ask PIN
2. Verify
3. Show balance

````python
def check_balance(self):

    entered_pin = input("Enter PIN : ")

    if entered_pin == self.pin:

        print(self.balance)

    else:

        print("Incorrect PIN")

    self.menu()
````

# Step 10 : Withdraw Money

Algorithm

1. Verify PIN
2. Ask amount
3. Check sufficient balance
4. Deduct amount
5. Print remaining balance

````python
def withdraw(self):

    entered_pin = input("Enter PIN : ")

    if entered_pin == self.pin:

        amount = int(input("Enter Amount : "))

        if amount <= self.balance:

            self.balance -= amount

            print("Remaining Balance:", self.balance)

        else:

            print("Insufficient Balance")

    else:

        print("Incorrect PIN")

    self.menu()
````

# Interview Discussion: How to Improve This Design

For learning OOP, this implementation is good.

In a production system, improvements would include:

- Replace recursive `self.menu()` calls with a `while True` loop to avoid growing the call stack.
- Store the PIN securely (e.g., hashed), not as plain text.
- Validate inputs (numeric choices, positive amounts, PIN format).
- Prevent overdrafts and handle invalid withdrawal amounts.
- Separate business logic from the user interface (input/output) to make the code easier to test.
- Persist account data in a database instead of keeping it only in memory.
- Support multiple users/accounts instead of a single ATM state.
