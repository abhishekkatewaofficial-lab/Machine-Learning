# Step 1 : The Procedural Way

Imagine we are building a Student Management System.

Each student has

State

- Name
- Age
- CGPA

and can perform

Behavior

- Study
- Sleep
- Give Exam

A beginner usually writes

````python
student_name = "Abhi"
student_age = 20
student_cgpa = 8.8


def study():
    print(f"{student_name} is studying.")


def sleep():
    print(f"{student_name} is sleeping.")


def give_exam():
    print(f"{student_name} is giving an exam.")
````

Everything works.

Current Memory

```

------------------

↓

------------------

↓

8.8

------------------

sleep()

give_exam()
```

Notice something.

The data and the functions are completely separate.

````python
study()
sleep()
give_exam()
````

# Step 3 : Another Problem

Suppose

Which function should execute?

```

```

````python
student_name = "Abhi"

study()

student_name = "Rahul"

study()
````

Notice what happened.

Instead of keeping

```

```

```

+
```

# Step 5 : Introducing an Object

Instead of writing

```

```

Let's imagine one thing called

```

```

```

│

├── Name

├── Age

├── Study()

├── Sleep()

└── Give Exam()
```

This single entity is called an

**Object**.

# What is an Object?

An Object is a real-world entity that combines

State (Data)

and

Behavior (Functions)

into one unit.

Example

```
Student

State

Name

Age

CGPA

----------------------

Behavior

Study()

Sleep()

Give Exam()
```

Everything related to one student stays together.

# Interview Thought Process

If an interviewer asks

"What is an Object?"

Don't answer

"Object is an instance of a class."

That's technically correct,

but it doesn't explain anything.

A better answer is:

"An object is a real-world entity that bundles together state (data) and behavior (methods) into a single unit."

Example

A Student object stores

- Name
- Age
- CGPA

and knows how to

- Study
- Sleep
- Give Exam.

# Object-Oriented Programming (OOP) - Part 3

# Understanding `self`

If you ask beginners,

"What is the most confusing thing in OOP?"

Most will answer

```
self
```

Today's goal is NOT to memorize

```
self = current object
```

Instead,

we'll understand

- Why `self` exists
- Why Python needs it
- What happens internally
- Why we don't pass it manually

# Step 1 : Let's Create an Empty Class

We already know

```
Class

↓

Object
```

Let's create one object.

````python
class student:
    pass

student1 = student()
````

Question

What if every student should be able to introduce themselves?

We need a method.

Example

```
introduce()
```

Let's add one.

````python
class student:

    def introduce():

        print("Hello World")
````

Now create an object.

````python
student1 = student()
student1.introduce()
````

TypeError:

introduce()

takes 0 positional arguments

but 1 was given

Question

We didn't pass any argument.

Then where did

```
1 argument
```

come from?

This is the mystery we'll solve.

# Step 2 : Python Does Something Behind the Scenes

You wrote

```python
student1.introduce()
```

You think Python executes

```
introduce()
```

Actually,

Python secretly converts it into

```python
Student.introduce(student1)
```

Notice

Python automatically passes

```
student1
```

to the method.

Visual Representation

You write

```
student1.introduce()
```

Python internally performs

```
Student.introduce(student1)
```

The object is automatically sent as the first argument.

Now everything makes sense.

Our method

```python
def introduce():
```

accepts

```
0 arguments
```

But Python sends

```
student1
```

Hence

```
TypeError
```

# Step 3 : Receive That Object

If Python automatically sends

```
student1
```

then our method should receive it.

Let's add one parameter.

````python
class Student:

    def introduce(self):

        print("Hello")


student1 = Student()

student1.introduce()
````

Question

What is

```
self
```

It is simply the parameter that receives

the object Python automatically passes.

Visual Representation

```
student1.introduce()

        │

        ▼

Python converts into

Student.introduce(student1)

        │

        ▼

self

↓

student1
```

````python
class Student:

    def introduce(self):

        print(self)


student1 = Student()

student1.introduce()
````

Look carefully.

```
self
```

is

the Student object.

Nothing more.

Nothing less.

# Step 4 : Multiple Objects

Question

What happens if we have

```
student1

student2

student3
```

Will

```
self
```

always refer to

```
student1
```

No.

It always refers to

**the object that called the method.**

````python
class Student:

    def introduce(self):

        print(self)


student1 = Student()

student2 = Student()

student1.introduce()

student2.introduce()
````

Visual Representation

```
student1.introduce()

↓

self

↓

student1

----------------------------

student2.introduce()

↓

self

↓

student2
```

There is no fixed object.

`self` changes automatically depending on who called the method.

# Step 5 : Is the Name `self` Special?

No.

This also works.

```python
def introduce(current_student):
```

Python doesn't care about the name.

It only cares that the **first parameter** receives the object.

However,

every Python developer uses

```
self
```

as the standard convention.

````python
class Student:

    def introduce(current_student):

        print(current_student)


student = Student()

student.introduce()
````

Although this works,

never write code like this in interviews or production.

Always use

```python
self
```

because it's the Python convention.

# Interview Thought Process

Suppose Google asks:

"What is `self`?"

A beginner answers:

"It refers to the current object."

A stronger answer is:

"`self` is the first parameter of an instance method. Python automatically passes the object that invoked the method as this first argument, and by convention we name that parameter `self`."

This answer explains both **what** `self` is and **why** it exists.

# End of Part 3

Today we learned

✅ Why `self` exists.

✅ Python automatically passes the calling object.

✅ `self` receives that object.

✅ `self` is not a keyword.

✅ `self` changes depending on which object calls the method.
