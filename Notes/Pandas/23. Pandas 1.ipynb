# What is Pandas?

**Pandas** is an open-source Python library used for working with structured and tabular data.

It provides powerful and efficient data structures that make data cleaning, transformation, analysis, and manipulation simple.

The two most important data structures in Pandas are:

- **Series** → One-dimensional labeled array.
- **DataFrame** → Two-dimensional labeled table.

Think of Pandas as **Excel on steroids, but with Python**.

# Why was Pandas Created?

Imagine you have a CSV file containing one million employee records.

Using only Python lists and dictionaries, tasks like:

- filtering rows
- sorting
- grouping
- finding averages
- handling missing values
- merging datasets

become slow, lengthy, and difficult to maintain.

Pandas was created to solve these problems efficiently while providing a clean and expressive API.

# Why Not Pure Python?

Python can work with data using lists and dictionaries, but it becomes inconvenient for large datasets.

Example tasks:

- Read CSV
- Remove missing values
- Find average salary department-wise
- Merge employee and department tables
- Calculate statistics

These operations require many loops and manual logic in pure Python.

Pandas performs the same tasks using optimized C/NumPy implementations, resulting in cleaner code and much better performance.

# What is a Pandas Series?

A **Series** is a one-dimensional labeled array capable of storing any data type.

Each element in a Series has:

- A value
- An associated index (label)

Unlike Python lists, every value in a Series is mapped to an index.

Think of it as:

Index  →  Value

# Visual Representation

        Series

Index      Value

0          15

1          28

2          34

3          40

4          52

Every value has a corresponding index.

# Why was Series Created?

Python lists are excellent for storing values, but they lack several features required for data analysis.

For example:

- Labels
- Automatic alignment
- Missing value handling
- Fast vectorized operations
- Statistical methods

Series provides all of these in a single data structure.

# Internal Structure of Series

Conceptually, a Series consists of three main components.

1. Data
2. Index
3. Data Type (dtype)

               Series

        +-----------------+

Data  → | 10 20 30 40 50 |

Index → | 0  1  2  3  4  |

dtype → int64

Every Series always contains these components.

# Characteristics of Series

- One-dimensional
- Mutable
- Labeled
- Homogeneous dtype (mostly)
- Fast vectorized operations
- Supports missing values

```python
numbers = pd.Series([10, 20, 30, 40, 50])

numbers
```

# Creating Series from a Tuple

```python
marks = pd.Series((85, 90, 78, 92, 88))

marks
```

# Creating Series from a NumPy Array

```python
import numpy as np  

arr = np.array([5,10,15,20])

series = pd.Series(arr)

series
```

# Creating Series from a Dictionary

Dictionary keys automatically become the index.

```python
student = {
    "Math": 95,
    "Science": 90,
    "English": 88
}

marks = pd.Series(student)

marks
```

# Creating Series from a Scalar

A scalar is a single value.

When creating a Series from a scalar, an index must also be provided.

```python
age = pd.Series(21, index=["Alice", "Bob", "Charlie"])

age
```

# Custom Index

```python
temperature = pd.Series(
    [31, 34, 29, 36],
    index=["Mon", "Tue", "Wed", "Thu"]
)

temperature
```

# Duplicate Index

Pandas allows duplicate index labels.

```python
data = pd.Series(
    [100, 200, 300],
    index=["A", "A", "B"]
)

data
```

# Can a Series Store Mixed Data Types?

Yes.

But the dtype usually becomes `object`, which is slower than numeric dtypes.

Whenever possible, keep the Series homogeneous.

# Interview Question

Q. Why is Series faster than a Python list for numerical operations?

Answer:

Because Series is built on top of NumPy arrays, which store homogeneous data in contiguous memory and perform vectorized operations implemented in optimized C code.

# Common Mistakes

❌ Forgetting to provide an index while creating a scalar Series.

❌ Mixing multiple data types unnecessarily.

❌ Assuming Series behaves exactly like a Python list.

❌ Thinking dictionary values become the index (they do not).
