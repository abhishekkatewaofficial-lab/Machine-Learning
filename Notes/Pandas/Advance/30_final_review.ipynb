## 1. I/O & Creation
```python
df = pd.read_csv('file.csv', usecols=['id', 'value'], dtype={'id': 'int32'})
df = pd.read_parquet('file.parquet')
df.to_csv('file.csv', index=False)
```

## 2. Exploration
```python
df.shape
df.info()
df.describe()
df['col'].value_counts(normalize=True, dropna=False)
df.isna().sum()
```

## 3. Filtering & Selection
```python
df[['col1', 'col2']]               # select columns
df.loc[df['col'] > 0, 'col2']      # select rows/cols by condition
df.query("col1 > 0 and col2 == 'A'") # clean filtering
df.nlargest(5, 'col')              # top 5
```

## 4. Cleaning & Modification
```python
df.rename(columns={'old': 'new'})  # rename
df.drop(columns=['col'])           # drop cols
df.dropna(subset=['col'])          # drop NaNs
df['col'].fillna(0)                # fill NaNs
df.assign(new_col=df['A'] * 2)     # chainable modification
```

## 5. Aggregation
```python
# Basic groupby
df.groupby('category')['value'].mean()

# Named aggregation (Best Practice)
df.groupby('category').agg(
    count=('id', 'count'),
    total=('value', 'sum')
)

# Window function (Group level feature)
df['dept_avg'] = df.groupby('dept')['salary'].transform('mean')
```

## 7. Strings & Dates
```python
df['str_col'].str.lower().str.replace('old', 'new')
df['str_col'].str.extract(r'(pattern)')

df['date'] = pd.to_datetime(df['date_str'])
df['date'].dt.year
df['date'].dt.day_name()
```

## 8. Vectorized Logic (If/Else)
```python
import numpy as np
df['flag'] = np.where(df['val'] > 0, 1, 0)
df['tier'] = pd.cut(df['val'], bins=[0, 10, 100], labels=['Low', 'High'])
```
