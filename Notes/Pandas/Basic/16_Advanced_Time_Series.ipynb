# Pro-Level: Time Series Analysis

Handling stock prices, server logs, or weather data requires advanced time-series methods.


## 1. Advanced Resampling

We don't just take the mean; we can `.ohlc()` (Open-High-Low-Close) which is standard in finance.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.
- **`.ohlc()`**: Computes the Open, High, Low, and Close values for time-series data.
- **`.resample(rule)`**: Groups time-series data into chunks (e.g. '4H' for 4 hours) so you can aggregate them.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

dates = pd.date_range('2023-01-01 09:00:00', periods=12, freq='H') # Hourly data
prices = [100, 102, 101, 105, 104, 108, 107, 110, 109, 112, 115, 114]
df = pd.DataFrame({'Price': prices}, index=dates)

# Resample to 4-Hour blocks and calculate OHLC
display(df.resample('4H').ohlc())
```

## 2. Expanding Windows (`.expanding()`)

While `.rolling()` looks at a fixed window (e.g. last 3 days), `.expanding()` looks at *all data up to the current point* (e.g. "Total cumulative average").


### 🔍 Functions & Attributes Used in this Cell:
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.expanding()`**: Provides expanding transformations (cumulative calculations from the start of the series to the current point).
- **`.rolling(window)`**: Provides rolling window calculations (e.g. a 7-day moving average).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df['3H_Rolling_Avg'] = df['Price'].rolling(window=3).mean()
df['Expanding_Avg'] = df['Price'].expanding().mean()
display(df.head(6))
```

## 3. Rate of Change (`.diff()`, `.pct_change()`)

Instead of manually using `.shift()`, we have built-in methods for calculating day-over-day changes.


### 🔍 Functions & Attributes Used in this Cell:
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.pct_change()`**: Calculates the percentage change between the current and a prior element.
- **`.diff(periods)`**: Calculates the difference between the current element and a prior element.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df['Absolute_Change'] = df['Price'].diff(periods=1)
df['Percentage_Change'] = df['Price'].pct_change(periods=1) * 100
display(df.head(4))
```

## 4. Timezones

Data is often in UTC, but you need it in Local Time.


### 🔍 Functions & Attributes Used in this Cell:
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.tz_convert(tz)`**: Converts a timezone-aware DatetimeIndex from one timezone to another.
- **`.tz_localize(tz)`**: Localizes a timezone-naive DatetimeIndex to a specific timezone.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Localize assigns a timezone (if none exists). Convert changes it to a new one.
df = df.tz_localize('UTC')
df_est = df.tz_convert('America/New_York')
display(df_est.head(2))
```
