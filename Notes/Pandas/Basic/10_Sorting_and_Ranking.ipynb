# Sorting and Ranking: Deep Dive

Organizing data is essential before grouping or exporting. We'll explore complex sorting parameters, handling NaNs, and different ranking methods.


## 1. Setup Data


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

data = {
    'Student': ['Eve', 'Alice', 'Charlie', 'Bob', 'David', 'Zack'],
    'Score': [95, 85, 90, 85, 92, np.nan],
    'Attendance': [90, 95, 85, 80, 99, 50]
}
df = pd.DataFrame(data)
display(df)
```

## 2. Advanced Sorting (`.sort_values()`)

We can sort by multiple columns and explicitly tell Pandas where to put missing values.


### 🔍 Functions & Attributes Used in this Cell:
- **`.sort_values(by, ascending, na_position)`**: Sorts the dataframe by one or more columns.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Sort by Score (Desc), then Attendance (Desc). Put NaNs FIRST.
display(df.sort_values(by=['Score', 'Attendance'], ascending=[False, False], na_position='first'))
```

**Output Explanation:**
Zack (who has a missing score) appears at the very top because of `na_position='first'`. Bob and Alice both scored 85, but Alice appears first because she had higher attendance.


## 3. Sorting the Index (`.sort_index()`)

After filtering, your index might be out of order. We can sort it, ascending or descending.


### 🔍 Functions & Attributes Used in this Cell:
- **`.sample(n, frac)`**: Returns a random sample of rows. `n` specifies exact count, `frac` specifies a percentage. Use `random_state` for reproducibility.
- **`.sort_index()`**: Sorts the dataframe by its row labels.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df_shuffled = df.sample(frac=1)
print("Shuffled Data:")
display(df_shuffled)

print("\nSorted Index (Descending):")
display(df_shuffled.sort_index(ascending=False))
```

## 4. Advanced Ranking (`.rank()`)

Ranking assigns a position (1st, 2nd). But what happens if there's a tie? We use the `method=` parameter.


### 🔍 Functions & Attributes Used in this Cell:
- **`.sort_values(by, ascending, na_position)`**: Sorts the dataframe by one or more columns.
- **`.rank(method)`**: Computes numerical data ranks (1st, 2nd, 3rd). `method='dense'` handles ties without skipping rank numbers.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# average: (default) Alice and Bob tie for 4th/5th, so they both get 4.5
# min: Alice and Bob both get 4
# dense: Ranks are packed. Alice and Bob get 4, the next person gets 5.

df['Rank_Avg'] = df['Score'].rank(method='average', ascending=False)
df['Rank_Min'] = df['Score'].rank(method='min', ascending=False)
df['Rank_Dense'] = df['Score'].rank(method='dense', ascending=False)

display(df.sort_values('Score', ascending=False))
```

> **Interview Tip:** When preparing leaderboards for games or sales, always use `.rank(method='dense')` so that a tie doesn't skip the next rank number (e.g., 1st, 2nd, 2nd, 3rd... instead of 1st, 2nd, 2nd, 4th...).
