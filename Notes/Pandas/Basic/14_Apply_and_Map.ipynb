# Custom Functions: apply, map, and applymap

When built-in Pandas methods fail, we fall back on `.apply()`.


## 1. Setup Data


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Height_cm': [170, 182],
    'Weight_kg': [65, 80]
})
display(df)
```

## 2. `.apply()` on Columns (Axis=0) vs Rows (Axis=1)

If we apply on a column (Series), it passes one cell at a time. If we apply on `axis=1`, it passes the entire ROW to our function.


### 🔍 Functions & Attributes Used in this Cell:
- **`.apply(func, axis)`**: Applies a custom function along an axis (cell-by-cell for Series, or row-by-row for DataFrames if `axis=1`).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Series Apply (Cell by Cell)
df['Height_m'] = df['Height_cm'].apply(lambda x: x / 100)

# DataFrame Apply (Row by Row)
# We can calculate BMI because the row contains both height and weight!
def calculate_bmi(row):
    height_m = row['Height_cm'] / 100
    return row['Weight_kg'] / (height_m ** 2)

df['BMI'] = df.apply(calculate_bmi, axis=1)
display(df)
```

## 3. `.map()` for Dictionary Translation

Extremely fast for translating values.


### 🔍 Functions & Attributes Used in this Cell:
- **`.map(arg)`**: Maps values of a Series using an input correspondence (a dictionary or a function).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
mapping = {'Alice': 'Female', 'Bob': 'Male'}
df['Gender'] = df['Name'].map(mapping)
display(df)
```

## 4. `.applymap()` (or `.map()` in newer Pandas) for Entire DataFrame

Applies a function to EVERY SINGLE CELL in the entire DataFrame.


### 🔍 Functions & Attributes Used in this Cell:
- **`.applymap(func)`**: Applies a function to every single cell across the entire DataFrame.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df_numeric = df[['Height_cm', 'Weight_kg']]

# Format all numbers as strings with 1 decimal
display(df_numeric.applymap(lambda x: f"{x:.1f}"))
```
