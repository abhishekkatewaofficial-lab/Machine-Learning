# Series Operations: The Deep Dive

Pandas Series allow for fast mathematical operations and built-in statistical methods without using loops (vectorization). We will cover dozens of methods here.


## 1. Setup Mock Dataset


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

# Monthly revenue (with a missing value and an outlier)
revenue = pd.Series([10.5, 12.0, 11.5, np.nan, 15.0, 99.0, 14.2, 11.5, 13.0, 11.5])
display(revenue)
```

## 2. Core Statistical Methods

Pandas automatically ignores `NaN` values in these calculations by default (unlike raw Numpy).


```python
print("Sum:", revenue.sum())
print("Mean (Average):", revenue.mean())
print("Median:", revenue.median())
print("Mode (Most frequent):", revenue.mode().values)
print("Standard Deviation:", revenue.std())
print("Variance:", revenue.var())
print("Min/Max:", revenue.min(), "/", revenue.max())
```

## 3. Cumulative Methods

Great for tracking growth over time (e.g., Year-to-Date totals).


```python
# Note how the NaN is handled (it's skipped, but the sum continues)
display(revenue.cumsum())
```

> **Interview Tip:** Other cumulative methods include `.cummax()`, `.cummin()`, and `.cumprod()`. These are heavily used in finance algorithms.


## 4. Value Counts & Unique (Deep Dive)

`.value_counts()` is arguably the most used method in EDA.


```python
print("--- Standard Counts ---")
display(revenue.value_counts())

print("\n--- Normalized (Percentages) & Include NaNs ---")
display(revenue.value_counts(normalize=True, dropna=False))

print("\n--- Binning Continuous Data ---")
display(revenue.value_counts(bins=3))
```

**Output Explanation:**
- `dropna=False` ensures `NaN` is counted as a category.
- `bins=3` automatically groups continuous numbers into 3 equal-width ranges and counts how many items fall into each! Very powerful.


## 5. Finding Extremes (`.nlargest()`, `.nsmallest()`, `.idxmax()`)

Sorting the entire dataset just to find the top 2 values is inefficient. Use these instead.


```python
print("Top 2 Revenues:")
display(revenue.nlargest(2))

print("\nBottom 3 Revenues:")
display(revenue.nsmallest(3))

print("\nIndex of the absolute maximum value:", revenue.idxmax())
```

## 6. Value Modification (`.clip()`, `.between()`, `.round()`)

We have a massive outlier (99.0). We can clip it.


### 🔍 Functions & Attributes Used in this Cell:
- **`.between(left, right)`**: Returns a boolean mask showing if values fall between `left` and `right` (inclusive).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Clip forces all values to be within a boundary (e.g., min 10, max 20)
clipped = revenue.clip(lower=10, upper=20)
print("Clipped:")
display(clipped)

# Check which values are between 11 and 13 (inclusive)
print("\nBetween 11 and 13:")
display(revenue.between(11, 13))

# Rounding to 0 decimal places
print("\nRounded:")
display(revenue.round(0))
```
