# Column Operations: The Deep Dive

Data manipulation often involves creating new features (Feature Engineering), cleaning up column names, and dropping unused features.


## 1. Setup Data


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

df = pd.DataFrame({
    'Price_USD': [100, 200, 300],
    'Tax_Rate': [0.1, 0.15, 0.2]
}, index=['Item1', 'Item2', 'Item3'])
display(df)
```

## 2. Dynamic Renaming (`.rename()`, `.add_prefix()`, `.add_suffix()`)

You can rename columns using dictionaries, or apply string functions to all column names instantly.


### 🔍 Functions & Attributes Used in this Cell:
- **`.add_prefix(prefix)`**: Adds a string to the start of every column name.
- **`.rename(columns, index)`**: Renames labels using a dictionary or a function (e.g. `str.lower`).
- **`.add_suffix(suffix)`**: Adds a string to the end of every column name.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Standard Dictionary Rename ---")
display(df.rename(columns={'Price_USD': 'price', 'Tax_Rate': 'tax'}))

print("\n--- Lowercase all columns dynamically ---")
display(df.rename(columns=str.lower))

print("\n--- Add Prefix / Suffix (Great for joined tables) ---")
display(df.add_prefix('Q1_').add_suffix('_2023'))
```

## 3. Creating Multiple Columns Cleanly (`.assign()`)

The standard way `df['New'] = ...` works, but `.assign()` allows you to create multiple columns at once and chain methods.


### 🔍 Functions & Attributes Used in this Cell:
- **`.assign(**kwargs)`**: Creates multiple new columns simultaneously. Excellent for method chaining.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Create Final_Price and Discount_Price simultaneously
new_df = df.assign(
    Final_Price = lambda x: x['Price_USD'] * (1 + x['Tax_Rate']),
    Discount_Price = lambda x: x['Price_USD'] * 0.9
)
display(new_df)
```

## 4. Dropping Columns/Rows safely (`.drop()`)

Use `.drop()` with `axis=0` for rows, and `axis=1` (or `columns=`) for columns.


### 🔍 Functions & Attributes Used in this Cell:
- **`.drop(labels, axis)`**: Drops specified rows (`axis=0`) or columns (`axis=1`).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Drop a column
display(new_df.drop(columns=['Tax_Rate']))

# Drop a row by its index label
display(new_df.drop(index=['Item2']))
```
