# Series Indexing: The Deep Dive

How do we extract specific data from a Series? We use `.loc`, `.iloc`, `.at`, `.iat`, and `.get()`. Mastering these is non-negotiable for Data Science.


## 1. Setup Mock Dataset


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

# Let's create a realistic mock series of daily active users (DAU)
dates = pd.date_range('2023-01-01', periods=7)
dau = pd.Series([1050, 1100, 1080, 1200, 1500, 1600, 1150], index=dates, name='DAU')
display(dau)
```

## 2. Label-based selection: `.loc` and `.at`

`.loc` is for labels. `.at` is a faster version of `.loc` that ONLY works for single scalar values.


### 🔍 Functions & Attributes Used in this Cell:
- **`.loc[]`**: Label-based indexer. `.loc['RowName', 'ColName']`. **Inclusive** of the end bound when slicing.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Extract a single day
print("Using .loc:", dau.loc['2023-01-04'])

# Faster scalar access
print("Using .at:", dau.at['2023-01-04'])

# Extract multiple days (returns a Series)
display(dau.loc[['2023-01-01', '2023-01-07']])
```

> **Performance Tip:** If you are iterating over a Series and looking up single values (which you should avoid, but sometimes it's necessary), use `.at[]` instead of `.loc[]`. It is significantly faster because it skips overhead.


## 3. Integer-based selection: `.iloc` and `.iat`

`.iloc` uses the hidden numerical position (0, 1, 2...). `.iat` is the faster scalar version.


### 🔍 Functions & Attributes Used in this Cell:
- **`.iloc[]`**: Integer-position based indexer. `.iloc[0, 1]`. **Exclusive** of the end bound when slicing.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Extract the first element
print("First item (.iloc):", dau.iloc[0])
print("First item (.iat):", dau.iat[0])

# Extract the last element using negative indexing
print("Last item:", dau.iloc[-1])
```

## 4. Slicing with `.loc` vs `.iloc`

This is a major source of bugs in Pandas. `.loc` slicing is INCLUSIVE of the end bound. `.iloc` slicing is EXCLUSIVE (like normal Python).


### 🔍 Functions & Attributes Used in this Cell:
- **`.loc[]`**: Label-based indexer. `.loc['RowName', 'ColName']`. **Inclusive** of the end bound when slicing.
- **`.iloc[]`**: Integer-position based indexer. `.iloc[0, 1]`. **Exclusive** of the end bound when slicing.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Using iloc (exclusive of index 3) ---")
display(dau.iloc[0:3])  # Gets index 0, 1, 2

print("\n--- Using loc (INCLUSIVE of Jan 3) ---")
display(dau.loc['2023-01-01':'2023-01-03'])
```

> **Common Mistake:** If your index is numeric (e.g., `[1, 2, 3]`), `.loc[0:2]` will look for the literal labels 0, 1, and 2 (inclusive). `.iloc[0:2]` will grab the 1st and 2nd items in memory. Mixing them up causes subtle bugs.


## 5. Safe Access with `.get()`

If you ask `.loc` for a label that doesn't exist, it crashes (KeyError). `.get()` returns a default value instead.


```python
print(dau.get('2025-01-01', default='Date not found!'))
```

## 6. Slicing by Step

You can skip elements by adding a step size: `[start:stop:step]`.


### 🔍 Functions & Attributes Used in this Cell:
- **`.iloc[]`**: Integer-position based indexer. `.iloc[0, 1]`. **Exclusive** of the end bound when slicing.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Get every 2nd day (alternate days)
display(dau.iloc[::2])
```
