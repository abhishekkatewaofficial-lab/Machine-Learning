# Series Mastery: Fundamentals & Deep Dive

A Pandas `Series` is a one-dimensional array. Think of it as a single column in an Excel sheet or a super-powered Python list/dictionary. In this notebook, we will explore EVERY essential attribute and method for Series creation and inspection.


## 1. Creating a Series from a List

You can easily convert a standard Python list into a Pandas Series.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

# Basic Example
students = ['Alice', 'Bob', 'Charlie', 'David']
series_students = pd.Series(students)

display(series_students)
```

**Output Explanation:**
- The left column `(0, 1, 2, 3)` is the **Index** (automatically generated).
- The right column is the **Data**.
- `dtype: object` means the data contains strings/text.


## 2. Custom Indexing & Naming (`name`, `index.name`)

Instead of `0, 1, 2`, you can define your own labels for the index. We can also give the Series itself a name, which becomes the column name if we convert it to a DataFrame later.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Intermediate Example
marks = [85.5, 92.0, 78.5, 90.0]
subjects = ['Math', 'Physics', 'Chemistry', 'Biology']

# Creating the series with index, dtype, and a name
series_marks = pd.Series(data=marks, index=subjects, dtype='float32', name='Final_Scores')
series_marks.index.name = 'Subject'

display(series_marks)
```

**Output Explanation:**
Notice the bottom now says `Name: Final_Scores, dtype: float32`. The index column now has a header called `Subject`. Specifying `dtype='float32'` saves memory compared to the default `float64`.


## 3. Creating from Dictionaries and Scalars

Dictionaries naturally map to Series (Keys -> Index, Values -> Data). You can also broadcast a single scalar value.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
employee_salaries = {'John': 50000, 'Emma': 65000, 'Sam': 55000}
series_salaries = pd.Series(employee_salaries)
display(series_salaries)

# Broadcasting a scalar
base_score = pd.Series(100, index=['Q1', 'Q2', 'Q3', 'Q4'])
display(base_score)
```

> **Interview Tip:** When converting a dictionary to a Series in modern Pandas, the order of the index is guaranteed to match the insertion order of the dictionary.


## 4. Comprehensive Attributes

Let's deeply inspect the Series using attributes (no parentheses).


```python
print("Shape (Dimensions):", series_marks.shape)
print("Size (Total elements):", series_marks.size)
print("Data Type:", series_marks.dtype)
print("Index Labels:", series_marks.index)
print("Raw Numpy Array:", series_marks.values)
print("Is it empty?:", series_marks.empty)
print("Are all values unique?:", series_marks.is_unique)
print("Does it have NaN values?:", series_marks.hasnans)
```

## 5. Type Conversion (`.astype()`)

Converting data types is a massive part of data cleaning.


### 🔍 Functions & Attributes Used in this Cell:
- **`.astype(dtype)`**: Casts a pandas object to a specified data type (e.g., converting float64 to int32 for memory savings).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Let's convert float32 to int32 (truncates the decimal)
series_marks_int = series_marks.astype('int32')
display(series_marks_int)

# Convert to string
series_str = series_marks.astype(str)
display(series_str)
```

## 6. Converting Series to DataFrame (`.to_frame()`)

Sometimes a Series isn't enough, and you want to bump it up to a 2D DataFrame.


```python
df_marks = series_marks.to_frame()
display(df_marks)
print("New type:", type(df_marks))
```
