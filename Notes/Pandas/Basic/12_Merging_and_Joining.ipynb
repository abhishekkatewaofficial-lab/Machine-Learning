# Joining Data: Concatenation and Merging

Combining datasets accurately is critical. We will explore `.concat()`, `.merge()` (Inner/Left/Right/Outer), `.join()`, and the advanced `.merge_asof()`.


## 1. Setup Data


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

customers = pd.DataFrame({'Cust_ID': [1, 2, 3, 4], 'Name': ['Alice', 'Bob', 'Charlie', 'David']})
orders = pd.DataFrame({'Order_ID': [101, 102, 103], 'Cust_ID': [1, 2, 2], 'Amount': [250, 150, 300]})
print("Customers:"); display(customers)
print("Orders:"); display(orders)
```

## 2. Advanced Concatenation (`pd.concat()`)

We can concat vertically (`axis=0`) or horizontally (`axis=1`). We can also add `keys` to track where data came from.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.
- **`pd.concat(objs, axis)`**: Glues DataFrames together vertically (`axis=0`) or horizontally (`axis=1`).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
q1_sales = pd.DataFrame({'Sales': [100, 200]}, index=['Jan', 'Feb'])
q2_sales = pd.DataFrame({'Sales': [150, 300]}, index=['Apr', 'May'])

# Vertical concat with keys
all_sales = pd.concat([q1_sales, q2_sales], keys=['Q1', 'Q2'])
display(all_sales)
```

## 3. SQL-Style Merging (`pd.merge()`)

We can do `inner`, `left`, `right`, or `outer` joins. The `indicator=True` parameter is a lifesaver for debugging.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.merge(left, right, how, on)`**: SQL-style joining of DataFrames (inner, left, right, outer).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- LEFT JOIN with Indicator ---")
# Keeps all customers, even if they have no orders.
left_join = pd.merge(customers, orders, on='Cust_ID', how='left', indicator=True)
display(left_join)
```

**Output Explanation:**
The `_merge` column tells us exactly what happened. Charlie and David are `left_only`, meaning they exist in the customers table but have 0 orders.


## 4. Index Joining (`.join()`)

If you want to merge two tables based on their indexes (not columns), `.join()` is a faster shorthand.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.
- **`.join(other)`**: Quickly merges DataFrames on their indexes instead of their columns.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df1 = pd.DataFrame({'A': [1, 2]}, index=['X', 'Y'])
df2 = pd.DataFrame({'B': [3, 4]}, index=['X', 'Z'])

display(df1.join(df2, how='outer'))
```

## 5. Pro-Level: `pd.merge_asof()`

Imagine joining stock prices (recorded at 10:00:01) with a trade execution (recorded at 10:00:05). An exact `merge` fails. `merge_asof()` joins on the *closest prior time*.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.
- **`pd.merge_asof()`**: Matches rows to the closest (but not exactly equal) prior key. Incredible for financial time-series matching.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
quotes = pd.DataFrame({'Time': [10, 20, 30], 'Price': [100.5, 101.0, 100.8]})
trades = pd.DataFrame({'Time': [12, 28], 'Volume': [500, 200]})

# For each trade, find the most recent quote price BEFORE the trade happened
display(pd.merge_asof(trades, quotes, on='Time'))
```
