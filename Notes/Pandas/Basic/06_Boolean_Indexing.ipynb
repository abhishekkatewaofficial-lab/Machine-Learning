# Boolean Indexing: Advanced Filtering

Filtering data isn't just about True/False arrays. We can use `.isin()`, `.between()`, and the ultra-fast `.query()` method.


## 1. Setup Mock Dataset


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Age': [25, 45, 30, 50, 28, 60],
    'Department': ['HR', 'IT', 'IT', 'Sales', 'HR', 'IT'],
    'Salary': [50000, 120000, 80000, 95000, 55000, 150000]
}
df = pd.DataFrame(data)
display(df)
```

## 2. Standard Filtering (`&`, `|`, `~`)

Remember: ALWAYS use parentheses around conditions.


```python
# IT Department AND Age over 40
display(df[(df['Department'] == 'IT') & (df['Age'] > 40)])

# Using the ~ (NOT) operator to get everyone EXCEPT IT
display(df[~(df['Department'] == 'IT')])
```

## 3. List Matching (`.isin()`) and Range Matching (`.between()`)

Don't use `(Dept == 'HR') | (Dept == 'Sales')`. Use `.isin()`.


### 🔍 Functions & Attributes Used in this Cell:
- **`.isin(values)`**: Returns a boolean mask showing if each element matches any value in the provided list. Replaces messy `(A == 1) | (A == 2)` chains.
- **`.between(left, right)`**: Returns a boolean mask showing if values fall between `left` and `right` (inclusive).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# isin()
display(df[df['Department'].isin(['HR', 'Sales'])])

# between() (inclusive bounds)
display(df[df['Age'].between(30, 50)])
```

## 4. The `.query()` Method (Pro-Level)

`.query()` allows you to write filtering conditions as a clean Python string. It is often faster than standard bracket filtering for large datasets because it uses `numexpr` under the hood.


### 🔍 Functions & Attributes Used in this Cell:
- **`.query(expr)`**: Filters rows using a clean string expression like `'Age > 40 and Dept == "IT"'`. Extremely fast for large datasets.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
target_dept = 'IT'

# You can reference external variables using the @ symbol!
result = df.query("Age > 40 and Department == @target_dept and Salary >= 100000")
display(result)
```

**Output Explanation:**
Notice we used the standard Python `and` keyword instead of `&`, and we didn't need parentheses or `df['...']`. It is incredibly readable.


## 5. Replacing Data Based on Conditions (`.where()` and `.mask()`)

What if we don't want to drop the rows, but rather replace the values that don't meet a condition?


### 🔍 Functions & Attributes Used in this Cell:
- **`.mask(cond, other)`**: The exact opposite of `.where()`. Replaces values where the condition is True with `other`.
- **`.where(cond, other)`**: Keeps values where the condition is True. Replaces values where the condition is False with `other`.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# .where() keeps values that are True, and replaces values that are False
# Let's hide salaries of people under 40
df['Hidden_Salary'] = df['Salary'].where(df['Age'] >= 40, other='CONFIDENTIAL')

# .mask() does the exact opposite. It replaces values that ARE True.
df['Flagged'] = df['Department'].mask(df['Salary'] > 100000, other='HIGH EARNER')

display(df)
```
