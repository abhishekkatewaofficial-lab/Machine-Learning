# Handling Missing Data: The Deep Dive

Real-world data is messy. In Kaggle and industry, handling `NaN` (Not a Number) correctly is the difference between a great model and a broken one. We will explore `.dropna()`, `.fillna()`, and `.interpolate()` with advanced parameters.


## 1. Setup Mock Dataset


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

data = {
    'Patient_ID': [1, 2, 3, 4, 5, 6, 7],
    'Age': [25, np.nan, 35, 40, np.nan, np.nan, 50],
    'Heart_Rate': [70, 72, np.nan, 75, 76, np.nan, 80],
    'Test_Score': [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
}
df = pd.DataFrame(data)
display(df)
```

## 2. Advanced Detection (`.isna()`, `.notna()`)

We need to find exactly how much data is missing.


### 🔍 Functions & Attributes Used in this Cell:
- **`.isna()`**: Returns a boolean identical-sized object showing True for missing values (NaN/None).
- **`.notna()`**: Returns True for non-missing values.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("Missing Values per Column:")
display(df.isna().sum())

print("\nPercentage of Missing Values per Column:")
display((df.isna().sum() / len(df)) * 100)

print("\nFilter rows where Age is NOT missing:")
display(df[df['Age'].notna()])
```

## 3. Advanced Dropping (`.dropna(how, thresh, subset)`)

`.dropna()` has powerful arguments so you don't accidentally delete useful data.


### 🔍 Functions & Attributes Used in this Cell:
- **`.dropna(how, thresh, subset)`**: Drops rows or columns containing missing values. `thresh=3` requires at least 3 non-null values to keep the row.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Drop columns that are completely empty (how='all') ---")
display(df.dropna(axis=1, how='all')) # Test_Score is deleted

print("\n--- Require at least 3 non-NaN values to keep a row (thresh=3) ---")
display(df.dropna(thresh=3))

print("\n--- Only drop if 'Heart_Rate' is missing (subset=['Heart_Rate']) ---")
display(df.dropna(subset=['Heart_Rate']))
```

## 4. Advanced Filling (`.fillna()` and `.replace()`)

Filling with a single value is basic. We can forward-fill, backward-fill, or cap the limit.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.
- **`.replace(to_replace, value)`**: Replaces specific exact values (like a placeholder `999`) with another value.
- **`.ffill()`**: Forward fill. Propagates the last valid observation forward to fill gaps.
- **`.bfill()`**: Backward fill. Uses the next valid observation to fill gaps backwards.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Forward Fill (propagate last valid observation forward) ---")
# Often used in time series
display(df['Age'].ffill())

print("\n--- Backward Fill (use next valid observation) with limit=1 ---")
display(df['Age'].bfill(limit=1))

print("\n--- .replace() for weird placeholder values ---")
# Imagine 999 was used as a placeholder for missing
messy = pd.Series([10, 20, 999, 30])
display(messy.replace(999, np.nan))
```

## 5. Interpolation (`.interpolate()`)

For time series or continuous data, filling with a straight mean is bad. Interpolation estimates the missing value based on surrounding data points.


### 🔍 Functions & Attributes Used in this Cell:
- **`.interpolate(method)`**: Guesses missing values by interpolating between existing data points. Essential for time-series.
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
time_series = pd.Series([10.0, np.nan, np.nan, 40.0])
print("Original:")
display(time_series)

print("\nLinear Interpolation:")
# It perfectly guesses 20.0 and 30.0!
display(time_series.interpolate(method='linear'))
```

> **Interview Tip:** When predicting missing values using `.mean()` or `.median()`, ensure you only calculate the mean on the *training data*, not the entire dataset, to avoid "Data Leakage".
