# DataFrame Mastery: Fundamentals & Deep Dive

A `DataFrame` is a two-dimensional table with rows and columns. It's essentially a dictionary of `Series` objects sharing the same index. We will cover how to create, inspect, and analyze the memory profile of DataFrames.


## 1. Creating a DataFrame

You can create a DataFrame from a dictionary of lists, a list of dictionaries, or a 2D Numpy array.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

# Creating from a Dictionary of Lists (Most Common)
data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['HR', 'Engineering', 'Marketing', 'Engineering', 'Sales'],
    'Salary': [60000, 120000, 85000, 110000, 95000]
}
df = pd.DataFrame(data)
display(df)
```

## 2. Inspecting the Data (`.head()`, `.tail()`, `.sample()`)

Never print the whole DataFrame. Use these methods to peek at the data.


### 🔍 Functions & Attributes Used in this Cell:
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.tail(n)`**: Returns the last `n` rows of the DataFrame.
- **`.sample(n, frac)`**: Returns a random sample of rows. `n` specifies exact count, `frac` specifies a percentage. Use `random_state` for reproducibility.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("First 2 rows:")
display(df.head(2))

print("\nLast 2 rows:")
display(df.tail(2))

print("\nRandomly sample 2 rows (great for avoiding bias):")
display(df.sample(n=2, random_state=42))
```

> **Interview Tip:** When exploring a new dataset, always use `.sample()` instead of just `.head()`. `.head()` might only show you data from one specific day or category if the data is sorted. `.sample()` shows you a true random mix.


## 3. Metadata and Statistics (`.info()`, `.describe()`)

`.info()` tells you about missing values and data types. `.describe()` gives you mathematical summaries.


### 🔍 Functions & Attributes Used in this Cell:
- **`.describe()`**: Calculates statistical summaries (mean, std, min, max, quartiles). Pass `include='all'` to analyze string columns too.
- **`.info()`**: Prints a concise summary of the DataFrame including null counts and data types. Pass `memory_usage='deep'` for exact RAM usage.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Data Information ---")
# memory_usage='deep' gives the EXACT memory usage including strings
df.info(memory_usage='deep')

print("\n--- Statistical Summary ---")
# include='all' forces Pandas to describe string columns too!
display(df.describe(include='all'))
```

## 4. Structural Attributes

We can inspect the exact dimensions and labels.


```python
print("Shape (rows, columns):", df.shape)
print("Size (total cells):", df.size)
print("Columns:", df.columns)
print("Index:", df.index)
print("Axes:", df.axes) # Returns both index and columns
print("Empty?:", df.empty)
```

## 5. Setting and Resetting the Index

The default `0, 1, 2` index is often useless. We can set a specific column as the index.


### 🔍 Functions & Attributes Used in this Cell:
- **`.reset_index()`**: Resets the index to default integers (0, 1, 2...) and moves the old index into a standard column.
- **`.set_index(keys)`**: Sets the DataFrame index (row labels) using one or more existing columns.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Let's make Employee_ID the index
df_indexed = df.set_index('Employee_ID')
display(df_indexed)

# We can reset it back to normal
df_reset = df_indexed.reset_index()
display(df_reset)
```

> **Common Mistake:** `.set_index()` does not modify the original DataFrame unless you use `inplace=True`. Always assign it to a new variable (e.g. `df = df.set_index(...)`).
