# String and Datetime Operations: The Deep Dive

Strings and Dates are notoriously messy. The `.str` and `.dt` accessors give you immense power to parse, clean, and extract features.


## 1. Advanced String Operations (`.str.contains()`, `.str.extract()`)

Let's parse messy log data.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.
- **`.str.contains(pat)`**: (String Accessor) Returns boolean mask if the pattern is found within the string.
- **`.str.extract(pat)`**: (String Accessor) Extracts capture groups in a regex pattern into new columns.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

logs = pd.Series([
    "Error 404: User not found at 10:00AM",
    "Warning: High memory usage",
    "Error 500: Server crashed at 11:30PM"
])

# 1. Filter using regex-powered str.contains()
print("--- Only Error logs ---")
display(logs[logs.str.contains('Error', regex=False)])

# 2. Extract specific patterns using RegEx capture groups
print("\n--- Extract Error Codes ---")
# Looks for the word Error, a space, and captures digits (\d+)
codes = logs.str.extract(r'Error (\d+)')
display(codes)
```

## 2. Advanced Datetime Parsing (`pd.to_datetime()`)

Sometimes dates are in weird formats. You must specify `format=` to speed up parsing and prevent errors.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.to_datetime(arg, format)`**: Converts strings to native Pandas Timestamp objects. Use `format` to drastically speed up parsing.
- **`pd.Series()`**: Creates a 1-dimensional array. Takes `data`, `index`, and `name` as arguments.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
dates = pd.Series(['31-12-2023', '01-05-2023', 'InvalidDate'])

# errors='coerce' turns 'InvalidDate' into NaT (Not a Time) instead of crashing
df_dates = pd.to_datetime(dates, format='%d-%m-%Y', errors='coerce').to_frame('Date')
display(df_dates)
```

## 3. Datetime Feature Extraction (`.dt` accessor)

Machine learning models can't read "2023-12-31". We must extract the year, month, and day-of-week.


```python
df_dates['Year'] = df_dates['Date'].dt.year
df_dates['Month'] = df_dates['Date'].dt.month
df_dates['Day_Name'] = df_dates['Date'].dt.day_name()
df_dates['Is_Leap_Year'] = df_dates['Date'].dt.is_leap_year

display(df_dates)
```

## 4. Date Math (`pd.DateOffset()`, `pd.Timedelta()`)

You can add or subtract time directly!


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DateOffset()`**: Standard kind of date increment used for a date range (e.g., adding exactly 1 month).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Add exactly 3 months and 1 day to the dates
df_dates['Future_Date'] = df_dates['Date'] + pd.DateOffset(months=3, days=1)

# Calculate difference between two dates
df_dates['Difference'] = df_dates['Future_Date'] - df_dates['Date']
display(df_dates[['Date', 'Future_Date', 'Difference']])
```
