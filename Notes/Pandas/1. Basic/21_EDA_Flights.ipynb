Analyzing seasonality and trends using the Seaborn flights dataset.


## 1. Load and Reshape


### 🔍 Functions & Attributes Used in this Cell:
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.

### 🧠 Intuition & Deep Dive:

```python
import pandas as pd
import seaborn as sns

flights = sns.load_dataset('flights')
display(flights.head())
```

## 2. Creating a Datetime Index

We need to combine Year and Month into a proper Date.


### 🔍 Functions & Attributes Used in this Cell:
- **`.drop(labels, axis)`**: Drops specified rows (`axis=0`) or columns (`axis=1`).
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.astype(dtype)`**: Casts a pandas object to a specified data type (e.g., converting float64 to int32 for memory savings).
- **`.set_index(keys)`**: Sets the DataFrame index (row labels) using one or more existing columns.
- **`pd.to_datetime(arg, format)`**: Converts strings to native Pandas Timestamp objects. Use `format` to drastically speed up parsing.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
flights['Date'] = pd.to_datetime(flights['year'].astype(str) + '-' + flights['month'].astype(str))
flights.set_index('Date', inplace=True)
flights.drop(columns=['year', 'month'], inplace=True)
display(flights.head())
```

## 3. Seasonality Analysis

Let's calculate the 6-month rolling average to smooth out the noise and see the true trend.


### 🔍 Functions & Attributes Used in this Cell:
- **`.pct_change()`**: Calculates the percentage change between the current and a prior element.
- **`.dropna(how, thresh, subset)`**: Drops rows or columns containing missing values. `thresh=3` requires at least 3 non-null values to keep the row.
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.rolling(window)`**: Provides rolling window calculations (e.g. a 7-day moving average).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
flights['6M_Rolling_Avg'] = flights['passengers'].rolling(window=6).mean()
flights['YoY_Growth_%'] = flights['passengers'].pct_change(periods=12) * 100

display(flights.dropna().head(10))
```
