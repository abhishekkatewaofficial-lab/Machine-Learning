# Pro-Level: Advanced MultiIndex

A MultiIndex (hierarchical index) is essential when grouping data by multiple dimensions.


## 1. Creating a MultiIndex from scratch


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

# Creating from a Cartesian Product (Cross Join)
index = pd.MultiIndex.from_product(
    [['2022', '2023'], ['Q1', 'Q2']], 
    names=['Year', 'Quarter']
)

# Assign dummy data
df = pd.DataFrame({'Sales': [100, 150, 120, 180], 'Profit': [20, 30, 25, 40]}, index=index)
display(df)
```

## 2. Advanced Selection (`.xs()`)

Selecting data from an inner level of a MultiIndex using `.loc` requires slice objects, which is ugly. `.xs()` is the clean way.


### 🔍 Functions & Attributes Used in this Cell:
- **`.xs(key, level)`**: Returns a cross-section (row(s) or column(s)) from a Series/DataFrame using a MultiIndex.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Select all Q2s regardless of the Year ---")
display(df.xs('Q2', level='Quarter'))

print("\n--- Select Q2 but keep the Quarter level in the output ---")
display(df.xs('Q2', level='Quarter', drop_level=False))
```

## 3. Manipulating Levels (`.swaplevel()`, `.sortlevel()`)

Sometimes you want Quarter to be the primary outer level instead of Year.


### 🔍 Functions & Attributes Used in this Cell:
- **`.swaplevel()`**: Swaps the hierarchical levels of a MultiIndex.
- **`.sort_index()`**: Sorts the dataframe by its row labels.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df_swapped = df.swaplevel('Year', 'Quarter')
df_swapped = df_swapped.sort_index(level='Quarter') # Always sort index after swapping!
display(df_swapped)
```
