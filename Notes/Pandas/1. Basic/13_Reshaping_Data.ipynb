# Reshaping Data: Pivot, Melt, Stack, Unstack

Moving data between "wide" and "long" formats.


## 1. Pivot Table (`.pivot_table()`) and CrossTab (`pd.crosstab()`)

`.pivot_table()` handles duplicates by aggregating. `pd.crosstab()` is a specialized function just for counting frequencies between two columns.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.crosstab(index, columns)`**: Computes a simple cross-tabulation of two (or more) factors, showing frequency counts.
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.
- **`.pivot_table(index, columns, values, aggfunc)`**: Creates a spreadsheet-style pivot table.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

df = pd.DataFrame({
    'Date': ['Mon', 'Mon', 'Tue', 'Tue', 'Mon'],
    'City': ['NY', 'LA', 'NY', 'LA', 'NY'],
    'Sales': [100, 150, 200, 50, 300]
})

print("--- Pivot Table (Mean Sales) with Margins (Totals) ---")
display(df.pivot_table(index='Date', columns='City', values='Sales', aggfunc='mean', margins=True, fill_value=0))

print("\n--- CrossTab (Count of transactions) ---")
display(pd.crosstab(df['Date'], df['City']))
```

## 2. Melting (`.melt()`)

Turning wide columns back into rows.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.
- **`.melt(id_vars, value_vars)`**: Unpivots a DataFrame from wide format to long format.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
wide_df = pd.DataFrame({'Student': ['Alice', 'Bob'], 'Math': [90, 80], 'Science': [95, 85]})
display(wide_df)

print("\nMelted:")
display(wide_df.melt(id_vars=['Student'], var_name='Subject', value_name='Score'))
```

## 3. Stacking and Unstacking

These are like Pivot/Melt, but they work exclusively on MultiIndexes.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.
- **`.unstack()`**: Pivots the innermost row index level to the innermost column level.
- **`.stack()`**: Pivots the innermost column level to the innermost row index level.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
index = pd.MultiIndex.from_tuples([('A', 'Cat1'), ('A', 'Cat2'), ('B', 'Cat1')])
df_multi = pd.DataFrame({'Val': [10, 20, 30]}, index=index)
print("Original MultiIndex:")
display(df_multi)

print("\nUnstacked (Moves Level 1 to Columns):")
unstacked = df_multi.unstack()
display(unstacked)

print("\nStacked (Moves Columns back to Index):")
display(unstacked.stack())
```
