# EDA - Sales Time Series

Let's do one final real-world example: A messy sales dataset that requires time-series aggregation and grouping.


## 1. Generate Mock Sales Data


### 🔍 Functions & Attributes Used in this Cell:
- **`.loc[]`**: Label-based indexer. `.loc['RowName', 'ColName']`. **Inclusive** of the end bound when slicing.
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
np.random.seed(42)

sales_data = pd.DataFrame({
    'Date': np.random.choice(dates, 500),
    'Store': np.random.choice(['Store_A', 'Store_B', 'Store_C'], 500),
    'Revenue': np.random.randint(100, 1000, 500)
})
# Introduce some dirty data
sales_data.loc[10:20, 'Revenue'] = np.nan 
display(sales_data.head())
```

## 2. Clean the Data


### 🔍 Functions & Attributes Used in this Cell:
- **`.assign(**kwargs)`**: Creates multiple new columns simultaneously. Excellent for method chaining.
- **`.transform(func)`**: Applies a function to groups but broadcasts the result back to the original size of the DataFrame (crucial for Feature Engineering).
- **`.fillna(value, method)`**: Replaces missing values with a specified scalar value, or uses a method like `'ffill'` (forward fill).
- **`.isna()`**: Returns a boolean identical-sized object showing True for missing values (NaN/None).
- **`.groupby(by)`**: Splits data into groups based on some criteria, allowing you to apply functions to each group independently.
- **`.sort_values(by, ascending, na_position)`**: Sorts the dataframe by one or more columns.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# 1. Sort by Date
# 2. Fill missing revenue with the median revenue of that specific store

sales_clean = (
    sales_data
    .sort_values('Date')
    .assign(Revenue = lambda x: x.groupby('Store')['Revenue'].transform(lambda y: y.fillna(y.median())))
)

print("Missing values after clean:", sales_clean['Revenue'].isna().sum())
```

**Output Explanation:**
We used Method Chaining! We sorted by Date, then used `.assign()` and `.transform()` to fill `NaN`s with the median revenue of *that specific store*, not the global median. This is advanced!


## 3. Monthly Sales per Store


### 🔍 Functions & Attributes Used in this Cell:
- **`.set_index(keys)`**: Sets the DataFrame index (row labels) using one or more existing columns.
- **`.groupby(by)`**: Splits data into groups based on some criteria, allowing you to apply functions to each group independently.
- **`.unstack()`**: Pivots the innermost row index level to the innermost column level.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Set Date as index for resampling
sales_clean.set_index('Date', inplace=True)

# Pivot table with resampling (Grouper)
monthly_sales = sales_clean.groupby([pd.Grouper(freq='M'), 'Store'])['Revenue'].sum().unstack()

display(monthly_sales)
```

**Output Explanation:**
`pd.Grouper(freq='M')` allows us to group by Month directly from a datetime column/index. `.unstack()` pushes the 'Store' level up to become columns (similar to Pivot).
