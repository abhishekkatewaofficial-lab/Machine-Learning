## Part 1: The Matplotlib Foundation (Object-Oriented Interface)

### Intuition: The Figure and the Axes
Most beginners use `plt.plot()`. This is the "State-based" approach, and it breaks down as soon as you want to plot two things side-by-side.

Professionals use the **Object-Oriented (OO) Interface**: `fig, ax = plt.subplots()`.
- **Figure (`fig`)**: The blank canvas. The window itself.
- **Axes (`ax`)**: The actual plot (the box with the X and Y ticks). A Figure can hold multiple Axes.


### 🔍 Functions & Attributes Used in this Cell:
- **`plt.tight_layout()`**: Automatically adjusts subplot parameters so that labels and titles don't overlap.
- **`plt.subplots(nrows, ncols)`**: The Object-Oriented foundation of Matplotlib. Creates a Figure canvas and a grid of Axes (plots).
- **`.plot.scatter(x, y, c, alpha)`**: Generates a scatter plot. `alpha` controls transparency to reveal data density.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a global aesthetic style
plt.style.use('fivethirtyeight')

# 1. Create a Figure and a 1x2 Grid of Axes (2 plots side-by-side)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Generate some dummy data
x = np.linspace(0, 10, 100)

# 2. Plot on the first axes
ax1.plot(x, np.sin(x), color='blue', linestyle='--', label='Sine Wave')
ax1.set_title("Plot on AX1")
ax1.set_xlabel("Time")
ax1.set_ylabel("Amplitude")
ax1.legend()

# 3. Plot on the second axes
ax2.scatter(x, np.cos(x), color='red', marker='x', label='Cosine Scatter')
ax2.set_title("Plot on AX2")
ax2.legend()

# 4. Display the figure
plt.tight_layout() # Prevents overlapping text
plt.show()
```

> **Interview Tip:** When passing Pandas data into Matplotlib, you can pass the `.plot(ax=ax1)` parameter to tell Pandas *exactly* which box on your canvas to draw the chart in!

## Part 2: Real-World Dataset (The Diamonds Dataset)

Let's load the famous `diamonds` dataset built into Seaborn. It has 53,940 rows. Plotting 50,000 rows introduces unique challenges like overplotting.


### 🔍 Functions & Attributes Used in this Cell:
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.info()`**: Prints a concise summary of the DataFrame including null counts and data types. Pass `memory_usage='deep'` for exact RAM usage.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
diamonds = sns.load_dataset('diamonds')
display(diamonds.info())
display(diamonds.head())
```

### 1. Histograms (Distribution of a Single Variable)

**Intuition:** A histogram answers: "What is the most common price for a diamond? Are there outliers?"

We divide the prices into `bins` and count how many diamonds fall into each bin.


### 🔍 Functions & Attributes Used in this Cell:
- **`.plot.hist(bins)`**: Generates a histogram to visualize the distribution of continuous data.
- **`plt.subplots(nrows, ncols)`**: The Object-Oriented foundation of Matplotlib. Creates a Figure canvas and a grid of Axes (plots).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Pandas plot wrapper explicitly passing the 'ax'
diamonds['price'].plot.hist(bins=100, color='purple', edgecolor='black', ax=ax)

ax.set_title("Distribution of Diamond Prices (Highly Right-Skewed)")
ax.set_xlabel("Price in USD")
plt.show()
```

### 2. Bar Charts (Comparing Categories)

**Intuition:** Which diamond cut is the most common? 
*Crucial Rule:* Never use `.plot.bar()` on the raw dataset. You MUST aggregate the data first (e.g., using `.value_counts()` or `.groupby()`).


### 🔍 Functions & Attributes Used in this Cell:
- **`.plot.barh()`**: Generates a horizontal bar plot (excellent for long category names).
- **`plt.subplots(nrows, ncols)`**: The Object-Oriented foundation of Matplotlib. Creates a Figure canvas and a grid of Axes (plots).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
cut_counts = diamonds['cut'].value_counts()

fig, ax = plt.subplots(figsize=(8, 4))
# We use a horizontal bar chart so the category names are easy to read
cut_counts.plot.barh(color='teal', ax=ax)

ax.set_title("Count of Diamonds by Cut Quality")
ax.invert_yaxis() # Puts the largest bar at the top
plt.show()
```

### 3. Stacked Bar Charts (Multi-dimensional Categories)

**Intuition:** We want to know the count of diamonds by `cut`, BUT we also want to see the breakdown of `color` within each cut.

We must use a Pivot Table or CrossTab before plotting.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.crosstab(index, columns)`**: Computes a simple cross-tabulation of two (or more) factors, showing frequency counts.
- **`.plot.bar(stacked)`**: Generates a vertical bar plot. Use `stacked=True` for composite categories.
- **`plt.subplots(nrows, ncols)`**: The Object-Oriented foundation of Matplotlib. Creates a Figure canvas and a grid of Axes (plots).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
cut_color_cross = pd.crosstab(diamonds['cut'], diamonds['color'])

fig, ax = plt.subplots(figsize=(10, 6))
# stacked=True places the colors on top of each other
cut_color_cross.plot.bar(stacked=True, colormap='viridis', ax=ax)

ax.set_title("Diamond Colors within Each Cut Quality")
ax.set_ylabel("Total Count")
# Move legend outside the plot so it doesn't block data
ax.legend(title='Color', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### 4. Scatter Plots & The Overplotting Problem

**Intuition:** Does a larger Carat mean a higher Price?
We use a scatter plot. However, plotting 54,000 dots on top of each other creates a massive black blob.


### 🔍 Functions & Attributes Used in this Cell:
- **`plt.subplots(nrows, ncols)`**: The Object-Oriented foundation of Matplotlib. Creates a Figure canvas and a grid of Axes (plots).
- **`.plot.scatter(x, y, c, alpha)`**: Generates a scatter plot. `alpha` controls transparency to reveal data density.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bad Scatter Plot (Solid Blob)
diamonds.plot.scatter(x='carat', y='price', ax=ax1, color='blue')
ax1.set_title("Bad: 100% Opaque (Overplotting)")

# Good Scatter Plot (Alpha Transparency)
diamonds.plot.scatter(x='carat', y='price', alpha=0.05, ax=ax2, color='darkblue')
ax2.set_title("Good: alpha=0.05 (Shows Density)")

plt.show()
```

### 5. Hexbin Plots (The Ultimate Fix for Big Data)

**Intuition:** When `alpha` transparency isn't enough, we use a Hexbin. It divides the screen into hexagons. The color of the hexagon represents how many data points fell into that specific region. It's essentially a 2D Histogram.


### 🔍 Functions & Attributes Used in this Cell:
- **`plt.subplots(nrows, ncols)`**: The Object-Oriented foundation of Matplotlib. Creates a Figure canvas and a grid of Axes (plots).
- **`.plot.hexbin(gridsize, cmap)`**: Generates a hexagonal bin plot. The ultimate solution for visualizing massive overlapping data points.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
fig, ax = plt.subplots(figsize=(10, 7))

# gridsize determines how many hexagons across the x-axis
# cmap='inferno' makes dense areas glow bright yellow/white
hb = diamonds.plot.hexbin(x='carat', y='price', gridsize=50, cmap='inferno', ax=ax)

ax.set_title("Hexbin: Density of Carat vs Price")
plt.show()
```

**Output Explanation:**
Look how powerful this is! We can instantly see that the vast majority of diamonds are clustered around 0.3 carats and under $2000. This was completely hidden in the standard scatter plot!

### 6. Box Plots (Identifying Outliers)

**Intuition:** The box shows the middle 50% of the data (the Interquartile Range). The line inside is the median. The "whiskers" show the expected range, and the dots outside the whiskers are mathematical outliers.


### 🔍 Functions & Attributes Used in this Cell:
- **`df.boxplot(column, by)`**: Generates a box plot, perfect for identifying the Interquartile Range and statistical outliers.
- **`plt.subplots(nrows, ncols)`**: The Object-Oriented foundation of Matplotlib. Creates a Figure canvas and a grid of Axes (plots).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Group prices by Cut quality
diamonds.boxplot(column='price', by='cut', ax=ax, grid=False, sym='r.') # sym='r.' makes outliers red dots

ax.set_title("Price Spread by Diamond Cut")
plt.suptitle('') # Removes the automatic Pandas title
ax.set_ylabel("Price ($)")
plt.show()
```

## Part 3: Advanced Seaborn

Pandas plotting is great, but `seaborn` requires 90% less code for complex statistical charts.

### 7. Violin Plots (Box Plot + KDE)

**Intuition:** Box plots hide the actual shape of the distribution. A Violin plot is a box plot with a KDE density curve mirrored on both sides. It shows outliers *and* where the data is clumped.


### 🔍 Functions & Attributes Used in this Cell:
- **`sns.violinplot()`**: A Seaborn plot that combines a boxplot with a KDE to show both outliers and data density.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
plt.figure(figsize=(12, 6))
# inner='quartile' draws lines for the 25%, 50%, and 75% marks inside the violin
sns.violinplot(data=diamonds, x='cut', y='price', palette='muted', inner='quartile')
plt.title("Violin Plot: Density and Spread")
plt.show()
```

### 8. PairPlots (The Exploratory Nuke)

**Intuition:** When you get a new dataset, you want to see how *every numeric column* correlates with *every other numeric column*.

`sns.pairplot` draws a massive matrix of scatter plots and histograms automatically. 
*(Warning: Running this on 54k rows takes a minute, so we will sample 1000 rows first)*.


### 🔍 Functions & Attributes Used in this Cell:
- **`sns.pairplot()`**: A Seaborn nuke. Plots every numeric column against every other numeric column instantly.
- **`.sample(n, frac)`**: Returns a random sample of rows. `n` specifies exact count, `frac` specifies a percentage. Use `random_state` for reproducibility.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Sample 1000 rows for speed, and select only 4 columns to avoid a 10x10 grid
sample_df = diamonds[['carat', 'depth', 'table', 'price', 'cut']].sample(1000, random_state=42)

# hue='cut' colors the dots based on their category!
sns.pairplot(sample_df, hue='cut', palette='husl', corner=True) # corner=True removes duplicate mirrored plots
plt.show()
```

### 9. FacetGrids (Trellis Plots)

**Intuition:** What if you want to draw a histogram of `price`, but you want a separate, mini-histogram for EVERY SINGLE `color` category?

You *could* write a `for` loop and create 7 subplots manually... or you could use a FacetGrid.


### 🔍 Functions & Attributes Used in this Cell:
- **`sns.FacetGrid()`**: A Seaborn Trellis plot. It automatically creates dozens of subplots based on categorical variables.
- **`.map(arg)`**: Maps values of a Series using an input correspondence (a dictionary or a function).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Create a grid where every row is a 'color'
grid = sns.FacetGrid(diamonds, col='color', col_wrap=4, height=3)

# Tell the grid to draw a histogram of 'price' in every single box
grid.map(sns.histplot, 'price', bins=30, color='darkgreen')

grid.fig.suptitle("Price Distribution Split by Color Category", y=1.05)
plt.show()
```
