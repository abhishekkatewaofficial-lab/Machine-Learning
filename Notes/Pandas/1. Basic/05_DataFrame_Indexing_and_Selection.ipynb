# DataFrame Selection: The Deep Dive

Selecting specific rows and columns is the bread-and-butter of Pandas. We will use `.loc`, `.iloc`, `.filter()`, and `.select_dtypes()`.


## 1. Setup Data


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

data = {
    'Math_Score': [85, 90, 78, 92],
    'Physics_Score': [88, 95, 80, 89],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Passed': [True, True, False, True]
}
df = pd.DataFrame(data, index=['ID1', 'ID2', 'ID3', 'ID4'])
display(df)
```

## 2. Advanced Column Selection (`.filter()` and `.select_dtypes()`)

You already know `df['Name']` selects one column and `df[['Name', 'Passed']]` selects two. But what if you have 1,000 columns?


### 🔍 Functions & Attributes Used in this Cell:
- **`.select_dtypes(include, exclude)`**: Subsets columns based on their data type. Perfect for isolating numeric vs categorical data.
- **`.filter(like, items, regex)`**: Subsets rows or columns based on their labels. `like='Score'` keeps columns containing the word Score.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# 1. Select columns dynamically using .filter()
# Let's get all columns that contain the word 'Score'
display(df.filter(like='Score'))

# 2. Select columns by data type using .select_dtypes()
# Let's get only the numeric columns
display(df.select_dtypes(include=['int64', 'float64']))

# Let's get everything EXCEPT the boolean column
display(df.select_dtypes(exclude=['bool']))
```

**Output Explanation:**
These methods are life-savers for Machine Learning, where you often need to separate numeric features from categorical text features.


## 3. Row and Column Selection with `.loc`

`.loc[rows, columns]` uses labels. Both rows and columns are INCLUSIVE when slicing.


### 🔍 Functions & Attributes Used in this Cell:
- **`.loc[]`**: Label-based indexer. `.loc['RowName', 'ColName']`. **Inclusive** of the end bound when slicing.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Select ID2 to ID4, and Math_Score to Name ---")
display(df.loc['ID2':'ID4', 'Math_Score':'Name'])
```

## 4. Row and Column Selection with `.iloc`

`.iloc[row_positions, col_positions]` uses integers. It is EXCLUSIVE of the end bound.


### 🔍 Functions & Attributes Used in this Cell:
- **`.iloc[]`**: Integer-position based indexer. `.iloc[0, 1]`. **Exclusive** of the end bound when slicing.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
print("--- Select first two rows, first two columns ---")
display(df.iloc[0:2, 0:2])

print("\n--- Select the absolute last row and last column (scalar) ---")
print(df.iloc[-1, -1])
```

## 5. Modifying Data (`.pop()` and `.insert()`)

You can extract a column completely (removing it from the DataFrame), or insert a new column at a specific location.


### 🔍 Functions & Attributes Used in this Cell:
- **`.pop(item)`**: Extracts a column as a Series AND completely deletes it from the original DataFrame simultaneously.
- **`.insert(loc, column, value)`**: Inserts a new column at a specific integer location, shifting other columns to the right.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# .pop() extracts the column AND deletes it from the original DataFrame
names = df.pop('Name')
print("Popped Column:")
display(names)

# .insert(location, column_name, values)
df.insert(0, 'First_Name', names) # Insert at position 0 (the far left)
print("\nDataFrame after pop and insert:")
display(df)
```
