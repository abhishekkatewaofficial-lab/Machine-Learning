## 1. Setup Messy Data


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Product': [' Apple ', 'BANANA', 'Orange', 'mango', 'Apple'],
    'Price': [1.2, 0.5, 0.8, 1.5, 1.2],
    'Qty': [100, 200, 150, np.nan, 100]
})
```

## 2. The Chained Pipeline

We will clean strings, fill NaNs, add a column, group, and sort all in one continuous flow.


### 🔍 Functions & Attributes Used in this Cell:
- **`.agg(func)`**: Applies one or multiple aggregation functions (mean, sum, max) to groups. Can use Named Aggregations for clean outputs.
- **`.assign(**kwargs)`**: Creates multiple new columns simultaneously. Excellent for method chaining.
- **`.fillna(value, method)`**: Replaces missing values with a specified scalar value, or uses a method like `'ffill'` (forward fill).
- **`.groupby(by)`**: Splits data into groups based on some criteria, allowing you to apply functions to each group independently.
- **`.sort_values(by, ascending, na_position)`**: Sorts the dataframe by one or more columns.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
clean_summary = (
    df
    # 1. Clean strings
    .assign(Product=lambda x: x['Product'].str.strip().str.title())
    # 2. Fill missing quantities with median
    .assign(Qty=lambda x: x['Qty'].fillna(x['Qty'].median()))
    # 3. Create Total Revenue
    .assign(Revenue=lambda x: x['Price'] * x['Qty'])
    # 4. Group by Product and sum
    .groupby('Product', as_index=False)
    .agg({'Revenue': 'sum', 'Qty': 'sum'})
    # 5. Sort by Revenue descending
    .sort_values('Revenue', ascending=False)
)
display(clean_summary)
```

> **Interview Tip:** Code written like this is extremely easy to read, test, and debug. You can comment out a specific `.assign()` line to see how it affects the pipeline without breaking variable references.
