# Pro-Level: Memory Optimization

Kaggle datasets are often 5GB+. If you have 16GB of RAM, basic Pandas will crash. You MUST downcast.


## 1. Generating a Large Dataset


### 🔍 Functions & Attributes Used in this Cell:
- **`.info()`**: Prints a concise summary of the DataFrame including null counts and data types. Pass `memory_usage='deep'` for exact RAM usage.
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'ID': np.random.randint(0, 100, size=1000000), # int64 by default
    'Value': np.random.rand(1000000), # float64 by default
    'Status': np.random.choice(['A', 'B', 'C'], size=1000000) # object by default
})
print("Original Memory:")
df.info(memory_usage='deep')
```

## 2. Downcasting Numerics

Pandas uses 64-bit integers (`int64`). `int64` can store numbers up to 9 quintillion. Our ID only goes up to 100. We only need `int8` (stores up to 127).


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.to_numeric(arg, downcast)`**: Converts an argument to a numeric type, optionally downcasting it to the smallest possible integer/float type to save memory.
- **`.info()`**: Prints a concise summary of the DataFrame including null counts and data types. Pass `memory_usage='deep'` for exact RAM usage.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df['ID'] = pd.to_numeric(df['ID'], downcast='integer') # automatically picks int8
df['Value'] = pd.to_numeric(df['Value'], downcast='float') # float32

print("After Downcasting Numerics:")
df.info(memory_usage='deep')
```

## 3. Categoricals

Storing 'A', 'B', 'C' a million times as strings is a massive waste of RAM.


### 🔍 Functions & Attributes Used in this Cell:
- **`.astype(dtype)`**: Casts a pandas object to a specified data type (e.g., converting float64 to int32 for memory savings).
- **`.info()`**: Prints a concise summary of the DataFrame including null counts and data types. Pass `memory_usage='deep'` for exact RAM usage.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df['Status'] = df['Status'].astype('category')

print("\nFully Optimized Memory:")
df.info(memory_usage='deep')
```

**Output Explanation:**
Memory usage dropped from ~65MB to ~6MB. A 90% reduction! This is how you win Kaggle.
