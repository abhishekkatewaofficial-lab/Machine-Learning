## 1. Advanced CSV Reading

Let's say you have a 10GB CSV. You can't read it normally. You must read it in chunks.


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.read_csv(filepath, chunksize)`**: Loads a CSV file into a DataFrame. Using `chunksize` reads massive files in batches.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

# Write a dummy CSV first
with open('huge.csv', 'w') as f:
    f.write("date,user,action\n2023-01-01,Alice,login\n2023-01-02,Bob,logout")

# Read in chunks of 1 row at a time (Imagine this was 1,000,000 rows at a time)
chunk_iterator = pd.read_csv('huge.csv', chunksize=1, parse_dates=['date'])

for chunk in chunk_iterator:
    print("Processing Chunk:")
    display(chunk)
    # Here you would process the chunk and append to a database, avoiding RAM crashes.
```

## 2. Parquet (The Future of Data Storage)

CSV is slow and takes up a lot of space. Parquet is compressed and keeps data types (no more parsing dates!).


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.read_csv(filepath, chunksize)`**: Loads a CSV file into a DataFrame. Using `chunksize` reads massive files in batches.
- **`pd.read_parquet(path)`**: Reads a Parquet file rapidly.
- **`.to_parquet(path)`**: Saves the DataFrame as a compressed Parquet file (preserves data types, unlike CSV).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
df = pd.read_csv('huge.csv')
# This requires pyarrow or fastparquet to be installed
try:
    df.to_parquet('data.parquet')
    print("Saved as Parquet!")
    
    # Read it back
    df_pq = pd.read_parquet('data.parquet')
    display(df_pq)
except ImportError:
    print("Please pip install pyarrow to run parquet commands.")
```
