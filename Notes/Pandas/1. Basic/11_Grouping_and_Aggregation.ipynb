# Grouping & Aggregation: The Split-Apply-Combine Strategy

This is the most powerful tool in Pandas. We will cover `.groupby()`, `.agg()`, named aggregations, `.transform()`, and `.filter()`.


## 1. Setup Realistic HR Data


### 🔍 Functions & Attributes Used in this Cell:
- **`pd.DataFrame()`**: Creates a 2-dimensional table. Basically a dictionary of Series sharing an index.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
import pandas as pd

df = pd.DataFrame({
    'Dept': ['IT', 'HR', 'IT', 'Sales', 'HR', 'Sales', 'IT'],
    'Role': ['Dev', 'Recruiter', 'Manager', 'Rep', 'Manager', 'Rep', 'Dev'],
    'Salary': [80000, 60000, 120000, 75000, 90000, 80000, 85000],
    'Bonus': [5000, 2000, 15000, 10000, 5000, 12000, 6000]
})
display(df)
```

## 2. Advanced `.agg()` with Named Aggregations

We can apply different functions to different columns, and explicitly name the output columns at the exact same time.


### 🔍 Functions & Attributes Used in this Cell:
- **`.agg(func)`**: Applies one or multiple aggregation functions (mean, sum, max) to groups. Can use Named Aggregations for clean outputs.
- **`.groupby(by)`**: Splits data into groups based on some criteria, allowing you to apply functions to each group independently.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
summary = df.groupby('Dept').agg(
    Avg_Salary=('Salary', 'mean'),
    Max_Bonus=('Bonus', 'max'),
    Employee_Count=('Role', 'count')
)
display(summary)
```

## 3. Grouping by Multiple Columns

You can group by a list of columns. This naturally creates a MultiIndex.


### 🔍 Functions & Attributes Used in this Cell:
- **`.groupby(by)`**: Splits data into groups based on some criteria, allowing you to apply functions to each group independently.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
multi_group = df.groupby(['Dept', 'Role'])['Salary'].mean().to_frame()
display(multi_group)
```

## 4. The `.transform()` Method (Crucial for Machine Learning)

`.agg()` shrinks the DataFrame. What if you want to calculate the department average, but attach it to EVERY original row?


### 🔍 Functions & Attributes Used in this Cell:
- **`.transform(func)`**: Applies a function to groups but broadcasts the result back to the original size of the DataFrame (crucial for Feature Engineering).
- **`.groupby(by)`**: Splits data into groups based on some criteria, allowing you to apply functions to each group independently.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# This calculates the mean salary per department, and broadcasts it back to the original size
df['Dept_Avg_Salary'] = df.groupby('Dept')['Salary'].transform('mean')

# Now we can calculate if an employee is paid above their department average!
df['Above_Dept_Avg'] = df['Salary'] > df['Dept_Avg_Salary']
display(df)
```

**Output Explanation:**
This is called "Feature Engineering". We couldn't have done this easily with `.agg()` because `.agg()` returns 3 rows, but we need 7 rows to match our original DataFrame.


## 5. The `.filter()` Method

Use `.filter()` to drop entire groups that don't meet a condition.


### 🔍 Functions & Attributes Used in this Cell:
- **`.groupby(by)`**: Splits data into groups based on some criteria, allowing you to apply functions to each group independently.
- **`.filter(like, items, regex)`**: Subsets rows or columns based on their labels. `like='Score'` keeps columns containing the word Score.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
# Only keep Departments that have more than 2 employees
display(df.groupby('Dept').filter(lambda x: len(x) > 2))
```
