## 1. Load and Inspect

```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

titanic = sns.load_dataset('titanic')
display(titanic.info())
```

## 2. Feature Engineering

Let's create new features based on family size.


### 🔍 Functions & Attributes Used in this Cell:
- **`.head(n)`**: Returns the first `n` rows of the DataFrame. Essential for peeking at data without crashing your screen.
- **`.astype(dtype)`**: Casts a pandas object to a specified data type (e.g., converting float64 to int32 for memory savings).

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
titanic['Family_Size'] = titanic['sibsp'] + titanic['parch'] + 1
titanic['Is_Alone'] = (titanic['Family_Size'] == 1).astype(int)

display(titanic[['sibsp', 'parch', 'Family_Size', 'Is_Alone']].head())
```

## 3. Multi-dimensional Analysis

What is the survival rate by Class, Gender, AND whether they were alone?


### 🔍 Functions & Attributes Used in this Cell:
- **`.pivot_table(index, columns, values, aggfunc)`**: Creates a spreadsheet-style pivot table.

### 🧠 Intuition & Deep Dive:
Every function in this block was explicitly chosen to handle edge cases or optimize performance. Read the docstrings above to understand the specific arguments used in the code block below!

```python
pivot = titanic.pivot_table(
    index=['pclass', 'sex'], 
    columns='Is_Alone', 
    values='survived', 
    aggfunc=['mean', 'count']
)
display(pivot)
```

**Output Explanation:**
We can see the average survival rate AND the absolute count of passengers in each bucket simultaneously!
