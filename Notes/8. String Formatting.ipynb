```python
person = {'name': 'Jenn','age': 23}

sentence = 'My name is ' + person['name'] + ' and I am ' + str(person['age']) + ' years old.'
print(sentence)
```

```python
person = {'name': 'Jenn','age': 23}
sentence = 'My name is {} and I am {} years old'.format (person['name'], person['age'])
print (sentence)
```

```python
person = {'name': 'Jenn','age': 23}
sentence = 'My name is {0} and I am {1} years old'.format (person['name'], person['age'])
print (sentence)
```

```python
person = {'name': 'Jenn','age': 23}
sentence = 'My name is {1} and I am {0} years old'.format (person['name'], person['age'])
print (sentence)
```

```python
person = {'name': 'Jenn','age': 23}
sentence = 'My name is {0[name]} and I am {1[name]} years old'.format (person, person)
print (sentence)
```

```python
sentence = 'My name is {name} and I am {age} years old.'.format(name='Jenn', age=23)
print(sentence)
```

```python
person = {'name': 'Jenn','age': 23}
sentence = 'My name is {name} and I am {age} years old'.format (**person)
print (sentence)
```
