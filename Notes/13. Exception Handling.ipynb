# 1. What is Exception Handling?
Exception Handling is a mechanism to handle runtime errors gracefully without terminating the program abruptly.
Without exception handling, a program crashes when an error occurs.


# Example
num = int(input("Enter a number: "))
print(10 / num)

# 2. Basic try-except

```python
#try:
    # Risky code
#except:
    # Code to handle exception
```

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)

except:
    print("An error occurred.")
```

Explanation
Code inside try is executed first.
If an exception occurs, Python immediately jumps to except.
Program continues instead of crashing

# 3. Catching Specific Exceptions

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print(result)

except ZeroDivisionError:
    print("Cannot divide by zero.")

except ValueError:
    print("Please enter a valid integer.")
```

Explanation
ZeroDivisionError occurs when dividing by zero.
ValueError occurs when conversion to integer fails.
Different exceptions can be handled differently.

# 4. Multiple Exceptions in One Block

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)

except (ValueError, ZeroDivisionError):
    print("Invalid input or division by zero.")
```

Explanation
Multiple exceptions can be grouped together.
Useful when handling them in the same way.

# 5. Getting Error Message

```python
try:
    num = int(input("Enter a number: "))
    print(10 / num)

except Exception as e:
    print("Error:", e)
```

Explanation
e stores the exception object.
Helps in debugging.
Exception is the parent class of most built-in exceptions.

# 6. try-except-else

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num

except ZeroDivisionError:
    print("Division by zero is not allowed.")

else:
    print("Result:", result)
```

#  Explanation
else executes only if no exception occurs.
Useful for code that should run after successful execution.

# 7. try-except-finally

```python
try:
    file = open("loda lassan.txt", "r")
    data = file.read()

except FileNotFoundError:
    print("File not found.")

finally:
    print("Execution completed.")
```

# Explanation
finally always executes.
Runs whether an exception occurs or not.
Commonly used for resource cleanup.

## 8. Complete Flow (try-except-else-finally)

```python
try:
    num = int(input("Enter a number: "))
    result = 100 / num

except ZeroDivisionError:
    print("Cannot divide by zero.")

except ValueError:
    print("Invalid input.")

else:
    print("Result:", result)

finally:
    print("Program ended.")
```

# Execution Flow
Execute try.
If exception occurs → matching except.
If no exception → else.
finally always runs.
